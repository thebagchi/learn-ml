# The Sum Rule of Probability

This notebook explains the **sum rule of probability**, which is used to calculate the probability of at least one of two events occurring ($P(A \cup B)$).


## The Sum Rule

For any two events, A and B, the probability that either A or B (or both) occurs is given by:

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

### Why do we subtract the intersection?
When we add $P(A)$ and $P(B)$, we are adding the probability of the intersection, $P(A \cap B)$, twice. The Venn diagram below shows that the intersection area is part of both circles. To correct for this double-counting, we must subtract it once.

### Special Case: Mutually Exclusive Events
If A and B are **mutually exclusive**, they cannot happen at the same time, so their intersection is empty ($P(A \cap B) = 0$). In this case, the sum rule simplifies to:

$P(A \cup B) = P(A) + P(B)$


```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Rectangle
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Diagram 1: General Case (Overlapping Events) ---
ax = axes[0]
ax.set_title('General Case: Overlapping Events', fontsize=12)
ax.add_patch(Rectangle((0, 0), 10, 6, fill=False, linewidth=2, edgecolor='black'))
ax.add_patch(Circle((4.1, 3), 2.0, facecolor='#90caf9', edgecolor='#1565c0', alpha=0.7, linewidth=2))
ax.add_patch(Circle((5.9, 3), 2.0, facecolor='#ffcc80', edgecolor='#ef6c00', alpha=0.7, linewidth=2))

ax.text(3.1, 4.9, 'A', fontsize=12, weight='bold')
ax.text(6.7, 4.9, 'B', fontsize=12, weight='bold')
ax.text(5.0, 3.0, 'A ∩ B', ha='center', va='center', fontsize=10, weight='bold')
ax.text(5.0, 0.5, 'The intersection is counted twice, so we subtract it once.', ha='center', fontsize=10)
ax.text(5.0, 0.1, '$P(A \\cup B) = P(A) + P(B) - P(A \\cap B)$', ha='center', fontsize=10, weight='bold')

ax.set_xlim(-0.2, 10.2)
ax.set_ylim(-0.2, 6.2)
ax.set_aspect('equal')
ax.axis('off')

# --- Diagram 2: Mutually Exclusive Events ---
ax = axes[1]
ax.set_title('Special Case: Mutually Exclusive Events', fontsize=12)
ax.add_patch(Rectangle((0, 0), 10, 6, fill=False, linewidth=2, edgecolor='black'))
ax.add_patch(Circle((3.2, 3), 1.7, facecolor='#90caf9', edgecolor='#1565c0', alpha=0.7, linewidth=2))
ax.add_patch(Circle((6.8, 3), 1.7, facecolor='#ffcc80', edgecolor='#ef6c00', alpha=0.7, linewidth=2))

ax.text(2.6, 4.6, 'A', fontsize=12, weight='bold')
ax.text(7.2, 4.6, 'B', fontsize=12, weight='bold')
ax.text(5.0, 0.5, 'No overlap, so $P(A \\cap B) = 0$.', ha='center', fontsize=10)
ax.text(5.0, 0.1, '$P(A \\cup B) = P(A) + P(B)$', ha='center', fontsize=10, weight='bold')

ax.set_xlim(-0.2, 10.2)
ax.set_ylim(-0.2, 6.2)
ax.set_aspect('equal')
ax.axis('off')

plt.tight_layout()
plt.show()
```

## Numerical Example

Let's verify the sum rule with a concrete example:
- $P(A) = 0.50$
- $P(B) = 0.40$
- $P(A \cap B) = 0.20$

Using the sum rule:

$P(A \cup B) = P(A) + P(B) - P(A \cap B) = 0.50 + 0.40 - 0.20 = 0.70$


```python
p_a = 0.50
p_b = 0.40
p_ab = 0.20  # P(A ∩ B)

p_union = p_a + p_b - p_ab

print(f"P(A)     = {p_a:.2f}")
print(f"P(B)     = {p_b:.2f}")
print(f"P(A ∩ B) = {p_ab:.2f}")
print(f"P(A ∪ B) = P(A) + P(B) - P(A ∩ B) = {p_a} + {p_b} - {p_ab} = {p_union:.2f}")
```
