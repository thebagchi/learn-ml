# Normal Distribution

The **Normal (Gaussian) distribution** is the most widely used continuous distribution. It arises naturally from the Central Limit Theorem — sums of many independent random variables converge to it regardless of the original distribution.

| Property | Value |
|---|---|
| **Parameters** | $\mu$ (location), $\sigma$ (scale, $\sigma > 0$) |
| **Support** | $(-\infty,\, +\infty)$ |
| **PDF** | $\displaystyle f(x) = \frac{1}{\sigma\sqrt{2\pi}} \exp\!\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)$ |
| **Mean** | $\mu$ |
| **Median** | $\mu$ |
| **Mode** | $\mu$ |
| **Std Dev** | $\sigma$ |
| **Skewness** | $0$ (symmetric) |
| **Kurtosis** | $3$ (mesokurtic) |
| **Parameter estimation** | $\hat{\mu} = \bar{x},\quad \hat{\sigma} = s$ (sample mean and std) |

## Visualising the Normal Distribution

### Probability Density Function (PDF)

The PDF describes the **relative likelihood** of the random variable taking a given value. For the Normal distribution:

$$f(x \mid \mu, \sigma) = \frac{1}{\sigma\sqrt{2\pi}} \exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

- The curve is **bell-shaped and symmetric** about $\mu$.
- A smaller $\sigma$ produces a **taller, narrower** peak; a larger $\sigma$ gives a **flatter, wider** curve.
- The total area under the PDF always equals $1$.

---

### Cumulative Distribution Function (CDF)

The CDF gives the **probability that the variable is less than or equal to** $x$:

$$F(x) = \int_{-\infty}^{x} f(t)\,dt = \frac{1}{2}\left[1 + \text{erf}\!\left(\frac{x - \mu}{\sigma\sqrt{2}}\right)\right]$$

- $F(x)$ is **S-shaped (sigmoid)**, ranging from $0$ to $1$.
- The inflection point (steepest rise) occurs at $x = \mu$.
- A larger $\sigma$ stretches the S-curve horizontally, reflecting more spread.

---

### Normal Q-Q Plot (Quantile–Quantile)

A Q-Q plot checks whether a sample follows a Normal distribution by comparing its **empirical quantiles** against the **theoretical quantiles** of $\mathcal{N}(0,1)$:

- Each point $(q_{\text{theory}},\, q_{\text{sample}})$ is plotted; a reference line $y = \sigma x + \mu$ is drawn.
- If the data are normally distributed, the points **fall closely along the straight line**.
- **Systematic deviations** reveal departures from normality: curved tails → skewness, heavy tails → high kurtosis.
- The correlation $r$ between the plotted pairs quantifies the goodness-of-fit; values close to $1$ indicate approximate normality.


```python
x = np.linspace(-5, 5, 400)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Normal Distribution', fontsize=12, fontweight='bold')

# PDF — vary σ
for sigma, color, label in [(0.5, '#C62828', 'σ=0.5'), (1, '#1565C0', 'σ=1 (std normal)'), (2, '#2E7D32', 'σ=2')]:
    axes[0].plot(x, stats.norm(0, sigma).pdf(x), color=color, linewidth=2, label=label)
axes[0].set_title('PDF — varying σ (μ=0)')
axes[0].set_xlabel('x'); axes[0].set_ylabel('Density')
axes[0].legend()

# CDF
for sigma, color, label in [(0.5, '#C62828', 'σ=0.5'), (1, '#1565C0', 'σ=1'), (2, '#2E7D32', 'σ=2')]:
    axes[1].plot(x, stats.norm(0, sigma).cdf(x), color=color, linewidth=2, label=label)
axes[1].set_title('CDF')
axes[1].set_xlabel('x'); axes[1].set_ylabel('F(x)')
axes[1].legend()

# Q-Q plot
np.random.seed(42)
sample = np.random.normal(2, 1.5, 200)
(osm, osr), (slope, intercept, r) = stats.probplot(sample, dist='norm')
axes[2].scatter(osm, osr, color='#42A5F5', s=15, alpha=0.7)
axes[2].plot([min(osm), max(osm)], [slope*min(osm)+intercept, slope*max(osm)+intercept],
             color='#E53935', linewidth=2, linestyle='--')
axes[2].set_title(f'Normal Q-Q (r={r:.4f})')
axes[2].set_xlabel('Theoretical quantiles'); axes[2].set_ylabel('Sample quantiles')

plt.tight_layout()
plt.show()

# 68-95-99.7 rule
print("68-95-99.7 rule (standard normal):")
for k, pct in [(1, 68.27), (2, 95.45), (3, 99.73)]:
    actual = (stats.norm.cdf(k) - stats.norm.cdf(-k)) * 100
    print(f"  P(|X| ≤ {k}σ) = {actual:.2f}%  (rule of thumb: {pct}%)")
```

```python
# Sample data plot — histogram + fitted PDF + rug plot
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Sample Data  (n=200,  μ=2,  σ=1.5)', fontsize=12, fontweight='bold')

mu_hat, sigma_hat = sample.mean(), sample.std(ddof=1)
x_fit = np.linspace(sample.min() - 1, sample.max() + 1, 300)

# ── Histogram + fitted PDF ───────────────────────────────────────────────────
axes[0].hist(sample, bins=25, density=True,
             color='#90CAF9', edgecolor='white', linewidth=0.6, label='Sample')
axes[0].plot(x_fit, stats.norm(mu_hat, sigma_hat).pdf(x_fit),
             color='#C62828', linewidth=2.5, label=f'Fitted PDF\nμ̂={mu_hat:.2f}, σ̂={sigma_hat:.2f}')
axes[0].axvline(mu_hat, color='#FF7043', linewidth=1.5, linestyle='--', label=f'Mean={mu_hat:.2f}')
axes[0].set_title('Histogram + Fitted PDF')
axes[0].set_xlabel('x'); axes[0].set_ylabel('Density')
axes[0].legend(fontsize=8)

# ── Empirical CDF vs theoretical CDF ────────────────────────────────────────
sorted_sample = np.sort(sample)
ecdf = np.arange(1, len(sorted_sample) + 1) / len(sorted_sample)
axes[1].step(sorted_sample, ecdf, color='#1565C0', linewidth=2, label='Empirical CDF')
axes[1].plot(x_fit, stats.norm(mu_hat, sigma_hat).cdf(x_fit),
             color='#C62828', linewidth=2, linestyle='--', label='Fitted CDF')
axes[1].set_title('Empirical vs Fitted CDF')
axes[1].set_xlabel('x'); axes[1].set_ylabel('F(x)')
axes[1].legend(fontsize=8)

# ── Box plot + strip (rug) ───────────────────────────────────────────────────
axes[2].boxplot(sample, vert=False, patch_artist=True, widths=0.4,
                boxprops=dict(facecolor='#BBDEFB', color='#1565C0'),
                medianprops=dict(color='#C62828', linewidth=2),
                whiskerprops=dict(color='#1565C0'),
                capprops=dict(color='#1565C0'),
                flierprops=dict(marker='o', color='#FF7043', markersize=4, alpha=0.6))
axes[2].scatter(sample, np.random.uniform(0.85, 1.15, len(sample)),
                color='#42A5F5', s=8, alpha=0.4)
axes[2].axvline(mu_hat,          color='#FF7043', linewidth=1.5, linestyle='--', label=f'Mean={mu_hat:.2f}')
axes[2].axvline(mu_hat + sigma_hat, color='#2E7D32', linewidth=1, linestyle=':', label=f'+1σ={mu_hat+sigma_hat:.2f}')
axes[2].axvline(mu_hat - sigma_hat, color='#2E7D32', linewidth=1, linestyle=':', label=f'-1σ={mu_hat-sigma_hat:.2f}')
axes[2].set_title('Box Plot + Strip')
axes[2].set_xlabel('x'); axes[2].set_yticks([])
axes[2].legend(fontsize=8)

plt.tight_layout()
plt.show()

print(f"Sample statistics:  n={len(sample)}  mean={mu_hat:.4f}  std={sigma_hat:.4f}"
      f"  skew={stats.skew(sample):.4f}  kurtosis={stats.kurtosis(sample):.4f}")
```

## Key Takeaways

| | |
|---|---|
| **Shape** | Symmetric bell curve — skewness $= 0$, kurtosis $= 3$ |
| **CLT** | Sample means of any distribution converge to Normal as $n \to \infty$ |
| **Estimation** | $\hat{\mu} = \bar{x}$ (MLE = MOM);  $\hat{\sigma} = s$ |
| **68-95-99.7** | $68\%$ within $\pm 1\sigma$,  $95\%$ within $\pm 2\sigma$,  $99.7\%$ within $\pm 3\sigma$ |
| **Standard form** | $\displaystyle Z = \frac{X - \mu}{\sigma} \sim \mathcal{N}(0,1)$ — enables all probability table lookups |
| **Outlier check** | $\lvert Z \rvert > 3$ is unusual (probability $0.3\%$) |
| **Q-Q plot** | Points on the diagonal $\Rightarrow$ data is normally distributed |
