# Dot Product of Vectors

This notebook demonstrates how to compute the dot product (also known as inner product or scalar product) of two vectors using PyTorch.

## Dot Product of Vectors

### Definition and Setup

**Let $\mathbf{x}$ and $\mathbf{y}$ be two vectors in the same space**, i.e., $\mathbf{x}, \mathbf{y} \in \mathbb{R}^d$

**Dot product between two vectors $\mathbf{x}$ and $\mathbf{y}$ is denoted as:**

$$\langle \mathbf{x}, \mathbf{y} \rangle = \mathbf{x}^T \mathbf{y}$$

### Mathematical Formula

**Multiplication of the transpose of the first vector $\mathbf{x}$ with the another vector $\mathbf{y}$:**

$$\langle \mathbf{x}, \mathbf{y} \rangle = \mathbf{x}^T \mathbf{y} = \sum_{i=1}^{d} x_i y_i = a$$

where:
- $\mathbf{x}^T$ is the **transpose** (row vector form) of $\mathbf{x}$
- The result is a **single scalar value** $a$

### Key Interpretation

**Dot product between the two vectors give a single value:**
- That single value is interpreted as: **how close are the two vectors in the space**
- It measures the **similarity** and **alignment** between the two vectors

### Geometric Interpretation

**For $\mathbf{x}, \mathbf{y} \in \mathbb{R}^2$ (2D vectors), geometrically:**

- **It is the length of the projection of vector $\mathbf{x}$ onto the vector $\mathbf{y}$**
- In other words, the dot product tells us: if we project vector $\mathbf{x}$ onto the direction of vector $\mathbf{y}$, how long is that projection?

$$\langle \mathbf{x}, \mathbf{y} \rangle = |\mathbf{x}| |\mathbf{y}| \cos(\theta)$$

where $\theta$ is the angle between the two vectors, and $|\mathbf{x}|$, $|\mathbf{y}|$ are their magnitudes.

### Expanded Representation

For explicit understanding, the dot product can be written as:

$$\mathbf{x}^T \mathbf{y} = \begin{bmatrix} x_1 & x_2 & \cdots & x_d \end{bmatrix} \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_d \end{bmatrix} = x_1 y_1 + x_2 y_2 + \cdots + x_d y_d$$

### Key Properties

- **Produces a scalar**: The dot product always results in a single number, not a vector

- **Commutative**: $\langle \mathbf{x}, \mathbf{y} \rangle = \langle \mathbf{y}, \mathbf{x} \rangle$

- **Linear**: $c(\langle \mathbf{x}, \mathbf{y} \rangle) = \langle c\mathbf{x}, \mathbf{y} \rangle = \langle \mathbf{x}, c\mathbf{y} \rangle$

- **Orthogonality**: If vectors are perpendicular, $\langle \mathbf{x}, \mathbf{y} \rangle = 0$

- **Parallel vectors**: Maximum dot product occurs when vectors point in the same direction

- **Opposite vectors**: Negative dot product when vectors point in opposite directions

```python
import torch
import numpy as np

# Create two vectors
u = torch.tensor([2.0, 3.0, 4.0])
v = torch.tensor([1.0, 2.0, 3.0])

print(f"u = {u}")
print(f"v = {v}")
```

```python
# Compute dot product manually (element-wise multiplication and sum)
dot_product_manual = torch.sum(u * v)
print(f"Dot product (manual): u · v = {dot_product_manual}")
print(f"Calculation: (2×1) + (3×2) + (4×3) = 2 + 6 + 12 = {dot_product_manual.item()}")
```

```python
# Using torch.dot() function
dot_product = torch.dot(u, v)
print(f"Dot product (torch.dot): u · v = {dot_product}")
print(f"Result is a scalar: {dot_product.shape}")
```

## Dot Product with Different Scenarios

Let's explore dot product with orthogonal vectors, parallel vectors, and other cases.

```python
# Orthogonal vectors (perpendicular, angle = 90°)
u_ortho = torch.tensor([1.0, 0.0, 0.0])
v_ortho = torch.tensor([0.0, 1.0, 0.0])
dot_ortho = torch.dot(u_ortho, v_ortho)
print(f"Orthogonal vectors: u · v = {dot_ortho}")
print(f"(Orthogonal vectors have dot product of 0)\n")

# Parallel vectors (same direction, angle = 0°)
u_para = torch.tensor([1.0, 2.0, 3.0])
v_para = torch.tensor([2.0, 4.0, 6.0])  # v_para = 2 * u_para
dot_para = torch.dot(u_para, v_para)
print(f"Parallel vectors: u · v = {dot_para}")
print(f"(Parallel vectors have maximum dot product)\n")

# Opposite vectors (opposite direction, angle = 180°)
u_opp = torch.tensor([1.0, 2.0, 3.0])
v_opp = torch.tensor([-1.0, -2.0, -3.0])  # v_opp = -u_opp
dot_opp = torch.dot(u_opp, v_opp)
print(f"Opposite vectors: u · v = {dot_opp}")
print(f"(Opposite vectors have minimum (negative) dot product)")
```

## Geometric Interpretation of Dot Product

The dot product can be interpreted geometrically. When two vectors have the same direction, their dot product is positive and maximum. When perpendicular, it's zero. When opposite, it's negative.

```python
import matplotlib.pyplot as plt
from iitm.libs.plot_vectors import plot_vectors

# Compute magnitudes (norms) of vectors
norm_u = torch.norm(u)
norm_v = torch.norm(v)
dot_prod = torch.dot(u, v)

# Compute angle using dot product formula: u · v = |u| |v| cos(θ)
cos_theta = dot_prod / (norm_u * norm_v)
theta_rad = torch.acos(cos_theta)
theta_deg = torch.rad2deg(theta_rad)

print(f"Magnitude of u: |u| = {norm_u:.4f}")
print(f"Magnitude of v: |v| = {norm_v:.4f}")
print(f"Dot product: u · v = {dot_prod:.4f}")
print(f"cos(θ) = {cos_theta:.4f}")
print(f"Angle between vectors: θ = {theta_deg:.2f}°")
```

## Commutative Property of Dot Product

The dot product is commutative: u · v = v · u

```python
dot_u_v = torch.dot(u, v)
dot_v_u = torch.dot(v, u)

print(f"u · v = {dot_u_v}")
print(f"v · u = {dot_v_u}")
print(f"Equal: {torch.equal(dot_u_v, dot_v_u)}")
```

## Dot Product with Different Data Types

PyTorch supports different data types for dot product computation.

```python
# Integer vectors
u_int = torch.tensor([1, 2, 3])
v_int = torch.tensor([4, 5, 6])
dot_int = torch.dot(u_int, v_int)
print(f"Integer vectors: u · v = {dot_int}  (dtype: {dot_int.dtype})")
print(f"Calculation: (1×4) + (2×5) + (3×6) = 4 + 10 + 18 = {dot_int.item()}\n")

# Float vectors
u_float = torch.tensor([1.5, 2.5, 3.5])
v_float = torch.tensor([2.0, 3.0, 4.0])
dot_float = torch.dot(u_float, v_float)
print(f"Float vectors: u · v = {dot_float}  (dtype: {dot_float.dtype})")
```

## Applications of Dot Product

The dot product has many important applications in linear algebra and machine learning:
- **Projection**: Finding how much one vector points in the direction of another
- **Similarity**: Measuring how similar two vectors are
- **Angle Calculation**: Determining the angle between vectors
- **Orthogonality Test**: Checking if vectors are perpendicular
- **Machine Learning**: Computing predictions in neural networks, similarity metrics, etc.
