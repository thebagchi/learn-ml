# Linear Dependence and Independence

Linear dependence and independence are fundamental concepts in linear algebra that describe the relationships between vectors. Understanding these concepts is essential for grasping the structure of vector spaces, solving linear systems, and understanding the dimensions of data in machine learning.

## Definition

### Linear Independence
A set of vectors $\{\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k\}$ is **linearly independent** if the only solution to:

$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \cdots + c_k\mathbf{v}_k = \mathbf{0}$$

is $c_1 = c_2 = \cdots = c_k = 0$ (the trivial solution).

### Linear Dependence
A set of vectors is **linearly dependent** if there exist scalars $c_1, c_2, \ldots, c_k$ (not all zero) such that:

$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \cdots + c_k\mathbf{v}_k = \mathbf{0}$$

In other words, at least one vector can be expressed as a linear combination of the others.

## Key Properties

- Any set of vectors containing the **zero vector** is linearly dependent
- A single **non-zero vector** is linearly independent
- Two vectors are linearly dependent if and only if one is a **scalar multiple** of the other
- If vectors $\mathbf{v}_1, \mathbf{v}_2, \ldots, \mathbf{v}_k$ are linearly independent and $\mathbf{v}_{k+1}$ is not in their span, then $\mathbf{v}_1, \ldots, \mathbf{v}_k, \mathbf{v}_{k+1}$ are linearly independent
- The **rank** of a matrix equals the maximum number of linearly independent row (or column) vectors

## Linear Independence Examples

```python
# Example 1: Two linearly independent vectors in 2D
v1 = torch.tensor([1.0, 0.0])
v2 = torch.tensor([0.0, 1.0])

print("Vector v1:", v1.numpy())
print("Vector v2:", v2.numpy())

# Create matrix with vectors as columns
A = torch.column_stack([v1, v2])
print("\nMatrix with v1, v2 as columns:")
print(A)

# Rank equals number of linearly independent vectors
rank = torch.linalg.matrix_rank(A)
print(f"\nRank: {rank}")
print(f"Number of vectors: 2")
print(f"Linearly independent: {rank == 2}")
```

```python
# Example 2: Three vectors in 2D (must be linearly dependent)
v1 = torch.tensor([1.0, 0.0])
v2 = torch.tensor([0.0, 1.0])
v3 = torch.tensor([2.0, 3.0])

print("Vector v1:", v1.numpy())
print("Vector v2:", v2.numpy())
print("Vector v3:", v3.numpy())

# Create matrix with vectors as columns
A = torch.column_stack([v1, v2, v3])
print("\nMatrix with v1, v2, v3 as columns:")
print(A)

rank = torch.linalg.matrix_rank(A)
print(f"\nRank: {rank}")
print(f"Number of vectors: 3")
print(f"Linearly dependent: {rank < 3}")
print(f"\nNote: v3 = 2*v1 + 3*v2")
print(f"Verification: 2*v1 + 3*v2 = {2*v1 + 3*v2}")
```

## Linear Dependence Examples

```python
# Example 3: Two vectors where one is a scalar multiple of the other
v1 = torch.tensor([1.0, 2.0])
v2 = torch.tensor([2.0, 4.0])  # v2 = 2 * v1

print("Vector v1:", v1.numpy())
print("Vector v2:", v2.numpy())
print(f"\nv2 = 2 * v1: {torch.allclose(v2, 2*v1)}")

# Create matrix with vectors as columns
A = torch.column_stack([v1, v2])
print("\nMatrix with v1, v2 as columns:")
print(A)

rank = torch.linalg.matrix_rank(A)
print(f"\nRank: {rank}")
print(f"Number of vectors: 2")
print(f"Linearly dependent: {rank < 2}")
```

```python
# Example 4: Three vectors with linear dependence
v1 = torch.tensor([1.0, 0.0, 0.0])
v2 = torch.tensor([0.0, 1.0, 0.0])
v3 = torch.tensor([1.0, 1.0, 0.0])  # v3 = v1 + v2

print("Vector v1:", v1.numpy())
print("Vector v2:", v2.numpy())
print("Vector v3:", v3.numpy())
print(f"\nv3 = v1 + v2: {torch.allclose(v3, v1 + v2)}")

# Create matrix with vectors as columns
A = torch.column_stack([v1, v2, v3])
print("\nMatrix with v1, v2, v3 as columns:")
print(A)

rank = torch.linalg.matrix_rank(A)
print(f"\nRank: {rank}")
print(f"Number of vectors: 3")
print(f"Linearly dependent: {rank < 3}")
```

## Testing Linear Independence

```python
# Method: Use the rank test
# Vectors v1, v2, ..., vk are linearly independent if and only if
# rank(A) = k, where A has the vectors as columns

def are_linearly_independent(vectors):
    """
    Check if a list of vectors are linearly independent.
    
    Args:
        vectors: list of torch tensors (each is a vector)
    
    Returns:
        True if linearly independent, False otherwise
    """
    if len(vectors) == 0:
        return True
    
    A = torch.column_stack(vectors)
    rank = torch.linalg.matrix_rank(A)
    return rank == len(vectors)

# Test 1: Linearly independent
v1 = torch.tensor([1.0, 0.0])
v2 = torch.tensor([0.0, 1.0])
result = are_linearly_independent([v1, v2])
print(f"Test 1 - [1,0], [0,1]: Independent = {result}")

# Test 2: Linearly dependent
v1 = torch.tensor([1.0, 2.0])
v2 = torch.tensor([2.0, 4.0])
result = are_linearly_independent([v1, v2])
print(f"Test 2 - [1,2], [2,4]: Independent = {result}")

# Test 3: Three linearly independent vectors in 3D
v1 = torch.tensor([1.0, 0.0, 0.0])
v2 = torch.tensor([0.0, 1.0, 0.0])
v3 = torch.tensor([0.0, 0.0, 1.0])
result = are_linearly_independent([v1, v2, v3])
print(f"Test 3 - Standard basis in 3D: Independent = {result}")
```

## Linear Combinations and Span

```python
# When vectors are linearly dependent, one can be written as
# a linear combination of others

# Solving for coefficients: A @ c = b
v1 = torch.tensor([1.0, 0.0])
v2 = torch.tensor([0.0, 1.0])
target = torch.tensor([3.0, 5.0])  # We want to express this as c1*v1 + c2*v2

print("v1:", v1.numpy())
print("v2:", v2.numpy())
print("Target:", target.numpy())

# Create matrix A with v1, v2 as columns
A = torch.column_stack([v1, v2])

# Solve A @ c = target
c = torch.linalg.solve(A, target)

print(f"\nCoefficients c: {c.numpy()}")
print(f"Verification: {c[0]:.1f}*v1 + {c[1]:.1f}*v2 = {(c[0]*v1 + c[1]*v2).numpy()}")
print(f"Matches target: {torch.allclose(c[0]*v1 + c[1]*v2, target)}")
```

## Basis and Linear Independence

```python
# A basis is a set of linearly independent vectors that span a space
# The number of vectors in a basis is the dimension of the space

# Standard basis in 3D
e1 = torch.tensor([1.0, 0.0, 0.0])
e2 = torch.tensor([0.0, 1.0, 0.0])
e3 = torch.tensor([0.0, 0.0, 1.0])

print("Standard Basis in R³:")
print(f"e1: {e1.numpy()}")
print(f"e2: {e2.numpy()}")
print(f"e3: {e3.numpy()}")

A = torch.column_stack([e1, e2, e3])
rank = torch.linalg.matrix_rank(A)
print(f"\nRank: {rank}")
print(f"Is a basis: {rank == 3}")
print(f"Dimension: 3")
```

```python
# Another basis in 3D (not the standard one)
v1 = torch.tensor([1.0, 1.0, 0.0])
v2 = torch.tensor([1.0, 0.0, 1.0])
v3 = torch.tensor([0.0, 1.0, 1.0])

print("Alternative Basis in R³:")
print(f"v1: {v1.numpy()}")
print(f"v2: {v2.numpy()}")
print(f"v3: {v3.numpy()}")

A = torch.column_stack([v1, v2, v3])
rank = torch.linalg.matrix_rank(A)
print(f"\nRank: {rank}")
print(f"Is a basis: {rank == 3}")
print(f"Dimension: 3")
```

## Rank and Linear Independence

```python
# The rank of a matrix equals the number of linearly independent
# rows (or columns)

print("Matrix A with rank 2:")
A = torch.tensor([[1, 2, 3],
                   [2, 4, 6],
                   [3, 5, 7]], dtype=torch.float32)
print(A)

rank = torch.linalg.matrix_rank(A)
print(f"\nRank: {rank}")
print("\nAnalysis:")
print("Row 1: [1, 2, 3]")
print("Row 2: [2, 4, 6] = 2 × Row 1 (dependent)")
print("Row 3: [3, 5, 7] (independent from Row 1)")
print("\nNumber of linearly independent rows: 2")
print(f"Matches rank: {rank == 2}")
```

```python
# Full rank matrix
print("Full Rank Matrix (3×3):")
B = torch.tensor([[1, 0, 0],
                   [0, 1, 0],
                   [0, 0, 1]], dtype=torch.float32)
print(B)

rank_B = torch.linalg.matrix_rank(B)
print(f"\nRank: {rank_B}")
print(f"Number of rows/columns: 3")
print(f"Is full rank: {rank_B == 3}")
print("All rows are linearly independent")
```

## Applications in Machine Learning

```python
# Example: Feature selection in machine learning
# Linearly dependent features are redundant

# Suppose we have features: Height, Weight, and Height_in_cm
# Height_in_cm = 100 * Height (linear dependence)

n_samples = 5
height_m = torch.tensor([[1.7], [1.8], [1.6], [1.75], [1.9]], dtype=torch.float32)
weight_kg = torch.tensor([[70.0], [75.0], [65.0], [72.0], [80.0]], dtype=torch.float32)
height_cm = 100 * height_m  # Linearly dependent

print("Feature matrix with redundancy:")
X_redundant = torch.column_stack([height_m.squeeze(), weight_kg.squeeze(), height_cm.squeeze()])
print(X_redundant)

rank_redundant = torch.linalg.matrix_rank(X_redundant)
n_features = 3
print(f"\nNumber of features: {n_features}")
print(f"Rank: {rank_redundant}")
print(f"Linearly dependent features detected: {rank_redundant < n_features}")

print("\nFeature matrix without redundancy:")
X_clean = torch.column_stack([height_m.squeeze(), weight_kg.squeeze()])
print(X_clean)

rank_clean = torch.linalg.matrix_rank(X_clean)
n_features_clean = 2
print(f"\nNumber of features: {n_features_clean}")
print(f"Rank: {rank_clean}")
print(f"All features linearly independent: {rank_clean == n_features_clean}")
```

```python
# Example: Solving linear systems
# A system has a unique solution if and only if the coefficient matrix
# has full rank

# System 1: Full rank (unique solution)
print("System 1: Full Rank Coefficient Matrix")
A1 = torch.tensor([[1, 2],
                    [3, 4]], dtype=torch.float32)
b1 = torch.tensor([5, 6], dtype=torch.float32)

print(f"A = {A1}")
print(f"b = {b1}")

rank_A1 = torch.linalg.matrix_rank(A1)
print(f"Rank(A): {rank_A1}")
print(f"Number of equations: 2")
print(f"Full rank (unique solution): {rank_A1 == 2}")

x1 = torch.linalg.solve(A1, b1)
print(f"\nSolution: {x1}")

# System 2: Rank deficient (dependent equations)
print("\n" + "="*50)
print("System 2: Rank Deficient Coefficient Matrix")
A2 = torch.tensor([[1, 2],
                    [2, 4]], dtype=torch.float32)  # Row 2 = 2 * Row 1
b2 = torch.tensor([5, 10], dtype=torch.float32)  # b2 also follows the relationship

print(f"A = {A2}")
print(f"b = {b2}")

rank_A2 = torch.linalg.matrix_rank(A2)
print(f"Rank(A): {rank_A2}")
print(f"Number of equations: 2")
print(f"Rank deficient (infinitely many solutions): {rank_A2 < 2}")
print(f"Note: The equations are dependent, so there are infinitely many solutions")
```

## Summary

| Concept | Definition | Relationship |
|---------|-----------|---------------|
| **Linear Independence** | Vectors where only trivial linear combination = 0 | All vectors are essential |
| **Linear Dependence** | Vectors where non-trivial linear combination = 0 | At least one vector is redundant |
| **Basis** | Linearly independent vectors that span a space | Exactly span the space |
| **Rank** | Number of linearly independent rows/columns | = Number of independent vectors |
| **Full Rank** | Rank = min(m,n) for m×n matrix | All rows/columns are independent |
| **Invertibility** | Matrix has an inverse | Square full rank matrix only |
