# Unit Vectors

This notebook demonstrates how to work with unit vectors using PyTorch.

## What is a Unit Vector?

A unit vector is a vector with a magnitude (or norm) of exactly 1. Unit vectors are important in mathematics and physics because they represent direction without magnitude. They are often used as basis vectors to represent any direction in space.

### Mathematical Definition

A unit vector $\hat{\mathbf{u}}$ is a vector with magnitude equal to 1:

$$|\hat{\mathbf{u}}| = 1$$

Any non-zero vector $\mathbf{v}$ can be normalized to create a unit vector in the same direction:

$$\hat{\mathbf{v}} = \frac{\mathbf{v}}{|\mathbf{v}|} = \frac{\mathbf{v}}{\sqrt{v_1^2 + v_2^2 + \cdots + v_d^2}}$$

**Key Properties**:

- A unit vector has **magnitude (L2 norm) equal to 1**

- Unit vectors **preserve direction** — normalizing a vector doesn't change its direction, only its magnitude

- Unit vectors form the **basis** of vector spaces

- The **dot product of a vector with a unit vector** gives the projection of that vector onto the unit vector's direction

- **Common standard unit vectors in 2D:** $\hat{\mathbf{i}} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$, $\hat{\mathbf{j}} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$

- **Common standard unit vectors in 3D:** $\hat{\mathbf{i}} = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}$, $\hat{\mathbf{j}} = \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix}$, $\hat{\mathbf{k}} = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}$

```python
import torch
import numpy as np

# Create a vector
v = torch.tensor([3.0, 4.0])

print(f"Original vector: v = {v}")
print(f"Components: v1 = {v[0]}, v2 = {v[1]}")
```

```python
# Calculate the magnitude (L2 norm) of the vector
magnitude = torch.norm(v)
print(f"Magnitude of v: |v| = {magnitude}")
print(f"Calculation: sqrt(3² + 4²) = sqrt(9 + 16) = sqrt(25) = {magnitude.item()}")
```

```python
# Normalize the vector to create a unit vector
u_hat = v / magnitude
print(f"Unit vector: û = v / |v| = {u_hat}")
print(f"û1 = {u_hat[0]:.4f}, û2 = {u_hat[1]:.4f}")
```

```python
# Verify that the result is a unit vector (magnitude = 1)
magnitude_unit = torch.norm(u_hat)
print(f"Magnitude of unit vector: |û| = {magnitude_unit}")
print(f"Is it a unit vector (magnitude = 1)? {torch.allclose(magnitude_unit, torch.tensor(1.0))}")
```

## Using torch.nn.functional.normalize()

PyTorch provides a convenient function to normalize vectors.

```python
import torch.nn.functional as F

# Using PyTorch's normalize function
u_hat_pytorch = F.normalize(v, p=2, dim=0)
print(f"Unit vector (using F.normalize): û = {u_hat_pytorch}")
print(f"Magnitude: |û| = {torch.norm(u_hat_pytorch)}")
print(f"Same as manual normalization? {torch.allclose(u_hat, u_hat_pytorch)}")
```

## Standard Unit Vectors (Basis Vectors)

Standard unit vectors are predefined unit vectors along each axis.

```python
# 2D standard unit vectors
i_hat = torch.tensor([1.0, 0.0])  # Unit vector along x-axis
j_hat = torch.tensor([0.0, 1.0])  # Unit vector along y-axis

print("2D Standard Unit Vectors (Basis Vectors):")
print(f"î (i-hat) = {i_hat},  |î| = {torch.norm(i_hat)}")
print(f"ĵ (j-hat) = {j_hat},  |ĵ| = {torch.norm(j_hat)}")

# 3D standard unit vectors
i_hat_3d = torch.tensor([1.0, 0.0, 0.0])
j_hat_3d = torch.tensor([0.0, 1.0, 0.0])
k_hat_3d = torch.tensor([0.0, 0.0, 1.0])

print("\n3D Standard Unit Vectors (Basis Vectors):")
print(f"î = {i_hat_3d},  |î| = {torch.norm(i_hat_3d)}")
print(f"ĵ = {j_hat_3d},  |ĵ| = {torch.norm(j_hat_3d)}")
print(f"k̂ = {k_hat_3d},  |k̂| = {torch.norm(k_hat_3d)}")
```

## Properties of Unit Vectors

Let's explore some important properties of unit vectors.

```python
# Property 1: Dot product of orthogonal unit vectors is 0
dot_product_ortho = torch.dot(i_hat, j_hat)
print(f"Orthogonal unit vectors: î · ĵ = {dot_product_ortho}")
print(f"(Unit vectors along perpendicular axes have dot product of 0)\n")

# Property 2: Dot product of a unit vector with itself is 1
dot_product_self = torch.dot(i_hat, i_hat)
print(f"Unit vector with itself: î · î = {dot_product_self}")
print(f"(This equals |î|² = 1² = 1)\n")

# Property 3: Dot product of any vector with a unit vector gives the projection
v_3d = torch.tensor([3.0, 4.0, 5.0])
projection_x = torch.dot(v_3d, i_hat_3d)
print(f"Vector: v = {v_3d}")
print(f"Projection onto î: v · î = {projection_x}")
print(f"(This gives the x-component of the vector)")
```

## Normalizing Vectors of Different Dimensions

Unit vectors can be created for vectors of any dimension.

```python
# Different dimensional vectors
v_1d = torch.tensor([5.0])
v_2d = torch.tensor([3.0, 4.0])
v_3d = torch.tensor([1.0, 2.0, 2.0])
v_4d = torch.tensor([1.0, 2.0, 3.0, 4.0])

# Normalize each vector
u_1d = F.normalize(v_1d, p=2, dim=0)
u_2d = F.normalize(v_2d, p=2, dim=0)
u_3d = F.normalize(v_3d, p=2, dim=0)
u_4d = F.normalize(v_4d, p=2, dim=0)

print(f"1D: v = {v_1d} => û = {u_1d}, |û| = {torch.norm(u_1d):.4f}")
print(f"2D: v = {v_2d} => û = {u_2d}, |û| = {torch.norm(u_2d):.4f}")
print(f"3D: v = {v_3d} => û = {u_3d}, |û| = {torch.norm(u_3d):.4f}")
print(f"4D: v = {v_4d} => û = {u_4d}, |û| = {torch.norm(u_4d):.4f}")
```

## Batch Normalization of Multiple Vectors

PyTorch can efficiently normalize multiple vectors at once.

```python
# Create a batch of vectors (each column is a vector)
vectors = torch.tensor([
    [3.0, 5.0, 1.0],
    [4.0, 12.0, 2.0]
])

print(f"Original vectors (each column is a vector):")
print(vectors)
print(f"\nMagnitudes: {torch.norm(vectors, p=2, dim=0)}")

# Normalize each vector
unit_vectors = F.normalize(vectors, p=2, dim=0)

print(f"\nNormalized unit vectors:")
print(unit_vectors)
print(f"\nMagnitudes of unit vectors: {torch.norm(unit_vectors, p=2, dim=0)}")
```

## Applications of Unit Vectors

Unit vectors are fundamental in many areas:
- **Direction representation**: Specifying direction independently of magnitude
- **Basis vectors**: Forming orthonormal bases for vector spaces
- **Normalization**: Converting any vector to a comparable form
- **Machine Learning**: Normalizing input features and embeddings
- **Computer Graphics**: Representing normals to surfaces
- **Physics**: Representing unit directional vectors in force, velocity, etc.
- **Cosine Similarity**: Computing similarity between normalized vectors
