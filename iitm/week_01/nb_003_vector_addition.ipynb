# Vector Addition

This notebook demonstrates how to perform vector addition using PyTorch.

## What is Vector Addition?

Vector addition is the operation of adding two vectors element-wise. For two vectors of the same dimension, the result is a new vector where each element is the sum of the corresponding elements from the input vectors.

### Mathematical Definition

Let $\mathbf{x}_m, \mathbf{x}_n \in \mathbb{R}^d$ be two vectors in $d$-dimensional space:

$$\mathbf{x}_m = \begin{bmatrix} x_{m1} \\ x_{m2} \\ \vdots \\ x_{md} \end{bmatrix}, \quad \mathbf{x}_n = \begin{bmatrix} x_{n1} \\ x_{n2} \\ \vdots \\ x_{nd} \end{bmatrix}$$

**Vector Addition** is defined as: $\mathbf{y} = \mathbf{x}_m + \mathbf{x}_n$

$$\mathbf{x}_m + \mathbf{x}_n = \begin{bmatrix} x_{m1} + x_{n1} \\ x_{m2} + x_{n2} \\ \vdots \\ x_{md} + x_{nd} \end{bmatrix} = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_d \end{bmatrix}$$

**Key Properties**:

- Vector addition is an **element-wise operation** where corresponding elements from two vectors are added

- The result is always a **new vector** of the same dimension as the input vectors

- For vectors of dimension $d$: each element $y_i = x_{mi} + x_{ni}$ for $i = 1, 2, \ldots, d$

```python
import torch

# Create two vectors
v1 = torch.tensor([1.0, 2.0, 3.0])
v2 = torch.tensor([4.0, 5.0, 6.0])

print(f"v1 = {v1}")
print(f"v2 = {v2}")
```

```python
# Perform vector addition
result = v1 + v2
print(f"v1 + v2 = {result}")
```

```python
# Alternative way using torch.add()
result = torch.add(v1, v2)
print(f"torch.add(v1, v2) = {result}")
```

## In-place Addition

You can also perform in-place addition using the `+=` operator.

```python
# In-place addition
v3 = torch.tensor([1.0, 2.0, 3.0])
v4 = torch.tensor([4.0, 5.0, 6.0])
v3 += v4
print(f"v3 += v4  =>  {v3}")
```

## Vector Addition with Different Data Types

PyTorch supports different data types. Let's see an example with integers.

```python
# Integer vectors
v5 = torch.tensor([1, 2, 3])
v6 = torch.tensor([4, 5, 6])
result = v5 + v6
print(f"v5 + v6 = {result}  (dtype: {result.dtype})")
```
