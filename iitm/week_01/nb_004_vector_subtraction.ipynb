# Vector Subtraction

This notebook demonstrates how to perform vector subtraction using PyTorch.

## What is Vector Subtraction?

Vector subtraction is the operation of subtracting two vectors element-wise. For two vectors of the same dimension, the result is a new vector where each element is the difference of the corresponding elements from the input vectors.

### Mathematical Definition

Let $\mathbf{x}_m, \mathbf{x}_n \in \mathbb{R}^d$ be two vectors in $d$-dimensional space:

$$\mathbf{x}_m = \begin{bmatrix} x_{m1} \\ x_{m2} \\ \vdots \\ x_{md} \end{bmatrix}, \quad \mathbf{x}_n = \begin{bmatrix} x_{n1} \\ x_{n2} \\ \vdots \\ x_{nd} \end{bmatrix}$$

**Vector Subtraction** is defined as: $\mathbf{y} = \mathbf{x}_m - \mathbf{x}_n$

$$\mathbf{x}_m - \mathbf{x}_n = \begin{bmatrix} x_{m1} - x_{n1} \\ x_{m2} - x_{n2} \\ \vdots \\ x_{md} - x_{nd} \end{bmatrix} = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_d \end{bmatrix}$$

**Key Properties**:

- Vector subtraction is an **element-wise operation** where corresponding elements from two vectors are subtracted

- The result is always a **new vector** of the same dimension as the input vectors

- For vectors of dimension $d$: each element $y_i = x_{mi} - x_{ni}$ for $i = 1, 2, \ldots, d$

- Unlike addition, subtraction is **not commutative**: $\mathbf{x}_m - \mathbf{x}_n \neq \mathbf{x}_n - \mathbf{x}_m$ (in general)

```python
import torch

# Create two vectors
v1 = torch.tensor([5.0, 7.0, 9.0])
v2 = torch.tensor([1.0, 2.0, 3.0])

print(f"v1 = {v1}")
print(f"v2 = {v2}")
```

```python
# Perform vector subtraction
result = v1 - v2
print(f"v1 - v2 = {result}")
```

## Alternative Method: torch.sub()

PyTorch provides the `torch.sub()` function as an alternative way to perform vector subtraction.

```python
# Alternative way using torch.sub()
result = torch.sub(v1, v2)
print(f"torch.sub(v1, v2) = {result}")
```

```python
# Non-commutativity of subtraction: v2 - v1
result_reversed = v2 - v1
print(f"v2 - v1 = {result_reversed}")
print(f"Notice: v1 - v2 = {result}")
print(f"And: -(v1 - v2) = {-result} (which equals v2 - v1)")
```

## In-place Subtraction

You can also perform in-place subtraction using the `-=` operator.

```python
# In-place subtraction
v3 = torch.tensor([10.0, 15.0, 20.0])
v4 = torch.tensor([2.0, 3.0, 4.0])
v3 -= v4
print(f"v3 -= v4  =>  {v3}")
```

## Vector Subtraction with Different Data Types

PyTorch supports different data types. Let's see an example with integers.

```python
# Integer vectors
v5 = torch.tensor([10, 20, 30])
v6 = torch.tensor([3, 5, 7])
result = v5 - v6
print(f"v5 - v6 = {result}  (dtype: {result.dtype})")
```
