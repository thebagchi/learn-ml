## Geometrical Interpretation of Vectors

- A vector $\mathbf{x}$ is a point in $d$-dimensional space
- Geometrically, a vector is considered as a directed line from the origin to the point corresponding to the vector $\mathbf{x}$


```python
# Plotting the vectors [1, 2] and [3, 2] using our custom function

# Create the vectors using PyTorch
x1 = torch.tensor([1, 2])
x2 = torch.tensor([3, 2])

print("Vector x1:", x1)
print("Vector x2:", x2)

# Plot both vectors
plot_vectors([x1, x2], labels=['x₁ = [1, 2]ᵀ', 'x₂ = [3, 2]ᵀ'])
```
