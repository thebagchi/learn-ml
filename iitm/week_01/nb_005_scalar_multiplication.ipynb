# Scalar Multiplication

This notebook demonstrates how to perform scalar multiplication (also known as scalar product) on vectors using PyTorch.

## What is Scalar Multiplication?

Scalar multiplication is the operation of multiplying each element of a vector by a scalar (a single number). The result is a new vector with each component scaled by the given scalar value.

### Mathematical Definition

Let $\mathbf{x} \in \mathbb{R}^d$ be a vector in $d$-dimensional space and let $c \in \mathbb{R}$ be a scalar:

$$\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_d \end{bmatrix}$$

**Scalar Multiplication** is defined as: $\mathbf{y} = c \cdot \mathbf{x}$

$$c \cdot \mathbf{x} = \begin{bmatrix} c \cdot x_1 \\ c \cdot x_2 \\ \vdots \\ c \cdot x_d \end{bmatrix} = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_d \end{bmatrix}$$

**Key Properties**:

- Scalar multiplication is an **element-wise operation** where each component of the vector is multiplied by the scalar

- The result is a **new vector** of the same dimension as the input vector

- For vectors of dimension $d$: each element $y_i = c \cdot x_i$ for $i = 1, 2, \ldots, d$

- Scalar multiplication is **commutative**: $c \cdot \mathbf{x} = \mathbf{x} \cdot c$

- **Geometric interpretation**: The resulting vector is scaled by a factor of $|c|$, and if $c < 0$, the vector is also reversed in direction

```python
import torch

# Create a vector
v = torch.tensor([2.0, 3.0, 4.0])
scalar = 3.0

print(f"v = {v}")
print(f"scalar = {scalar}")
```

```python
# Perform scalar multiplication
result = scalar * v
print(f"{scalar} * v = {result}")
```

## Alternative Method: torch.mul()

PyTorch provides the `torch.mul()` function as an alternative way to perform scalar multiplication.

```python
# Alternative way using torch.mul()
result = torch.mul(v, scalar)
print(f"torch.mul(v, {scalar}) = {result}")
```

```python
# Commutativity of scalar multiplication
result_alt = v * scalar
print(f"v * {scalar} = {result_alt}")
print(f"Notice: {scalar} * v = {result}")
print(f"And: v * {scalar} = {result_alt}")
print(f"They are equal: {torch.equal(result, result_alt)}")
```

## Scalar Multiplication with Negative Scalars

When multiplying by a negative scalar, the vector is reversed in direction (scaled in the opposite direction).

```python
# Scalar multiplication with negative scalar
negative_scalar = -2.0
result_negative = negative_scalar * v
print(f"{negative_scalar} * v = {result_negative}")
```

## In-place Scalar Multiplication

You can also perform in-place scalar multiplication using the `*=` operator.

```python
# In-place scalar multiplication
v2 = torch.tensor([1.0, 2.0, 3.0])
v2 *= 5.0
print(f"v2 *= 5.0  =>  {v2}")
```

## Scalar Multiplication with Different Data Types

PyTorch supports different data types. Let's see examples with integers and floating-point numbers.

```python
# Integer vector with integer scalar
v_int = torch.tensor([2, 3, 4])
scalar_int = 4
result_int = scalar_int * v_int
print(f"Integer: {scalar_int} * {v_int} = {result_int}  (dtype: {result_int.dtype})")

# Float vector with float scalar
v_float = torch.tensor([1.5, 2.5, 3.5])
scalar_float = 2.5
result_float = scalar_float * v_float
print(f"Float: {scalar_float} * {v_float} = {result_float}  (dtype: {result_float.dtype})")
```
