# Rules of Differentiation: Quotient Rule

## Key Concepts

- Let $f(x)$ and $g(x)$ are functions of a single scalar variable $x$

## Quotient Rule

- Let $\frac{f(x)}{g(x)}$ be a function,

  $\dfrac{d}{dx}\left(\dfrac{f(x)}{g(x)}\right) = \dfrac{g(x) \dfrac{df(x)}{dx} - f(x) \dfrac{dg(x)}{dx}}{(g(x))^2}$

## Example

Let $f(x) = \frac{x^3 - x^2 + 1}{x}$

$\dfrac{d(f(x))}{dx} = \dfrac{d}{dx}\left(\dfrac{x^3 - x^2 + 1}{x}\right)$

$\dfrac{d(f(x))}{dx} = \dfrac{x \dfrac{d(x^3-x^2+1)}{dx} - (x^3-x^2+1) \dfrac{d(x)}{dx}}{x^2}$

$\dfrac{d(f(x))}{dx} = \dfrac{x(3x^2 - 2x) - (x^3 - x^2 + 1) \cdot 1}{x^2}$

$\dfrac{d(f(x))}{dx} = \dfrac{3x^3 - 2x^2 - x^3 + x^2 - 1}{x^2}$

$\dfrac{d(f(x))}{dx} = \dfrac{2x^3 - x^2 - 1}{x^2}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values (avoiding x=0 where function is undefined)
x_neg = np.linspace(-5, -0.1, 80)
x_pos = np.linspace(0.1, 5, 80)

# Define individual functions
def numerator(x):
    return x**3 - x**2 + 1

def denominator(x):
    return x

# Quotient function
def f(x):
    return numerator(x) / denominator(x)

# Derivatives of individual functions
def numerator_prime(x):
    return 3*x**2 - 2*x

def denominator_prime(x):
    return np.ones_like(x)  # derivative of x is 1

# Quotient derivative using quotient rule
def f_prime(x):
    return (denominator(x) * numerator_prime(x) - numerator(x) * denominator_prime(x)) / (denominator(x)**2)

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Numerator and Denominator
ax1 = axes[0, 0]
x_plot = np.linspace(-5, 5, 100)
ax1.plot(x_plot, numerator(x_plot), 'b-', linewidth=2, label='Numerator: $x^3 - x^2 + 1$', alpha=0.7)
ax1.plot(x_plot, denominator(x_plot), 'g-', linewidth=2, label='Denominator: $x$', alpha=0.7)
ax1.axvline(x=0, color='k', linestyle='--', linewidth=1, alpha=0.5)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('Numerator and Denominator', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1.legend(fontsize=11, loc='best')
ax1.set_ylim(-10, 10)

# Plot 2: Quotient function
ax2 = axes[0, 1]
ax2.plot(x_neg, f(x_neg), 'r-', linewidth=2.5, label='$f(x) = \\frac{x^3 - x^2 + 1}{x}$')
ax2.plot(x_pos, f(x_pos), 'r-', linewidth=2.5)
ax2.axvline(x=0, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Vertical asymptote at x=0')
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('f(x)', fontsize=12)
ax2.set_title('Quotient Function', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax2.legend(fontsize=11, loc='best')
ax2.set_ylim(-20, 20)

# Plot 3: Component derivatives
ax3 = axes[1, 0]
x_plot2 = np.linspace(-5, 5, 100)
ax3.plot(x_plot2, numerator_prime(x_plot2), 'b--', linewidth=2, label="Numerator': $3x^2 - 2x$", alpha=0.7)
ax3.plot(x_plot2, denominator_prime(x_plot2), 'g--', linewidth=2, label="Denominator': $1$", alpha=0.7)
ax3.set_xlabel('x', fontsize=12)
ax3.set_ylabel("y'", fontsize=12)
ax3.set_title('Component Derivatives', fontsize=13, fontweight='bold')
ax3.grid(True, alpha=0.3)
ax3.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax3.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax3.legend(fontsize=11, loc='best')

# Plot 4: Quotient derivative
ax4 = axes[1, 1]
ax4.plot(x_neg, f_prime(x_neg), 'r-', linewidth=2.5, label="$f'(x) = \\frac{2x^3 - x^2 - 1}{x^2}$")
ax4.plot(x_pos, f_prime(x_pos), 'r-', linewidth=2.5)
ax4.axvline(x=0, color='red', linestyle='--', linewidth=2, alpha=0.7, label='Vertical asymptote at x=0')
ax4.set_xlabel('x', fontsize=12)
ax4.set_ylabel("f'(x)", fontsize=12)
ax4.set_title('Quotient Rule Derivative', fontsize=13, fontweight='bold')
ax4.grid(True, alpha=0.3)
ax4.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax4.legend(fontsize=11, loc='best')
ax4.set_ylim(-20, 20)

plt.tight_layout()
plt.show()

# Print detailed analysis
print("=" * 95)
print("Quotient Rule Demonstration: d(f(x)/g(x))/dx = [g(x)*df(x)/dx - f(x)*dg(x)/dx] / (g(x))²")
print("=" * 95)
print("\nGiven:")
print("  Numerator f(x) = x³ - x² + 1")
print("  Denominator g(x) = x")
print("  h(x) = f(x) / g(x) = (x³ - x² + 1) / x")
print("\nDerivatives using Quotient Rule:")
print("  f'(x) = 3x² - 2x")
print("  g'(x) = 1")
print("  h'(x) = [g(x) · f'(x) - f(x) · g'(x)] / (g(x))²")
print("       = [x · (3x² - 2x) - (x³ - x² + 1) · 1] / x²")
print("       = [3x³ - 2x² - x³ + x² - 1] / x²")
print("       = (2x³ - x² - 1) / x²")
print("\n" + "=" * 95)
print("Verification at specific points (where x ≠ 0):")
print("=" * 95)
header = "x        f(x)         g(x)         h(x)         f'(x)        g'(x)        h'(x)"
print(header)
print("-" * 95)

for x_val in [-3, -2, -1, 1, 2, 3]:
    f_val = numerator(x_val)
    g_val = denominator(x_val)
    h_val = f(x_val)
    fp_val = numerator_prime(x_val)
    gp_val = denominator_prime(x_val)
    hp_val = f_prime(x_val)
    print(f"{x_val:<8} {f_val:<12.2f} {g_val:<12.2f} {h_val:<12.2f} {fp_val:<12.2f} {gp_val:<12.2f} {hp_val:<12.2f}")

print("\n" + "=" * 95)
print("Key Insight:")
print("=" * 95)
print("• Quotient rule: (f/g)' = (g·f' - f·g') / g²  (NOT simply f'/g')")
print("• Note: h'(x) = (2x³ - x² - 1) / x², which matches our analytical solution")
print("• This rule is essential for differentiating rational and fractional functions")
print("• Undefined at x = 0 (where denominator = 0)")
print("=" * 95)
```

## Geometric Representation

The quotient rule provides a method for finding the derivative of a function that is the ratio of two functions. The formula may appear complex, but it follows a pattern: (denominator × derivative of numerator - numerator × derivative of denominator) divided by (denominator squared). This rule is essential when dealing with rational functions and fractions in calculus.
