# Rules of Differentiation: Power Rule

## Key Concepts

- Let $f(x)$ is a function of a single scalar variable $x$

## Power Rule

- Let $f(x) = x^n$, where $n$ is constant and $n > 0$

  $\dfrac{dx^n}{dx} = nx^{n-1}$

## Examples

$\dfrac{d(2x^2)}{dx} = 2 \cdot 2x^{2-1} = 4x$

$\dfrac{d(x^{-1})}{dx} = -1 \cdot x^{-1-1} = -x^{-2} = -\dfrac{1}{x^2}$

**Example 2**: $f(x) = x^{-1}$

$\dfrac{d(x^{-1})}{dx} = -1 \cdot x^{-1-1} = -x^{-2} = -\dfrac{1}{x^2}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure with subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Define x values
x_pos = np.linspace(0.1, 3, 100)
x_all = np.linspace(-3, 3, 100)

# Power functions to plot
power_functions = [
    (lambda x: x**1, lambda x: 1*x**(1-1), 1, "$f(x) = x^1$"),
    (lambda x: x**2, lambda x: 2*x**(2-1), 2, "$f(x) = x^2$"),
    (lambda x: x**3, lambda x: 3*x**(3-1), 3, "$f(x) = x^3$"),
    (lambda x: x**0.5, lambda x: 0.5*x**(0.5-1), 0.5, "$f(x) = x^{0.5}$"),
    (lambda x: x**(-1), lambda x: -1*x**(-1-1), -1, "$f(x) = x^{-1}$"),
    (lambda x: x**(-2), lambda x: -2*x**(-2-1), -2, "$f(x) = x^{-2}$"),
]

for idx, (func, deriv, n, title) in enumerate(power_functions):
    row = idx // 3
    col = idx % 3
    ax = axes[row, col]
    
    # Choose appropriate x range
    if n > 0:
        x = x_pos
    else:
        x = x_pos
    
    y = func(x)
    y_deriv = deriv(x)
    
    # Plot function and its derivative
    ax.plot(x, y, 'b-', linewidth=2.5, label=f'$f(x)$')
    ax.plot(x, y_deriv, 'r-', linewidth=2.5, label=f"$f'(x) = {n}x^{{{n-1}}}$")
    
    # Add annotations
    if n > 0:
        ax.set_xlim(0, 3)
    else:
        ax.set_xlim(0.1, 3)
    
    if n >= 0:
        ax.set_ylim(-5, 10)
    else:
        ax.set_ylim(-10, 2)
    
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('x', fontsize=11)
    ax.set_ylabel('y', fontsize=11)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.legend(fontsize=10, loc='best')
    ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
    ax.axvline(x=0, color='k', linestyle='-', linewidth=0.5)

plt.tight_layout()
plt.show()

# Print the power rule applications
print("=" * 80)
print("Power Rule Applications: f(x) = x^n  =>  f'(x) = n·x^(n-1)")
print("=" * 80)
print("\n{:<30} {:<30}".format("Function", "Derivative"))
print("-" * 80)

examples = [
    ("f(x) = x^1", "f'(x) = 1·x^0 = 1"),
    ("f(x) = x^2", "f'(x) = 2·x^1 = 2x"),
    ("f(x) = x^3", "f'(x) = 3·x^2 = 3x^2"),
    ("f(x) = x^(1/2)", "f'(x) = (1/2)·x^(-1/2) = 1/(2√x)"),
    ("f(x) = x^(-1)", "f'(x) = -1·x^(-2) = -1/x^2"),
    ("f(x) = x^(-2)", "f'(x) = -2·x^(-3) = -2/x^3"),
]

for func, deriv in examples:
    print("{:<30} {:<30}".format(func, deriv))

print("\n" + "=" * 80)
print("Key Observations:")
print("=" * 80)
print("• For positive n: function and derivative are both positive for x > 0")
print("• For negative n: function and derivative have opposite signs (inverse relationship)")
print("• The derivative's exponent is always n-1 (one less than original)")
print("• The coefficient multiplies by n from the original power")
print("=" * 80)
```

## Geometric Representation

The power rule shows how the derivative changes as the exponent changes. When the exponent is larger, the function grows more rapidly, and the slope of the tangent line (the derivative) is steeper. The power rule is one of the most fundamental differentiation rules and is widely used in calculus and machine learning.
