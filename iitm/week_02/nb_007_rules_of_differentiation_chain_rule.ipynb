# Rules of Differentiation: Chain Rule

## Key Concepts

- Let $f(x)$ and $g(x)$ are functions of a single scalar variable $x$

## Chain Rule

- Let $f(u)$ be a function of $u$ and $u = g(x)$, so that the function is $f(g(x))$

  $\dfrac{d f(g(x))}{dx} = \dfrac{d f(g(x))}{du} \cdot \dfrac{dg(x)}{dx}$

  Or equivalently:

  $\dfrac{df}{dx} = \dfrac{df}{du} \cdot \dfrac{du}{dx}$

## Examples

**Example 1**: $y = \log(x^2)$

  $\dfrac{dy}{dx} = \dfrac{d\log(u)}{du} \cdot \dfrac{dx^2}{dx} = \dfrac{1}{u} \cdot 2x = \dfrac{1}{x^2} \cdot 2x = \dfrac{2}{x}$

**Example 2**: $y = \exp\left(-\dfrac{x^2}{2}\right)$

  $\dfrac{dy}{dx} = \dfrac{d\exp(u)}{du} \cdot \dfrac{d(-x^2/2)}{dx} = \exp(u) \cdot (-x) = \exp\left(-\dfrac{x^2}{2}\right) \cdot (-x)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(0.1, 4, 200)

# Example 1: y = log(x²)
def example1_u(x):
    return x**2

def example1_f(u):
    return np.log(u)

def example1_y(x):
    return example1_f(example1_u(x))

def example1_du_dx(x):
    return 2*x

def example1_df_du(u):
    return 1/u

def example1_dy_dx(x):
    return example1_df_du(example1_u(x)) * example1_du_dx(x)

# Example 2: y = exp(-x²/2)
def example2_u(x):
    return -x**2 / 2

def example2_f(u):
    return np.exp(u)

def example2_y(x):
    return example2_f(example2_u(x))

def example2_du_dx(x):
    return -x

def example2_df_du(u):
    return np.exp(u)

def example2_dy_dx(x):
    return example2_df_du(example2_u(x)) * example2_du_dx(x)

# Create figure with subplots
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(3, 2, hspace=0.3, wspace=0.3)

# Example 1: y = log(x²)
ax1_1 = fig.add_subplot(gs[0, 0])
ax1_2 = fig.add_subplot(gs[1, 0])
ax1_3 = fig.add_subplot(gs[2, 0])

# Plot 1a: u(x) and f(u)
ax1_1.plot(x, example1_u(x), 'b-', linewidth=2, label='$u(x) = x^2$', alpha=0.7)
ax1_1_twin = ax1_1.twinx()
u_vals = np.linspace(0.1, 16, 100)
ax1_1_twin.plot(u_vals, example1_f(u_vals), 'g-', linewidth=2, label='$f(u) = \\log(u)$', alpha=0.7)
ax1_1.set_xlabel('x', fontsize=11)
ax1_1.set_ylabel('$u(x) = x^2$', fontsize=11, color='b')
ax1_1_twin.set_ylabel('$f(u) = \\log(u)$', fontsize=11, color='g')
ax1_1.set_title('Example 1: Component Functions', fontsize=12, fontweight='bold')
ax1_1.grid(True, alpha=0.3)
ax1_1.tick_params(axis='y', labelcolor='b')
ax1_1_twin.tick_params(axis='y', labelcolor='g')

# Plot 1b: Composite function
ax1_2.plot(x, example1_y(x), 'r-', linewidth=2.5, label='$y(x) = \\log(x^2)$')
ax1_2.set_xlabel('x', fontsize=11)
ax1_2.set_ylabel('y', fontsize=11)
ax1_2.set_title('Example 1: Composite Function', fontsize=12, fontweight='bold')
ax1_2.grid(True, alpha=0.3)
ax1_2.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1_2.legend(fontsize=10)

# Plot 1c: Derivatives
ax1_3.plot(x, example1_dy_dx(x), 'r-', linewidth=2.5, label="$\\frac{dy}{dx} = \\frac{df}{du} \\cdot \\frac{du}{dx}$")
ax1_3.plot(x, example1_du_dx(x), 'b--', linewidth=1.5, alpha=0.6, label="$\\frac{du}{dx} = 2x$")
ax1_3.plot(x, example1_df_du(example1_u(x)), 'g--', linewidth=1.5, alpha=0.6, label="$\\frac{df}{du} = \\frac{1}{u}$")
ax1_3.set_xlabel('x', fontsize=11)
ax1_3.set_ylabel("y'", fontsize=11)
ax1_3.set_title('Example 1: Chain Rule Derivative', fontsize=12, fontweight='bold')
ax1_3.grid(True, alpha=0.3)
ax1_3.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1_3.legend(fontsize=10)

# Example 2: y = exp(-x²/2)
ax2_1 = fig.add_subplot(gs[0, 1])
ax2_2 = fig.add_subplot(gs[1, 1])
ax2_3 = fig.add_subplot(gs[2, 1])

# Plot 2a: u(x) and f(u)
ax2_1.plot(x, example2_u(x), 'b-', linewidth=2, label='$u(x) = -\\frac{x^2}{2}$', alpha=0.7)
ax2_1_twin = ax2_1.twinx()
u_vals2 = np.linspace(-8, 0, 100)
ax2_1_twin.plot(u_vals2, example2_f(u_vals2), 'g-', linewidth=2, label='$f(u) = e^u$', alpha=0.7)
ax2_1.set_xlabel('x', fontsize=11)
ax2_1.set_ylabel('$u(x) = -\\frac{x^2}{2}$', fontsize=11, color='b')
ax2_1_twin.set_ylabel('$f(u) = e^u$', fontsize=11, color='g')
ax2_1.set_title('Example 2: Component Functions', fontsize=12, fontweight='bold')
ax2_1.grid(True, alpha=0.3)
ax2_1.tick_params(axis='y', labelcolor='b')
ax2_1_twin.tick_params(axis='y', labelcolor='g')

# Plot 2b: Composite function
ax2_2.plot(x, example2_y(x), 'r-', linewidth=2.5, label='$y(x) = e^{-x^2/2}$')
ax2_2.set_xlabel('x', fontsize=11)
ax2_2.set_ylabel('y', fontsize=11)
ax2_2.set_title('Example 2: Composite Function (Gaussian)', fontsize=12, fontweight='bold')
ax2_2.grid(True, alpha=0.3)
ax2_2.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax2_2.legend(fontsize=10)

# Plot 2c: Derivatives
ax2_3.plot(x, example2_dy_dx(x), 'r-', linewidth=2.5, label="$\\frac{dy}{dx} = \\frac{df}{du} \\cdot \\frac{du}{dx}$")
ax2_3.set_xlabel('x', fontsize=11)
ax2_3.set_ylabel("y'", fontsize=11)
ax2_3.set_title('Example 2: Chain Rule Derivative', fontsize=12, fontweight='bold')
ax2_3.grid(True, alpha=0.3)
ax2_3.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax2_3.legend(fontsize=10)

plt.suptitle('Chain Rule: $\\frac{dy}{dx} = \\frac{df}{du} \\cdot \\frac{du}{dx}$', fontsize=14, fontweight='bold', y=0.995)
plt.show()

# Print detailed analysis
print("=" * 100)
print("Chain Rule Demonstration: dy/dx = (df/du) · (du/dx)")
print("=" * 100)
print("\n" + "=" * 100)
print("EXAMPLE 1: y = log(x²)")
print("=" * 100)
print("\nStep 1: Identify the composition")
print("  Inner function: u(x) = x²")
print("  Outer function: f(u) = log(u)")
print("  Composite: y = f(u(x)) = log(x²)")
print("\nStep 2: Find individual derivatives")
print("  du/dx = 2x")
print("  df/du = 1/u = 1/x²")
print("\nStep 3: Apply chain rule")
print("  dy/dx = (df/du) · (du/dx)")
print("       = (1/u) · (2x)")
print("       = (1/x²) · (2x)")
print("       = 2/x")

print("\n" + "=" * 100)
print("EXAMPLE 2: y = exp(-x²/2)")
print("=" * 100)
print("\nStep 1: Identify the composition")
print("  Inner function: u(x) = -x²/2")
print("  Outer function: f(u) = exp(u)")
print("  Composite: y = f(u(x)) = exp(-x²/2)")
print("\nStep 2: Find individual derivatives")
print("  du/dx = -x")
print("  df/du = exp(u) = exp(-x²/2)")
print("\nStep 3: Apply chain rule")
print("  dy/dx = (df/du) · (du/dx)")
print("       = exp(-x²/2) · (-x)")
print("       = -x · exp(-x²/2)")

print("\n" + "=" * 100)
print("Verification at specific points (Example 1: y = log(x²)):")
print("=" * 100)
print(f"{'x':<8} {'u(x)':<15} {'f(u)':<15} {'du/dx':<15} {'df/du':<15} {'dy/dx':<15}")
print("-" * 100)

for x_val in [0.5, 1, 2, 3, 4]:
    u_val = example1_u(x_val)
    f_val = example1_f(u_val)
    du_dx = example1_du_dx(x_val)
    df_du = example1_df_du(u_val)
    dy_dx = example1_dy_dx(x_val)
    print(f"{x_val:<8} {u_val:<15.4f} {f_val:<15.4f} {du_dx:<15.4f} {df_du:<15.4f} {dy_dx:<15.4f}")

print("\n" + "=" * 100)
print("Key Insight:")
print("=" * 100)
print("• Chain rule is crucial for composite functions: (f∘g)' = (f'∘g) · g'")
print("• The derivative is the product of individual derivatives at each level")
print("• Essential for neural networks and backpropagation in deep learning")
print("• Can be applied repeatedly for multiple levels of composition")
print("=" * 100)
```

## Geometric Representation

The chain rule is one of the most important rules in calculus. It describes how to find the derivative of a composite function (a function composed of other functions). The key insight is that the total rate of change is the product of the rates of change at each level of composition. The chain rule is essential for differentiating complex functions and is widely used in machine learning for backpropagation.
