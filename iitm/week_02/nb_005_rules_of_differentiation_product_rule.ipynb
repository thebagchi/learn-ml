# Differentiation: Rules of Differentiation - Product Rule

## Key Concepts

- Let $f(x)$ and $g(x)$ are functions of a single scalar variable $x$

## Product Rule

- Let $f(x) \cdot g(x)$ be a function,

  $\dfrac{d(f(x)g(x))}{dx} = f(x) \dfrac{dg(x)}{dx} + g(x) \dfrac{df(x)}{dx}$

## Example

Let $f(x) = (2x^2 - 2)(x + 4)$

Using the product rule where $g(x) = 2x^2 - 2$ and $u(x) = x + 4$:

$\dfrac{df(x)}{dx} = \dfrac{d(g(x) \cdot u(x))}{dx} = g(x) \dfrac{d(u(x))}{dx} + u(x) \dfrac{dg(x)}{dx}$

$= (2x^2 - 2) \cdot 1 + (x + 4) \cdot 4x$

$= 2x^2 - 2 + 4x^2 + 16x = 6x^2 + 16x - 2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-2, 4, 100)

# Define individual functions
def g(x):
    return 2*x**2 - 2

def u(x):
    return x + 4

# Product of functions
def f(x):
    return g(x) * u(x)

# Derivatives of individual functions
def g_prime(x):
    return 4*x

def u_prime(x):
    return np.ones_like(x)  # derivative of (x + 4) is 1

# Product derivative using product rule
def f_prime(x):
    return g(x) * u_prime(x) + u(x) * g_prime(x)

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Individual functions
ax1 = axes[0, 0]
ax1.plot(x, g(x), 'b-', linewidth=2, label='$g(x) = 2x^2 - 2$', alpha=0.7)
ax1.plot(x, u(x), 'g-', linewidth=2, label='$u(x) = x + 4$', alpha=0.7)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('Component Functions', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax1.legend(fontsize=11, loc='best')

# Plot 2: Product function
ax2 = axes[0, 1]
ax2.plot(x, f(x), 'r-', linewidth=2.5, label='$f(x) = g(x) \\cdot u(x)$')
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('f(x)', fontsize=12)
ax2.set_title('Product Function: $f(x) = (2x^2 - 2)(x + 4)$', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax2.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax2.legend(fontsize=11, loc='best')

# Plot 3: Component derivatives
ax3 = axes[1, 0]
ax3.plot(x, g_prime(x), 'b--', linewidth=2, label="$g'(x) = 4x$", alpha=0.7)
ax3.plot(x, u_prime(x), 'g--', linewidth=2, label="$u'(x) = 1$", alpha=0.7)
ax3.set_xlabel('x', fontsize=12)
ax3.set_ylabel("y'", fontsize=12)
ax3.set_title('Component Derivatives', fontsize=13, fontweight='bold')
ax3.grid(True, alpha=0.3)
ax3.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax3.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax3.legend(fontsize=11, loc='best')

# Plot 4: Product derivative
ax4 = axes[1, 1]
ax4.plot(x, f_prime(x), 'r-', linewidth=2.5, label="$f'(x) = g(x) \\cdot u'(x) + u(x) \\cdot g'(x)$")
ax4.set_xlabel('x', fontsize=12)
ax4.set_ylabel("f'(x)", fontsize=12)
ax4.set_title('Product Rule Derivative: $f\'(x) = 6x^2 + 16x - 2$', fontsize=13, fontweight='bold')
ax4.grid(True, alpha=0.3)
ax4.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax4.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax4.legend(fontsize=11, loc='best')

plt.tight_layout()
plt.show()

# Print detailed analysis
print("=" * 90)
print("Product Rule Demonstration: d(f(x) * u(x))/dx = f(x) * du(x)/dx + u(x) * df(x)/dx")
print("=" * 90)
print("\nGiven:")
print("  g(x) = 2x² - 2")
print("  u(x) = x + 4")
print("  f(x) = g(x) · u(x) = (2x² - 2)(x + 4)")
print("\nDerivatives using Product Rule:")
print("  g'(x) = 4x")
print("  u'(x) = 1")
print("  f'(x) = g(x) · u'(x) + u(x) · g'(x)")
print("       = (2x² - 2) · 1 + (x + 4) · 4x")
print("       = 2x² - 2 + 4x² + 16x")
print("       = 6x² + 16x - 2")
print("\n" + "=" * 90)
print("Verification at specific points:")
print("=" * 90)
header = "x        g(x)         u(x)         f(x)         g'(x)        u'(x)        f'(x)"
print(header)
print("-" * 90)

for x_val in [-2, -1, 0, 1, 2]:
    f_val = f(x_val)
    g_val = g(x_val)
    u_val = u(x_val)
    gp_val = g_prime(x_val)
    up_val = u_prime(x_val)
    fp_val = f_prime(x_val)
    print(f"{x_val:<8} {g_val:<12.2f} {u_val:<12.2f} {f_val:<12.2f} {gp_val:<12.2f} {up_val:<12.2f} {fp_val:<12.2f}")

print("\n" + "=" * 90)
print("Key Insight:")
print("=" * 90)
print("• Product rule: (f·u)' = f·u' + u·f'  (NOT simply f'·u')")
print("• Note: f'(x) = 6x² + 16x - 2, which matches our analytical solution")
print("• This rule is essential for differentiating polynomials and complex expressions")
print("=" * 90)
```

## Geometric Representation

The product rule shows how to find the derivative of a product of two functions. Unlike the sum rule, the derivative of a product is not simply the product of derivatives. Instead, we must use the formula: the derivative equals the first function times the derivative of the second, plus the second function times the derivative of the first. This is a fundamental rule in calculus with many applications.
