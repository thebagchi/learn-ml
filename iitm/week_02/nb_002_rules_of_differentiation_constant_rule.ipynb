# Differentiation: Rules of Differentiation

## Key Concepts

- Let $f(x)$ is a function of a single scalar variable $x$

## Constant Rule

- Let $f(x) = k$, where $k$ is constant and $k > 0$

  $\dfrac{dk}{dx} = 0$

```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure and subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Define x values
x = np.linspace(-5, 5, 100)

# Example 1: f(x) = 3
k1 = 3
y1 = np.ones_like(x) * k1

axes[0].plot(x, y1, 'b-', linewidth=2.5, label=f'$f(x) = {k1}$')
axes[0].plot([-5, 5], [k1, k1], 'b-', linewidth=2.5)
# Add tangent lines at different points to show zero slope
points = [-3, 0, 3]
for point in points:
    axes[0].plot(point, k1, 'ro', markersize=8)
    axes[0].plot([point-1, point+1], [k1, k1], 'r--', linewidth=1.5, alpha=0.7)

axes[0].set_xlabel('x', fontsize=12)
axes[0].set_ylabel('f(x)', fontsize=12)
axes[0].set_title(f'$f(x) = {k1}$ (Constant Function)', fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(-1, 7)
axes[0].axhline(y=0, color='k', linestyle='-', linewidth=0.5)
axes[0].axvline(x=0, color='k', linestyle='-', linewidth=0.5)
axes[0].legend(fontsize=11)

# Example 2: f(x) = -2
k2 = -2
y2 = np.ones_like(x) * k2

axes[1].plot(x, y2, 'g-', linewidth=2.5, label=f'$f(x) = {k2}$')
axes[1].plot([-5, 5], [k2, k2], 'g-', linewidth=2.5)
# Add tangent lines at different points to show zero slope
for point in points:
    axes[1].plot(point, k2, 'ro', markersize=8)
    axes[1].plot([point-1, point+1], [k2, k2], 'r--', linewidth=1.5, alpha=0.7)

axes[1].set_xlabel('x', fontsize=12)
axes[1].set_ylabel('f(x)', fontsize=12)
axes[1].set_title(f'$f(x) = {k2}$ (Constant Function)', fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim(-5, 2)
axes[1].axhline(y=0, color='k', linestyle='-', linewidth=0.5)
axes[1].axvline(x=0, color='k', linestyle='-', linewidth=0.5)
axes[1].legend(fontsize=11)

plt.tight_layout()
plt.show()

# Print explanation
print("=" * 70)
print("Geometric Representation: Constant Rule")
print("=" * 70)
print("\nFor a constant function f(x) = k:")
print("  • The graph is a horizontal line")
print("  • At every point, the tangent line is also horizontal")
print("  • A horizontal line has slope = 0")
print("  • Therefore, f'(x) = 0 for all x")
print("\nExamples:")
print(f"  1. f(x) = 3  →  f'(x) = 0")
print(f"  2. f(x) = -2  →  f'(x) = 0")
print(f"  3. f(x) = 5  →  f'(x) = 0")
print("\nRed dots show points on the curve")
print("Red dashed lines show tangent lines (all horizontal with slope 0)")
print("=" * 70)
```

## Geometric Representation

The constant rule states that if $f(x) = k$ (a horizontal line), then $f'(x) = 0$.

**Geometric Interpretation**: A horizontal line has zero slope. Since the derivative represents the slope of the tangent line at any point on the curve, a constant function has a slope of 0 everywhere.
