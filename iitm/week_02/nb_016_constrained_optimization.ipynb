# Constrained Optimization

## Introduction

**Constrained optimization** is a process of optimizing an objective function with respect to some variables in the presence of constraints on those variables.

### Key Differences from Unconstrained Optimization

- In **unconstrained optimization**, we can choose any value for the variables
- In **constrained optimization**, the variables must satisfy certain restrictions (constraints)
- The solution must be feasible (satisfy all constraints) and optimal

### Problem Components

Let $\mathbf{x}$ be a vector of $d$ variables

A general constrained optimization problem may be written as:

$\text{optimize } f(\mathbf{x})$
$\text{subject to:}$
$g_i(\mathbf{x}) = 0 \quad \text{for } i = 1, \ldots, n_1$
$h_j(\mathbf{x}) > 0 \quad \text{for } j = 1, \ldots, n_2$
$p_k(\mathbf{x}) < 0 \quad \text{for } k = 1, \ldots, n_3$

### Constraint Types

- $g_i(\mathbf{x}) = 0$ for $i = 1, \ldots, n_1$ are **equality constraints**
- $h_j(\mathbf{x}) > 0$ for $j = 1, \ldots, n_2$ and $p_k(\mathbf{x}) < 0$ for $k = 1, \ldots, n_3$ are **inequality constraints** that must be satisfied
- $f(\mathbf{x})$ is the **objective function** that needs to be optimized subject to the constraints

### Important Note

Optimization means either **maximization or minimization** depending on the problem context.

## Steps for Solving Constrained Optimization

### General Approach

Steps involved in solving a constrained optimization problem:

1. **Identify the objective function** — The function to be optimized (minimized or maximized)

2. **Identify the constraints** — All restrictions on the variables

3. **Find the best solution** — Determine the feasible point that optimizes the objective function

### Methods for Constrained Optimization

There are many methods for constrained optimization, including:

- **Lagrange multiplier method** — Popular and widely used
- Penalty methods
- Interior point methods
- Sequential quadratic programming (SQP)

### Key Insight: Lagrange Multiplier Method

The **Lagrange multiplier method** is one of the most important techniques because it:

- **Converts constrained optimization problem into an unconstrained problem** using Lagrange multipliers
- There will be one Lagrange multiplier per constraint
- Once converted to unconstrained form, we can apply unconstrained optimization techniques

## The Lagrange Multiplier Method

### Overview

The Lagrange multiplier method converts a constrained optimization problem into an unconstrained problem by introducing additional variables (multipliers) for each constraint.

### Steps in Lagrange Multiplier Method

Let $f(x)$ be a function of single variable $x$

#### Step 1: Identify the Primal Objective Function

$\text{optimize } f(x)$
$\text{such that } g(x) = 0$
$h(x) > 0$
$p(x) < 0$

where $f(x)$ is the **primary objective function**

#### Step 2: Formulate Lagrangian Objective Function

Convert the constrained optimization problem to an unconstrained problem by treating all constraints as secondary objective functions and adding or subtracting them to the primary objective function by considering Lagrange multipliers as weights to addition/subtraction:

$L(x, \alpha, \beta, \gamma) = f(x) - \alpha(g(x)) - \beta(h(x)) + \gamma(p(x))$

where the coefficients $\alpha$, $\beta$ and $\gamma$ are **Lagrange multipliers**

#### Step 3: Obtain the Optimality Condition(s)

Take the derivative of the Lagrangian with respect to the primal variable and set it to zero:

$\frac{\partial L(x, \alpha, \beta, \gamma)}{\partial x} = 0$

#### Step 4: Obtain the Dual Objective Function

The dual objective function is obtained by substituting the optimality condition(s) in the Lagrangian objective function. This dual objective function will be a function of only the Lagrangian multipliers:

$L_D(\alpha, \beta, \gamma)$

#### Step 5: Optimize the Dual Objective Function

Optimize the dual objective function:

$\text{optimize } L_D(\alpha, \beta, \gamma)$

This gives the optimal values for $\alpha$, $\beta$ and $\gamma$

#### Step 6: Find the Optimal Primal Solution

Substitute the optimal values of Lagrangian multipliers in the optimality condition to get the optimal value $x^*$

### Important Note

- If the **primal objective function is minimised**, then the **dual objective function will be maximised** and vice versa

## Constrained Optimization: Single Variable Example

### Problem Statement

Let $f(x) = \frac{1}{2}cx^2$ where $c$ is a constant

**Constrained optimization problem:** Find the value for $x^*$ for which the function $f(x) = \frac{1}{2}cx^2$ is minimum with the constraint that $ax - b < 0$

$x^* = \text{minimize } \frac{1}{2}cx^2$
$\text{subject to } ax - b < 0$

### Step 1: Formulate the Lagrangian

The Lagrangian objective function is:

$L(x, \lambda) = \frac{1}{2}cx^2 + \lambda(ax - b) \quad \text{...(1)}$

where $\lambda$ is the Lagrange multiplier

### Step 2: Obtain Optimality Condition

Take the derivative with respect to $x$ and set it to zero:

$\frac{df(x)}{dx} = 0 \quad \Rightarrow \quad \frac{d\left(\frac{1}{2}cx^2 + \lambda(ax-b)\right)}{dx} = 0$

$cx + \lambda a = 0 \quad \Rightarrow \quad x = -\frac{\lambda a}{c} \quad \text{...(2)}$

### Step 3: Obtain the Dual Objective Function

Substitute equation (2) in equation (1):

$L_D(\lambda) = \frac{1}{2}c\left(-\frac{\lambda a}{c}\right)^2 + \lambda\left(a\left(-\frac{\lambda a}{c}\right) - b\right)$

Simplifying:

$L_D(\lambda) = -\frac{a^2\lambda^2}{2c} - \lambda b$

### Step 4: Optimize the Dual Objective Function

Since the primal is minimization, we maximize the dual:

$\text{maximize } L_D(\lambda)$

$\frac{dL_D(\lambda)}{d\lambda} = 0 \quad \Rightarrow \quad \frac{d\left(-\frac{a^2\lambda^2}{2c} - \lambda b\right)}{d\lambda} = 0$

$-\frac{a^2 \cdot 2\lambda}{2c} - b = 0 \quad \Rightarrow \quad \lambda^* = -\frac{bc}{a^2} \quad \text{...(3)}$

### Step 5: Substitute to Get Optimal Primal Solution

Substitute equation (3) in equation (2):

$x^* = -\frac{\lambda^* a}{c} = -\frac{\left(-\frac{bc}{a^2}\right) \cdot a}{c} = \frac{b}{a}$

**The optimal solution is:** $x^* = \frac{b}{a}$

## Numerical Example: Constrained Optimization

### Problem Setup

Let $c = 1$, $a = -2$, and $b = 3$

**Objective function:** $f(x) = \frac{1}{2}x^2$

**Constraint:** $-2x - 3 < 0$ (which means $-2x < 3$, or $x > -1.5$)

**Find:** $x^* = \text{minimize } \frac{1}{2}x^2$ subject to $-2x - 3 < 0$

### Solution

Using the formula derived:

$x^* = \frac{b}{a} = \frac{3}{-2} = -1.5$

### Verification

- **Objective value:** $f(x^*) = \frac{1}{2}(-1.5)^2 = 1.125$
- **Constraint check:** $-2(-1.5) - 3 = 3 - 3 = 0$ ✓ (constraint is satisfied at boundary)
- **Feasibility:** The constraint $-2x - 3 < 0$ requires $x > -1.5$, so $x^* = -1.5$ is at the boundary

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
c = 1
a = -2
b = 3

# Objective function
def f(x):
    return 0.5 * c * x**2

# Constraint: ax - b < 0
def constraint(x):
    return a * x - b

# Optimal solution
x_optimal = b / a
f_optimal = f(x_optimal)

print("Constrained Optimization Example")
print(f"Objective function: f(x) = 0.5*x^2")
print(f"Constraint: -2x - 3 < 0 (i.e., x > -1.5)")
print()
print(f"Optimal solution: x* = {x_optimal}")
print(f"Optimal value: f(x*) = {f_optimal}")
print(f"Constraint value at x*: g(x*) = {constraint(x_optimal)}")
print()

# Visualization
x = np.linspace(-5, 3, 1000)
y = f(x)
constraint_line = constraint(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Left plot: Objective function and constraint
ax1.plot(x, y, 'b-', linewidth=2.5, label='f(x) = 0.5*x²')
ax1.axvline(x_optimal, color='g', linestyle='--', linewidth=2, alpha=0.7, label='Feasible region boundary')
ax1.fill_betweenx([0, 50], -1.5, 3, alpha=0.2, color='green', label='Feasible region: x > -1.5')
ax1.plot(x_optimal, f_optimal, 'r*', markersize=20, label=f'Constrained minimum at ({x_optimal}, {f_optimal})')
ax1.set_xlim(-5, 3)
ax1.set_ylim(-5, 20)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_title('Constrained Optimization: Function and Feasible Region')
ax1.legend()

# Right plot: Constraint visualization
ax2.plot(x, constraint_line, 'r-', linewidth=2.5, label='g(x) = -2x - 3')
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax2.axvline(x=-1.5, color='g', linestyle='--', linewidth=2, alpha=0.7, label='Constraint boundary')
ax2.fill_between(x, constraint_line, 50, where=(constraint_line < 0), alpha=0.2, color='green', label='Feasible region: g(x) < 0')
ax2.plot(x_optimal, 0, 'r*', markersize=20, label=f'Boundary point: x* = {x_optimal}')
ax2.set_xlim(-5, 3)
ax2.set_ylim(-15, 10)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x')
ax2.set_ylabel('g(x) = -2x - 3')
ax2.set_title('Constraint Visualization')
ax2.legend()

plt.tight_layout()
plt.show()
```

## Constrained Optimization: Multiple Variables

### General Problem Formulation

For functions with multiple variables, the methodology extends naturally:

**Minimize or Maximize:** $f(\mathbf{x})$

**Subject to:**
- Equality constraints: $g_i(\mathbf{x}) = 0$ for $i = 1, \ldots, n_1$
- Inequality constraints: $h_j(\mathbf{x}) \leq 0$ for $j = 1, \ldots, n_2$

### Lagrangian for Multiple Variables

$L(\mathbf{x}, \lambda, \mu) = f(\mathbf{x}) + \sum_{i=1}^{n_1} \lambda_i g_i(\mathbf{x}) + \sum_{j=1}^{n_2} \mu_j h_j(\mathbf{x})$

where:
- $\lambda_i$ are Lagrange multipliers for equality constraints
- $\mu_j$ are Lagrange multipliers for inequality constraints (typically $\mu_j \geq 0$ for inequality constraints)

### Optimality Conditions (KKT Conditions)

The Karush-Kuhn-Tucker (KKT) conditions are necessary conditions for optimality in constrained optimization:

1. **Stationarity:** $\nabla f(\mathbf{x}^*) + \sum_i \lambda_i \nabla g_i(\mathbf{x}^*) + \sum_j \mu_j \nabla h_j(\mathbf{x}^*) = \mathbf{0}$

2. **Primal feasibility:** $g_i(\mathbf{x}^*) = 0$ and $h_j(\mathbf{x}^*) \leq 0$

3. **Dual feasibility:** $\mu_j \geq 0$

4. **Complementary slackness:** $\mu_j h_j(\mathbf{x}^*) = 0$ (either $\mu_j = 0$ or $h_j(\mathbf{x}^*) = 0$)

## Constrained Optimization: Two Variable Example

### Problem Statement

**Minimize:** $f(x_1, x_2) = x_1^2 + 2x_2^2$

**Subject to:** $x_1 + x_2 = 3$ (equality constraint)

### Solution Using Lagrange Multipliers

#### Step 1: Formulate the Lagrangian

$L(x_1, x_2, \lambda) = x_1^2 + 2x_2^2 - \lambda(x_1 + x_2 - 3)$

#### Step 2: Optimality Conditions

$\frac{\partial L}{\partial x_1} = 2x_1 - \lambda = 0 \quad \Rightarrow \quad x_1 = \frac{\lambda}{2}$

$\frac{\partial L}{\partial x_2} = 4x_2 - \lambda = 0 \quad \Rightarrow \quad x_2 = \frac{\lambda}{4}$

$\frac{\partial L}{\partial \lambda} = -(x_1 + x_2 - 3) = 0 \quad \Rightarrow \quad x_1 + x_2 = 3$

#### Step 3: Solve the System

Substitute the expressions for $x_1$ and $x_2$ into the constraint:

$\frac{\lambda}{2} + \frac{\lambda}{4} = 3 \quad \Rightarrow \quad \frac{3\lambda}{4} = 3 \quad \Rightarrow \quad \lambda = 4$

Therefore:

$x_1^* = \frac{4}{2} = 2$

$x_2^* = \frac{4}{4} = 1$

#### Step 4: Evaluate the Objective Function

$f(x_1^*, x_2^*) = 2^2 + 2(1)^2 = 4 + 2 = 6$

**Optimal solution:** $\mathbf{x}^* = [2, 1]^T$ with $f(\mathbf{x}^*) = 6$

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Objective function
def f(x):
    return x[0]**2 + 2*x[1]**2

# Constraint: x1 + x2 = 3
def constraint_eq(x):
    return x[0] + x[1] - 3

# Solve using Lagrange multiplier method (analytical solution)
x1_opt = 2
x2_opt = 1
f_opt = f([x1_opt, x2_opt])

print("Constrained Optimization with Equality Constraint")
print(f"Objective: minimize x₁² + 2x₂²")
print(f"Constraint: x₁ + x₂ = 3")
print()
print(f"Optimal solution: x₁* = {x1_opt}, x₂* = {x2_opt}")
print(f"Optimal value: f(x*) = {f_opt}")
print(f"Constraint satisfaction: x₁* + x₂* = {x1_opt + x2_opt}")
print()

# Verify with numerical solver
from scipy.optimize import NonlinearConstraint
constraints = {'type': 'eq', 'fun': constraint_eq}
result = minimize(f, x0=[1, 1], method='SLSQP', constraints=constraints)
print("Verification using numerical solver:")
print(f"Numerical x* = [{result.x[0]:.4f}, {result.x[1]:.4f}]")
print(f"Numerical f(x*) = {result.fun:.4f}")
```

```python
# Visualize the constrained optimization problem
fig = plt.figure(figsize=(14, 5))

# Create grid for objective function
x1 = np.linspace(-1, 5, 100)
x2 = np.linspace(-1, 5, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = X1**2 + 2*X2**2

# Left plot: 3D surface with constraint
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.7)

# Plot constraint line on the surface
x1_constraint = np.linspace(-1, 5, 100)
x2_constraint = 3 - x1_constraint
z_constraint = x1_constraint**2 + 2*x2_constraint**2
ax1.plot(x1_constraint, x2_constraint, z_constraint, 'r-', linewidth=3, label='Constraint: x₁ + x₂ = 3')

# Plot optimal point
ax1.scatter([x1_opt], [x2_opt], [f_opt], color='red', s=200, marker='*', label=f'Optimum: ({x1_opt}, {x2_opt})')
ax1.set_xlabel('x₁')
ax1.set_ylabel('x₂')
ax1.set_zlabel('f(x₁, x₂)')
ax1.set_title('3D View: Constrained Optimization')
ax1.legend()

# Right plot: 2D contour with constraint
ax2 = fig.add_subplot(122)
contours = ax2.contour(X1, X2, Z, levels=20, cmap='viridis')
ax2.clabel(contours, inline=True, fontsize=8)

# Plot constraint line
ax2.plot(x1_constraint, x2_constraint, 'r-', linewidth=2.5, label='Constraint: x₁ + x₂ = 3')

# Plot optimal point
ax2.plot(x1_opt, x2_opt, 'r*', markersize=20, label=f'Optimum at ({x1_opt}, {x2_opt})')

ax2.set_xlabel('x₁')
ax2.set_ylabel('x₂')
ax2.set_title('Contour Plot: Constrained Optimization')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xlim(-1, 5)
ax2.set_ylim(-1, 5)

plt.tight_layout()
plt.show()
```

## Comparison: Unconstrained vs Constrained

### Same Problem Without Constraint

**Minimize:** $f(x_1, x_2) = x_1^2 + 2x_2^2$ (no constraint)

**Solution:**

Setting gradients to zero:

$\frac{\partial f}{\partial x_1} = 2x_1 = 0 \quad \Rightarrow \quad x_1 = 0$

$\frac{\partial f}{\partial x_2} = 4x_2 = 0 \quad \Rightarrow \quad x_2 = 0$

**Unconstrained optimum:** $\mathbf{x} = [0, 0]^T$ with $f(\mathbf{x}) = 0$

### Comparison

| Aspect | Unconstrained | Constrained |
|--------|---------------|-------------|
| Objective | $f(x_1, x_2) = x_1^2 + 2x_2^2$ | Same + constraint $x_1 + x_2 = 3$ |
| Optimal point | $[0, 0]$ | $[2, 1]$ |
| Optimal value | $0$ | $6$ |
| Feasibility | Not on constraint | On constraint: $2 + 1 = 3$ |
| Nature | Global minimum | Local minimum subject to constraint |

**Key Insight:** The constraint forces the solution to lie on the constraint boundary, preventing us from reaching the global unconstrained minimum.

## Role of Constrained Optimization in Machine Learning

### Where Constraints Arise in ML

Constrained optimization appears frequently in machine learning:

1. **Support Vector Machines (SVMs)**
   - Maximize margin between classes subject to classification constraints
   - Constraints ensure correct classification with margin

2. **Resource Allocation**
   - Optimize model performance subject to computational/memory constraints
   - Training time or model size limitations

3. **Portfolio Optimization**
   - Minimize risk subject to return constraints
   - Constraints on asset allocation (weights sum to 1)

4. **Deep Learning with Regularization**
   - Minimize loss subject to weight magnitude constraints
   - L1/L2 regularization can be viewed as constraints

5. **Reinforcement Learning**
   - Learn policies subject to safety or fairness constraints
   - Ensure safe exploration during learning

### The ML Optimization Pipeline with Constraints

1. **Define objective** — Formulate loss/cost function
2. **Identify constraints** — Business rules, fairness, safety, resource limits
3. **Choose method** — Lagrange multipliers, penalty methods, etc.
4. **Convert to solvable form** — Use Lagrangian to create unconstrained equivalent
5. **Optimize** — Find feasible solution that optimizes objective
6. **Validate** — Ensure all constraints are satisfied

### Why Constrained Optimization Matters

- **Real-world problems have constraints** — Budget, fairness, safety, regulatory requirements
- **Unconstrained solutions may be infeasible** — May violate business rules or resource limits
- **Constraints guide the solution** — Shape where optimal solution lies
- **Lagrange multipliers provide insights** — Show how sensitive optimal value is to constraint changes

## Summary

### Key Takeaways

1. **Constrained optimization** optimizes an objective function while satisfying restrictions on variables

2. **Problem formulation** requires:
   - Objective function: $f(\mathbf{x})$
   - Equality constraints: $g_i(\mathbf{x}) = 0$
   - Inequality constraints: $h_j(\mathbf{x}) \leq 0$ (or $> 0$)

3. **Lagrange multiplier method** converts constrained problems into unconstrained problems:
   - Introduce one multiplier per constraint
   - Form Lagrangian: $L(\mathbf{x}, \lambda, \mu) = f(\mathbf{x}) + \sum \lambda_i g_i + \sum \mu_j h_j$
   - Apply unconstrained optimization techniques

4. **Steps to solve:**
   - Formulate Lagrangian
   - Apply first-order conditions (set gradients to zero)
   - Form dual objective function
   - Optimize dual function
   - Back-substitute to find primal solution

5. **KKT conditions** are necessary for optimality in constrained problems:
   - Stationarity, primal feasibility, dual feasibility, complementary slackness

6. **Practical differences:**
   - Unconstrained solutions are global optima
   - Constrained solutions are feasible optima (may be suboptimal without constraints)
   - Constraints can significantly change the optimal solution

7. **ML applications:**
   - SVMs use constrained optimization for margin maximization
   - Fairness constraints in algorithmic decision-making
   - Resource-constrained optimization (time, memory, energy)
   - Safety constraints in reinforcement learning
