# Second Derivative Test: Minima and Maxima

## Key Concepts

The **Second Derivative Test** is a method to determine whether a critical point is a local minimum or maximum:

**Procedure:**
1. Find critical points by solving $f'(x) = 0$
2. Evaluate the second derivative $f''(x)$ at each critical point
3. Apply the test:
   - If $f''(c) > 0$ at critical point $x = c$, then $f(c)$ is a **local minimum**
   - If $f''(c) < 0$ at critical point $x = c$, then $f(c)$ is a **local maximum**
   - If $f''(c) = 0$, the test is **inconclusive** (may be inflection point)

## Local vs Global Extrema

- **Local (Relative) Extrema**: A point where the function has a minimum or maximum value within a small neighborhood, but not necessarily over the entire domain.
- **Global (Absolute) Extrema**: The largest or smallest value of a function over its entire domain. A global extremum is also a local extremum, but not all local extrema are global.

To find global extrema:
1. Find all local extrema using the second derivative test
2. Evaluate the function at the endpoints of the domain (if bounded)
3. Compare all values to determine the global maximum and minimum

## Example: Identifying Local Extrema Using Second Derivative Test

Let $f(x) = x^3 - 3x^2$

**First derivative:**

  $f'(x) = 3x^2 - 6x = 3x(x-2)$

Critical points: $x = 0$ and $x = 2$

**Second derivative:**

  $f''(x) = 6x - 6$

**Applying Second Derivative Test:**
- At $x = 0$: $f''(0) = -6 < 0$ → **Local maximum**
- At $x = 2$: $f''(2) = 12 - 6 = 6 > 0$ → **Local minimum**

**Function values at critical points:**
- $f(0) = 0^3 - 3(0)^2 = 0$
- $f(2) = 2^3 - 3(2)^2 = 8 - 12 = -4$

For the domain $[−1, 4]$:
- $f(-1) = -1 - 3 = -4$ (endpoint)
- $f(0) = 0$ (local maximum)
- $f(2) = -4$ (local minimum, also global minimum on this domain)
- $f(4) = 64 - 48 = 16$ (endpoint, **global maximum** on this domain)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function: f(x) = x³ - 3x²
def f(x):
    return x**3 - 3*x**2

# First derivative: f'(x) = 3x² - 6x
def f_prime(x):
    return 3*x**2 - 6*x

# Second derivative: f''(x) = 6x - 6
def f_double_prime(x):
    return 6*x - 6

# Create x values
x = np.linspace(-1, 3.5, 300)
y = f(x)
y_prime = f_prime(x)
y_double_prime = f_double_prime(x)

# Critical points
critical_points = [0, 2]
critical_values = [f(cp) for cp in critical_points]
critical_second_deriv = [f_double_prime(cp) for cp in critical_points]

# Create figure
fig = plt.figure(figsize=(16, 12))
gs = fig.add_gridspec(3, 2, hspace=0.35, wspace=0.3)

# Main plots
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, 0])
ax3 = fig.add_subplot(gs[1, 1])
ax4 = fig.add_subplot(gs[2, 0])
ax5 = fig.add_subplot(gs[2, 1])

# Plot 1: Original function with critical points
ax1.plot(x, y, 'b-', linewidth=3, label='$f(x) = x^3 - 3x^2$')
ax1.plot(0, f(0), 'r*', markersize=25, label='x=0: Local Maximum', zorder=5)
ax1.plot(2, f(2), 'g*', markersize=25, label='x=2: Local Minimum', zorder=5)
ax1.axvline(x=0, color='red', linestyle='--', alpha=0.5)
ax1.axvline(x=2, color='green', linestyle='--', alpha=0.5)
ax1.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax1.grid(True, alpha=0.3)
ax1.set_ylabel('f(x)', fontsize=12, fontweight='bold')
ax1.set_title('Original Function: $f(x) = x^3 - 3x^2$', fontsize=14, fontweight='bold')
ax1.legend(fontsize=11, loc='upper right')
ax1.fill_between(x, y, alpha=0.15, color='blue')

# Plot 2: First derivative
ax2.plot(x, y_prime, 'g-', linewidth=2.5, label="$f'(x) = 3x^2 - 6x$")
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
ax2.plot([0, 2], [0, 0], 'r*', markersize=15, zorder=5)
ax2.axvline(x=0, color='red', linestyle='--', alpha=0.5)
ax2.axvline(x=2, color='green', linestyle='--', alpha=0.5)
ax2.fill_between(x[x < 0], y_prime[x < 0], 0, alpha=0.3, color='green', label='Increasing')
ax2.fill_between(x[(x >= 0) & (x <= 2)], y_prime[(x >= 0) & (x <= 2)], 0, alpha=0.3, color='red', label='Decreasing')
ax2.fill_between(x[x > 2], y_prime[x > 2], 0, alpha=0.3, color='green', label='Increasing')
ax2.grid(True, alpha=0.3)
ax2.set_ylabel("f'(x)", fontsize=12, fontweight='bold')
ax2.set_title('First Derivative: Critical Points', fontsize=13, fontweight='bold')
ax2.legend(fontsize=10, loc='upper left')

# Plot 3: Second derivative
ax3.plot(x, y_double_prime, 'r-', linewidth=2.5, label="$f''(x) = 6x - 6$")
ax3.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
ax3.plot(0, f_double_prime(0), 'r*', markersize=15, zorder=5)
ax3.plot(2, f_double_prime(2), 'g*', markersize=15, zorder=5)
ax3.axvline(x=0, color='red', linestyle='--', alpha=0.5, label="x=0: f''(0)=-6<0")
ax3.axvline(x=2, color='green', linestyle='--', alpha=0.5, label="x=2: f''(2)=6>0")
ax3.fill_between(x[x < 1], y_double_prime[x < 1], 0, alpha=0.3, color='red', label='Concave down')
ax3.fill_between(x[x > 1], y_double_prime[x > 1], 0, alpha=0.3, color='green', label='Concave up')
ax3.grid(True, alpha=0.3)
ax3.set_ylabel("f''(x)", fontsize=12, fontweight='bold')
ax3.set_title('Second Derivative Test', fontsize=13, fontweight='bold')
ax3.legend(fontsize=10, loc='upper left')

# Plot 4: Zoomed view at x=0 (maximum)
x_zoom_0 = np.linspace(-0.8, 0.8, 200)
y_zoom_0 = f(x_zoom_0)
ax4.plot(x_zoom_0, y_zoom_0, 'b-', linewidth=3, label='$f(x)$')
ax4.plot(0, f(0), 'r*', markersize=20, zorder=5)
# Draw tangent line
ax4.plot([-0.8, 0.8], [f(0), f(0)], 'r--', linewidth=2, alpha=0.7, label='Tangent (slope=0)')
ax4.grid(True, alpha=0.3)
ax4.set_xlabel('x', fontsize=12, fontweight='bold')
ax4.set_ylabel('f(x)', fontsize=12, fontweight='bold')
ax4.set_title('x=0: Local Maximum\n$f''(0)=-6<0$ (Concave Down)', fontsize=12, fontweight='bold', color='red')
ax4.legend(fontsize=10)
ax4.fill_between(x_zoom_0, y_zoom_0, alpha=0.2, color='red')

# Plot 5: Zoomed view at x=2 (minimum)
x_zoom_2 = np.linspace(1.2, 2.8, 200)
y_zoom_2 = f(x_zoom_2)
ax5.plot(x_zoom_2, y_zoom_2, 'b-', linewidth=3, label='$f(x)$')
ax5.plot(2, f(2), 'g*', markersize=20, zorder=5)
# Draw tangent line
ax5.plot([1.2, 2.8], [f(2), f(2)], 'g--', linewidth=2, alpha=0.7, label='Tangent (slope=0)')
ax5.grid(True, alpha=0.3)
ax5.set_xlabel('x', fontsize=12, fontweight='bold')
ax5.set_ylabel('f(x)', fontsize=12, fontweight='bold')
ax5.set_title('x=2: Local Minimum\n$f''(2)=6>0$ (Concave Up)', fontsize=12, fontweight='bold', color='green')
ax5.legend(fontsize=10)
ax5.fill_between(x_zoom_2, y_zoom_2, alpha=0.2, color='green')

plt.suptitle('Second Derivative Test: How f\'\'(x) Determines Minima and Maxima', 
             fontsize=15, fontweight='bold', y=0.995)
plt.show()

# Print comprehensive analysis
print("=" * 100)
print("SECOND DERIVATIVE TEST: f(x) = x³ - 3x²")
print("=" * 100)
print("\nStep 1: Find Critical Points (where f'(x) = 0)")
print(f"  f'(x) = 3x² - 6x = 3x(x-2)")
print(f"  Critical points: x = 0 and x = 2")

print("\nStep 2: Evaluate Second Derivative at Each Critical Point")
print("  f''(x) = 6x - 6")
print("  f''(0) = 6(0) - 6 = -6")
print("  f''(2) = 6(2) - 6 = 6")

print("\nStep 3: Apply Second Derivative Test")
print("  At x = 0:")
print("    • f''(0) = -6 < 0")
print("    • Therefore, x = 0 is a LOCAL MAXIMUM")
print("    • f(0) = 0³ - 3(0)² = 0")
print("    • Maximum value: (0, 0)")

print("  At x = 2:")
print("    • f''(2) = 6 > 0")
print("    • Therefore, x = 2 is a LOCAL MINIMUM")
print("    • f(2) = 2³ - 3(2)² = 8 - 12 = -4")
print("    • Minimum value: (2, -4)")

print("\n" + "=" * 100)
print("Detailed Analysis at Each Critical Point:")
print("=" * 100)
header = "{:<15} {:<12} {:<15} {:<15} {:<20} {:<30}".format("Point", "x-value", "f(x)", "f''(x)", "Type", "Interpretation")
print("\n" + header)
print("-" * 100)
row1 = "{:<15} {:<12} {:<15.2f} {:<15.2f} {:<20} {:<30}".format("Critical", "0", 0.0, -6.0, "Maximum", "Concave down ∩")
print(row1)
row2 = "{:<15} {:<12} {:<15} {:<15} {:<20} {:<30}".format("Point 1", "", "", "", "", "Slope changes from +ve to -ve")
print(row2)
print()
row3 = "{:<15} {:<12} {:<15.2f} {:<15.2f} {:<20} {:<30}".format("Critical", "2", -4.0, 6.0, "Minimum", "Concave up ∪")
print(row3)
row4 = "{:<15} {:<12} {:<15} {:<15} {:<20} {:<30}".format("Point 2", "", "", "", "", "Slope changes from -ve to +ve")
print(row4)

print("\n" + "=" * 100)
print("Key Takeaways:")
print("=" * 100)
print("1. Second Derivative Test provides a quick way to classify critical points")
print("2. f''(x) > 0 → Concave up → Local MINIMUM")
print("3. f''(x) < 0 → Concave down → Local MAXIMUM")
print("4. f''(x) = 0 → Test inconclusive → Need first derivative test or other analysis")
print("5. This is crucial for optimization problems in machine learning and calculus")
print("=" * 100)
```

```python

# Global vs Local Extrema Visualization
print("\n" + "=" * 100)
print("GLOBAL vs LOCAL EXTREMA")
print("=" * 100)
print("\nOn a bounded domain [−1, 4], we need to check:")
print("1. Local extrema (critical points): x = 0, x = 2")
print("2. Boundary points: x = -1, x = 4")
print("\nFunction values:")
print(f"  f(-1) = (-1)³ - 3(-1)² = -1 - 3 = {f(-1)}")
print(f"  f(0) = 0³ - 3(0)² = {f(0)} (Local maximum)")
print(f"  f(2) = 2³ - 3(2)² = 8 - 12 = {f(2)} (Local minimum)")
print(f"  f(4) = 4³ - 3(4)² = 64 - 48 = {f(4)}")
print(f"\nGLOBAL MAXIMUM on [-1, 4]: x = 4, f(4) = {f(4)}")
print(f"GLOBAL MINIMUM on [-1, 4]: x = 2, f(2) = {f(2)} (also local minimum)")
print("=" * 100)

# Create comparison plots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Local Extrema (Unbounded)
ax_local = axes[0]
x_unbounded = np.linspace(-1, 3.5, 300)
y_unbounded = f(x_unbounded)
ax_local.plot(x_unbounded, y_unbounded, 'b-', linewidth=3, label='$f(x) = x^3 - 3x^2$')
ax_local.plot(0, f(0), 'r*', markersize=30, label='Local Maximum: (0, 0)', zorder=5)
ax_local.plot(2, f(2), 'g*', markersize=30, label='Local Minimum: (2, -4)', zorder=5)
ax_local.axvline(x=0, color='red', linestyle='--', alpha=0.5)
ax_local.axvline(x=2, color='green', linestyle='--', alpha=0.5)
ax_local.axhline(y=f(0), color='red', linestyle=':', alpha=0.5)
ax_local.axhline(y=f(2), color='green', linestyle=':', alpha=0.5)
ax_local.grid(True, alpha=0.3)
ax_local.set_xlabel('x', fontsize=12, fontweight='bold')
ax_local.set_ylabel('f(x)', fontsize=12, fontweight='bold')
ax_local.set_title('Local Extrema (Entire Domain)', fontsize=13, fontweight='bold')
ax_local.legend(fontsize=11, loc='upper left')
ax_local.fill_between(x_unbounded, y_unbounded, alpha=0.1, color='blue')

# Plot 2: Global Extrema (Bounded Domain [-1, 4])
ax_global = axes[1]
x_bounded = np.linspace(-1, 4, 300)
y_bounded = f(x_bounded)
ax_global.plot(x_bounded, y_bounded, 'b-', linewidth=3, label='$f(x) = x^3 - 3x^2$', zorder=2)

# Mark critical points and endpoints
ax_global.plot(-1, f(-1), 'bs', markersize=15, label=f'Boundary: ({-1}, {f(-1)})', zorder=5)
ax_global.plot(0, f(0), 'r*', markersize=30, label=f'Local Max: (0, {f(0)})', zorder=5)
ax_global.plot(2, f(2), 'g*', markersize=30, label=f'Local Min: (2, {f(2)})', zorder=5)
ax_global.plot(4, f(4), 'mo', markersize=15, label=f'Global Max: (4, {f(4)})', zorder=5, markeredgewidth=2, markeredgecolor='purple', fillstyle='none')

# Highlight global extrema
ax_global.axhline(y=f(4), color='purple', linestyle='-.', alpha=0.7, linewidth=2, label='Global Maximum')
ax_global.axhline(y=f(2), color='orange', linestyle='-.', alpha=0.7, linewidth=2, label='Global Minimum')

# Vertical lines for domain bounds
ax_global.axvline(x=-1, color='gray', linestyle='--', alpha=0.5, linewidth=2)
ax_global.axvline(x=4, color='gray', linestyle='--', alpha=0.5, linewidth=2)
ax_global.fill_between(x_bounded, -5, 20, alpha=0.05, color='gray', label='Domain: [-1, 4]')

ax_global.grid(True, alpha=0.3)
ax_global.set_xlabel('x', fontsize=12, fontweight='bold')
ax_global.set_ylabel('f(x)', fontsize=12, fontweight='bold')
ax_global.set_title('Global Extrema on Bounded Domain [-1, 4]', fontsize=13, fontweight='bold')
ax_global.legend(fontsize=10, loc='upper left')
ax_global.set_xlim(-1.5, 4.5)
ax_global.set_ylim(-5, 20)
ax_global.fill_between(x_bounded, y_bounded, alpha=0.1, color='blue')

plt.tight_layout()
plt.suptitle('Comparison: Local Extrema vs Global Extrema', fontsize=14, fontweight='bold', y=1.00)
plt.show()

print("\nSUMMARY TABLE:")
print("=" * 100)
summary_header = "{:<20} {:<12} {:<12} {:<25}".format("Point Type", "x-value", "f(x)", "Classification")
print(summary_header)
print("-" * 100)
print("{:<20} {:<12} {:<12.1f} {:<25}".format("Boundary (left)", "-1", f(-1), "Candidate for global extremum"))
print("{:<20} {:<12} {:<12.1f} {:<25}".format("Critical Point 1", "0", f(0), "Local Maximum (f''(0)<0)"))
print("{:<20} {:<12} {:<12.1f} {:<25}".format("Critical Point 2", "2", f(2), "Local Minimum (f''(2)>0)"))
print("{:<20} {:<12} {:<12.1f} {:<25}".format("Boundary (right)", "4", f(4), "GLOBAL MAXIMUM on [-1,4]"))
print("-" * 100)
print(f"{'GLOBAL MINIMUM':<20} {'2':<12} {f(2):<12.1f} {'At Local Minimum':<25}")
print("=" * 100)
```

## Visualization and Interpretation

The plots above demonstrate how the second derivative test works:

- **Plot 1 (Function $f(x)$)**: Shows the original function with critical points marked. Notice the maximum at $x=0$ (peak) and minimum at $x=2$ (valley).

- **Plot 2 (First Derivative $f'(x)$)**: Displays where the slope is zero (critical points). The first derivative changes sign at critical points.

- **Plot 3 (Second Derivative $f''(x)$)**: Shows the curvature of the function. When $f''(x) > 0$, the function is concave up (curves like ∪) indicating a local minimum. When $f''(x) < 0$, the function is concave down (curves like ∩) indicating a local maximum.

- **Zoomed views at critical points**: The left and right panels zoom into the critical points to show the concavity more clearly. At $x=0$, the curve is concave down (maximum), and at $x=2$, the curve is concave up (minimum).

**Key Insight**: The second derivative test provides a reliable and efficient method to classify critical points without needing to analyze the sign changes of the first derivative.
