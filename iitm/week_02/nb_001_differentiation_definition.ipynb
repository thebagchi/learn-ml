# Differentiation: Definition

## Key Concepts

- **Differentiation**: The process of finding derivatives of a function is called differentiation

- **Derivative**: A derivative is the rate of change of a function with respect to a variable that defines the function

- Let $x$ be a single scalar variable

- If the function $f(x)$ is differentiable, then the differentiation is represented as:

  $\dfrac{df(x)}{dx} \text{ or } f'(x)$

- It is read as differentiation of $f(x)$ with respect to $x$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function f(x) = x^2
def f(x):
    return x**2

# Define the derivative f'(x) = 2x
def f_prime(x):
    return 2 * x

# Create x values
x = np.linspace(-3, 3, 100)
y = f(x)

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the function
ax.plot(x, y, 'b-', linewidth=2, label='$f(x) = x^2$')

# Demonstrate the derivative at specific points
points = [-2, -1, 0, 1, 2]
colors = ['red', 'orange', 'green', 'purple', 'brown']

for point, color in zip(points, colors):
    # Calculate function value and derivative at the point
    y_point = f(point)
    slope = f_prime(point)
    
    # Plot the point on the curve
    ax.plot(point, y_point, 'o', color=color, markersize=8)
    
    # Create tangent line: y - y0 = m(x - x0)
    x_range = np.linspace(point - 1, point + 1, 50)
    y_tangent = y_point + slope * (x_range - point)
    
    # Plot tangent line
    ax.plot(x_range, y_tangent, '--', color=color, linewidth=1.5, 
            label=f"Tangent at $x={point}$, slope=$f'({point})={slope}$")

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.set_title('Geometric Representation of Differentiation\n(Tangent Lines Show the Derivative at Different Points)', fontsize=14)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10, loc='upper center')
ax.set_xlim(-3, 3)
ax.set_ylim(-1, 10)

plt.tight_layout()
plt.show()

print("The derivative represents the slope of the tangent line at each point on the curve.")
print(f"\nFor f(x) = x², the derivative f'(x) = 2x")
print("\nExamples:")
for point in points:
    print(f"  At x = {point}: f({point}) = {f(point)}, f'({point}) = {f_prime(point)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define position function: s(t) = t^3 - 6t^2 + 9t + 5
def position(t):
    return t**3 - 6*t**2 + 9*t + 5

# Define velocity (first derivative): v(t) = 3t^2 - 12t + 9
def velocity(t):
    return 3*t**2 - 12*t + 9

# Define acceleration (second derivative): a(t) = 6t - 12
def acceleration(t):
    return 6*t - 12

# Create time array from 0 to 5 seconds
t = np.linspace(0, 5, 100)
s = position(t)
v = velocity(t)
a = acceleration(t)

# Create subplots
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Plot 1: Position
axes[0].plot(t, s, 'b-', linewidth=2)
axes[0].set_ylabel('Position s(t)', fontsize=12)
axes[0].set_title('$s(t) = t^3 - 6t^2 + 9t + 5$', fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].axhline(y=0, color='k', linestyle='-', linewidth=0.5)

# Plot 2: Velocity (derivative of position)
axes[1].plot(t, v, 'g-', linewidth=2)
axes[1].set_ylabel('Velocity v(t)', fontsize=12)
axes[1].set_title('$v(t) = \\frac{ds}{dt} = 3t^2 - 12t + 9$', fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].axhline(y=0, color='k', linestyle='-', linewidth=0.5)
axes[1].fill_between(t, 0, v, where=(v>=0), alpha=0.3, color='green', label='Moving forward')
axes[1].fill_between(t, 0, v, where=(v<0), alpha=0.3, color='red', label='Moving backward')
axes[1].legend()

# Plot 3: Acceleration (derivative of velocity, second derivative of position)
axes[2].plot(t, a, 'r-', linewidth=2)
axes[2].set_xlabel('Time t (seconds)', fontsize=12)
axes[2].set_ylabel('Acceleration a(t)', fontsize=12)
axes[2].set_title('$a(t) = \\frac{dv}{dt} = 6t - 12$', fontsize=12)
axes[2].grid(True, alpha=0.3)
axes[2].axhline(y=0, color='k', linestyle='-', linewidth=0.5)
axes[2].fill_between(t, 0, a, where=(a>=0), alpha=0.3, color='green', label='Accelerating')
axes[2].fill_between(t, 0, a, where=(a<0), alpha=0.3, color='red', label='Decelerating')
axes[2].legend()

plt.tight_layout()
plt.show()

# Print analysis at specific times
print("=" * 70)
print("Analysis of Motion: Distance, Speed, and Acceleration")
print("=" * 70)
print(f"{'Time(s)':<10} {'Position(m)':<15} {'Velocity(m/s)':<15} {'Accel(m/s²)':<15}")
print("-" * 70)

for time in [0, 1, 2, 3, 4, 5]:
    s_val = position(time)
    v_val = velocity(time)
    a_val = acceleration(time)
    print(f"{time:<10} {s_val:<15.2f} {v_val:<15.2f} {a_val:<15.2f}")

print("\n" + "=" * 70)
print("Key Insights:")
print("=" * 70)
print("• Velocity = 0 at t=1 and t=3 (turning points)")
print("• Velocity < 0 between t=1 and t=3 (moving backward)")
print("• Acceleration = 0 at t=2 (inflection point)")
print("• Acceleration < 0 for t<2 (decelerating)")
print("• Acceleration > 0 for t>2 (accelerating)")
```

## Gradient

The gradient is another term for the derivative of a function. It emphasizes the geometric interpretation:

- The **geometric meaning** of the derivative of a function $f(x)$ is the **slope of the tangent to the curve** $f(x)$ at $x$
  - The slope of the tangent to the curve $f(x)$ at $x$ is nothing but the **gradient** of the curve $f(x)$ at $x$

- **Gradient** is denoted as:

  $\nabla_x f(x) = \dfrac{df(x)}{dx} = f'(x)$
  
  - $\nabla$ is called "nabla" or "del" operator

  - $\nabla_x f(x)$ reads as "gradient of $f(x)$ with respect to $x$"  - The gradient represents the direction and steepness of the steepest ascent on the curve

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 0.5 * x**2 - 2*x + 3

# Define the gradient (derivative)
def gradient(x):
    return x - 2

# Create x values
x = np.linspace(-1, 5, 200)
y = f(x)

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 7))

# Plot the function
ax.plot(x, y, 'b-', linewidth=3, label='$f(x) = 0.5x^2 - 2x + 3$')

# Select points to show gradient
points = [0, 1, 2, 3, 4]
colors = ['red', 'orange', 'green', 'purple', 'brown']

for point, color in zip(points, colors):
    y_point = f(point)
    grad = gradient(point)
    
    # Plot the point
    ax.plot(point, y_point, 'o', color=color, markersize=10)
    
    # Draw tangent line showing gradient
    x_range = np.linspace(point - 1, point + 1, 50)
    y_tangent = y_point + grad * (x_range - point)
    ax.plot(x_range, y_tangent, '--', color=color, linewidth=2, alpha=0.7)
    
    # Add gradient value annotation
    ax.annotate(f'$\\nabla f({point})={grad:.1f}$', 
                xy=(point, y_point), 
                xytext=(10, 15), 
                textcoords='offset points',
                fontsize=10,
                bbox=dict(boxstyle='round,pad=0.3', facecolor=color, alpha=0.3),
                arrowprops=dict(arrowstyle='->', color=color))

ax.set_xlabel('x', fontsize=13)
ax.set_ylabel('f(x)', fontsize=13)
ax.set_title('Gradient: The Slope of the Tangent Line\n$\\nabla_x f(x) = \\frac{df(x)}{dx}$ at Different Points', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=12, loc='upper left')
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 8)

plt.tight_layout()
plt.show()

# Print gradient information
print("=" * 80)
print("Gradient: Slope of the Tangent Line at Different Points")
print("=" * 80)
print(f"\nFor f(x) = 0.5x² - 2x + 3")
print(f"Gradient (Derivative): ∇f(x) = df(x)/dx = x - 2")
print("\n" + "-" * 80)
print(f"{'Point x':<12} {'f(x)':<15} {'Gradient ∇f(x)':<20} {'Interpretation':<30}")
print("-" * 80)

for point in points:
    f_val = f(point)
    grad_val = gradient(point)
    if grad_val > 0:
        interpretation = "Increasing (↗)"
    elif grad_val < 0:
        interpretation = "Decreasing (↘)"
    else:
        interpretation = "Stationary (→)"
    print(f"{point:<12} {f_val:<15.2f} {grad_val:<20.2f} {interpretation:<30}")

print("\n" + "=" * 80)
print("Key Points:")
print("=" * 80)
print("• Gradient = Derivative = Slope of tangent line")
print("• Positive gradient means the curve is increasing at that point")
print("• Negative gradient means the curve is decreasing at that point")
print("• Zero gradient means the curve has a local extremum (max or min)")
print("• The gradient notation ∇ is commonly used in multivariable calculus")
print("=" * 80)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define the function: f(x) = x² - 4x + 3
def f_quad(x):
    return x**2 - 4*x + 3

# Define the gradient (derivative): ∇f(x) = 2x - 4
def gradient_quad(x):
    return 2*x - 4

# Create x values with wider range for better visualization
x_wide = np.linspace(-20, 20, 500)
y_wide = f_quad(x_wide)

# Create figure with larger size
fig = plt.figure(figsize=(16, 7))

# Create two subplots
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

# Plot 1: Function with gradient visualization
ax1.plot(x_wide, y_wide, 'b-', linewidth=3, label='$f(x) = x^2 - 4x + 3$')

# Mark key points with gradients
key_points = [(-10, 'red'), (0, 'orange'), (2, 'green'), (10, 'purple')]
point_values = [-10, 0, 2, 10]

for i, (point, color) in enumerate(key_points):
    y_point = f_quad(point)
    grad = gradient_quad(point)
    
    # Plot point
    ax1.plot(point, y_point, 'o', color=color, markersize=12, zorder=5)
    
    # Add tangent line to show gradient
    x_tangent = np.linspace(point - 3, point + 3, 100)
    y_tangent = y_point + grad * (x_tangent - point)
    ax1.plot(x_tangent, y_tangent, '--', color=color, linewidth=2.5, alpha=0.8)
    
    # Add gradient label
    grad_text = f'∇f({point}) = {grad}'
    slope_type = '+ve' if grad > 0 else ('-ve' if grad < 0 else 'zero')
    ax1.text(point, y_point + 50, grad_text + f'\n({slope_type})', 
             ha='center', fontsize=10, fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.5', facecolor=color, alpha=0.4))

# Mark minimum point specially
ax1.plot(2, f_quad(2), '*', color='green', markersize=25, zorder=6, label='Minimum (∇f = 0)')

ax1.set_xlabel('x', fontsize=13, fontweight='bold')
ax1.set_ylabel('f(x)', fontsize=13, fontweight='bold')
ax1.set_title('Gradient Visualization: $f(x) = x^2 - 4x + 3$\n$\\nabla_x f(x) = 2x - 4$', 
              fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.legend(fontsize=11, loc='upper right')
ax1.set_xlim(-20, 20)
ax1.set_ylim(-100, 400)

# Plot 2: Gradient (derivative) function
x_deriv = np.linspace(-20, 20, 500)
y_deriv = gradient_quad(x_deriv)

ax2.plot(x_deriv, y_deriv, 'r-', linewidth=3, label='$\\nabla_x f(x) = 2x - 4$')
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
ax2.axvline(x=0, color='k', linestyle='-', linewidth=0.8)
ax2.plot(2, 0, '*', color='green', markersize=25, zorder=5, label='Zero gradient at x = 2')

# Shade regions
ax2.fill_between(x_deriv[x_deriv < 2], y_deriv[x_deriv < 2], 0, alpha=0.3, color='red', label='Negative gradient (decreasing)')
ax2.fill_between(x_deriv[x_deriv > 2], y_deriv[x_deriv > 2], 0, alpha=0.3, color='green', label='Positive gradient (increasing)')

ax2.set_xlabel('x', fontsize=13, fontweight='bold')
ax2.set_ylabel('∇f(x)', fontsize=13, fontweight='bold')
ax2.set_title('Gradient Function: $\\nabla_x f(x) = 2x - 4$', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.legend(fontsize=10, loc='upper left')
ax2.set_xlim(-20, 20)
ax2.set_ylim(-50, 40)

plt.tight_layout()
plt.show()

# Create comprehensive table
print("=" * 95)
print("Detailed Gradient Analysis: f(x) = x² - 4x + 3, ∇f(x) = 2x - 4")
print("=" * 95)
print(f"\n{'x':<12} {'f(x)':<15} {'∇ₓf(x)':<15} {'Slope Type':<20} {'Meaning':<35}")
print("-" * 95)

analysis_points = [-10, -5, 0, 1, 2, 3, 5, 10]

for x_val in analysis_points:
    f_val = f_quad(x_val)
    grad_val = gradient_quad(x_val)
    
    if grad_val > 0:
        slope_type = "+ve (increasing)"
        meaning = "Function going up"
    elif grad_val < 0:
        slope_type = "-ve (decreasing)"
        meaning = "Function going down"
    else:
        slope_type = "ZERO"
        meaning = "Critical point (minimum)"
    
    print(f"{x_val:<12} {f_val:<15.1f} {grad_val:<15.1f} {slope_type:<20} {meaning:<35}")

print("\n" + "=" * 95)
print("Key Observations:")
print("=" * 95)
print("1. When x < 2: gradient is negative → function is decreasing")
print("2. When x = 2: gradient is zero → function reaches its minimum value (f(2) = -1)")
print("3. When x > 2: gradient is positive → function is increasing")
print("4. The gradient function is linear: ∇f(x) = 2x - 4 (a straight line)")
print("5. Finding where ∇f(x) = 0 helps locate extrema (maxima and minima)")
print("=" * 95)
```

## Extended Example: Gradient Analysis

Let $f(x) = x^2 - 4x + 3$

The gradient (derivative) is:

$\nabla_x f(x) = \dfrac{df(x)}{dx} = 2x - 4$

**Gradient at specific points:**
- $\nabla_{10} f(x) = 2(10) - 4 = 16$ (strong positive slope)
- $\nabla_{-10} f(x) = 2(-10) - 4 = -24$ (strong negative slope)
- $\nabla_0 f(x) = 2(0) - 4 = -4$ (negative slope)
- $\nabla_2 f(x) = 2(2) - 4 = 0$ (zero slope)

**Key Insight:** When the gradient of the curve $f(x)$ at $x$ is **zero**, then $f(x) = x^2 - 4x + 3$ has a **minimum value** at that $x$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function: f(x) = -x² - 4x + 3 (inverted parabola)
def f_quad_max(x):
    return -x**2 - 4*x + 3

# Define the gradient (derivative): ∇f(x) = -2x - 4
def gradient_quad_max(x):
    return -2*x - 4

# Create x values with wide range
x_wide = np.linspace(-20, 20, 500)
y_wide = f_quad_max(x_wide)

# Create figure with larger size
fig = plt.figure(figsize=(16, 7))

# Create two subplots
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)

# Plot 1: Function with gradient visualization
ax1.plot(x_wide, y_wide, 'b-', linewidth=3, label='$f(x) = -x^2 - 4x + 3$')

# Mark key points with gradients
key_points_max = [(10, 'red'), (0, 'orange'), (-2, 'green'), (-10, 'purple')]
point_values_max = [10, 0, -2, -10]

for i, (point, color) in enumerate(key_points_max):
    y_point = f_quad_max(point)
    grad = gradient_quad_max(point)
    
    # Plot point
    ax1.plot(point, y_point, 'o', color=color, markersize=12, zorder=5)
    
    # Add tangent line to show gradient
    x_tangent = np.linspace(point - 3, point + 3, 100)
    y_tangent = y_point + grad * (x_tangent - point)
    ax1.plot(x_tangent, y_tangent, '--', color=color, linewidth=2.5, alpha=0.8)
    
    # Add gradient label
    grad_text = f'∇f({point}) = {grad}'
    slope_type = '+ve' if grad > 0 else ('-ve' if grad < 0 else 'zero')
    ax1.text(point, y_point + 30, grad_text + f'\n({slope_type})', 
             ha='center', fontsize=10, fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.5', facecolor=color, alpha=0.4))

# Mark maximum point specially
ax1.plot(-2, f_quad_max(-2), '*', color='green', markersize=25, zorder=6, label='Maximum (∇f = 0)')

ax1.set_xlabel('x', fontsize=13, fontweight='bold')
ax1.set_ylabel('f(x)', fontsize=13, fontweight='bold')
ax1.set_title('Gradient Visualization: $f(x) = -x^2 - 4x + 3$\n$\\nabla_x f(x) = -2x - 4$', 
              fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.legend(fontsize=11, loc='lower right')
ax1.set_xlim(-20, 20)
ax1.set_ylim(-500, 100)

# Plot 2: Gradient (derivative) function
x_deriv = np.linspace(-20, 20, 500)
y_deriv = gradient_quad_max(x_deriv)

ax2.plot(x_deriv, y_deriv, 'r-', linewidth=3, label='$\\nabla_x f(x) = -2x - 4$')
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
ax2.axvline(x=0, color='k', linestyle='-', linewidth=0.8)
ax2.plot(-2, 0, '*', color='green', markersize=25, zorder=5, label='Zero gradient at x = -2')

# Shade regions
ax2.fill_between(x_deriv[x_deriv < -2], y_deriv[x_deriv < -2], 0, alpha=0.3, color='green', label='Positive gradient (increasing)')
ax2.fill_between(x_deriv[x_deriv > -2], y_deriv[x_deriv > -2], 0, alpha=0.3, color='red', label='Negative gradient (decreasing)')

ax2.set_xlabel('x', fontsize=13, fontweight='bold')
ax2.set_ylabel('∇f(x)', fontsize=13, fontweight='bold')
ax2.set_title('Gradient Function: $\\nabla_x f(x) = -2x - 4$', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3, linestyle='--')
ax2.legend(fontsize=10, loc='upper right')
ax2.set_xlim(-20, 20)
ax2.set_ylim(-40, 40)

plt.tight_layout()
plt.show()

# Create comprehensive table
print("=" * 100)
print("Detailed Gradient Analysis: f(x) = -x² - 4x + 3, ∇f(x) = -2x - 4 (Downward Parabola - MAXIMUM)")
print("=" * 100)
print(f"\n{'x':<12} {'f(x)':<15} {'∇ₓf(x)':<15} {'Slope Type':<20} {'Meaning':<35}")
print("-" * 100)

analysis_points_max = [-10, -5, -2, -1, 0, 5, 10]

for x_val in analysis_points_max:
    f_val = f_quad_max(x_val)
    grad_val = gradient_quad_max(x_val)
    
    if grad_val > 0:
        slope_type = "+ve (increasing)"
        meaning = "Function going up"
    elif grad_val < 0:
        slope_type = "-ve (decreasing)"
        meaning = "Function going down"
    else:
        slope_type = "ZERO"
        meaning = "Critical point (maximum)"
    
    print(f"{x_val:<12} {f_val:<15.1f} {grad_val:<15.1f} {slope_type:<20} {meaning:<35}")

print("\n" + "=" * 100)
print("Comparison: Minimum vs Maximum")
print("=" * 100)
print("MINIMUM (Example 1: f(x) = x² - 4x + 3):")
print("  • Upward-opening parabola (positive leading coefficient)")
print("  • Gradient changes: negative → zero → positive")
print("  • Critical point at x = 2 is a MINIMUM")
print()
print("MAXIMUM (Example 2: f(x) = -x² - 4x + 3):")
print("  • Downward-opening parabola (negative leading coefficient)")
print("  • Gradient changes: positive → zero → negative")
print("  • Critical point at x = -2 is a MAXIMUM")
print()
print("General Rule: When ∇f(x) = 0, the gradient is zero at critical points.")
print("             • If gradient changes from negative to positive: MINIMUM")
print("             • If gradient changes from positive to negative: MAXIMUM")
print("=" * 100)
```

## Extended Example 2: Gradient Analysis for Maximum

Let $f(x) = -x^2 - 4x + 3$

The gradient (derivative) is:

$\nabla_x f(x) = \dfrac{df(x)}{dx} = -2x - 4$

**Gradient at specific points:**
- $\nabla_{10} f(x) = -2(10) - 4 = -24$ (negative slope)
- $\nabla_{-10} f(x) = -2(-10) - 4 = 16$ (positive slope)
- $\nabla_0 f(x) = -2(0) - 4 = -4$ (negative slope)
- $\nabla_{-2} f(x) = -2(-2) - 4 = 0$ (zero slope)

**Key Insight:** When the gradient of the curve $f(x)$ at $x$ is **zero**, then $f(x) = -x^2 - 4x + 3$ has a **maximum value** at that $x$.
