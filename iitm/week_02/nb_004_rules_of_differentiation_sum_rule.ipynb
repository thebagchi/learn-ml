# Rules of Differentiation: Sum Rule

## Key Concepts

- Let $f(x)$ and $g(x)$ are functions of a single scalar variable $x$

## Sum Rule

- Let $f(x) + g(x)$ be a function,

  $\dfrac{d(f(x)+g(x))}{dx} = \dfrac{df(x)}{dx} + \dfrac{dg(x)}{dx}$

## Example

Let $f(x) = x^3 + x^2 - x$

$\dfrac{df(x)}{dx} = \dfrac{d(x^3+x^2-x)}{dx} = \dfrac{d(x^3)}{dx} + \dfrac{d(x^2)}{dx} + \dfrac{d(-x)}{dx} = 3x^2 + 2x - 1$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-2, 3, 100)

# Define individual functions
def f(x):
    return x**3

def g(x):
    return x**2

# Sum of functions
def f_plus_g(x):
    return f(x) + g(x)

# Derivatives
def f_prime(x):
    return 3*x**2

def g_prime(x):
    return 2*x

def f_plus_g_prime(x):
    return f_prime(x) + g_prime(x)

# Create figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Individual functions and their sum
ax1 = axes[0]
ax1.plot(x, f(x), 'b-', linewidth=2, label='$f(x) = x^3$', alpha=0.7)
ax1.plot(x, g(x), 'g-', linewidth=2, label='$g(x) = x^2$', alpha=0.7)
ax1.plot(x, f_plus_g(x), 'r-', linewidth=2.5, label='$f(x) + g(x) = x^3 + x^2$')
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('Functions: $f(x)$, $g(x)$, and their sum', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax1.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax1.legend(fontsize=11, loc='upper left')
ax1.set_ylim(-5, 15)

# Plot 2: Derivatives
ax2 = axes[1]
ax2.plot(x, f_prime(x), 'b--', linewidth=2, label="$f'(x) = 3x^2$", alpha=0.7)
ax2.plot(x, g_prime(x), 'g--', linewidth=2, label="$g'(x) = 2x$", alpha=0.7)
ax2.plot(x, f_plus_g_prime(x), 'r-', linewidth=2.5, label="$(f+g)'(x) = f'(x) + g'(x) = 3x^2 + 2x$")
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('dy/dx', fontsize=12)
ax2.set_title('Derivatives: Sum of derivatives', fontsize=13, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax2.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
ax2.legend(fontsize=11, loc='upper left')
ax2.set_ylim(-10, 20)

plt.tight_layout()
plt.show()

# Print detailed analysis
print("=" * 80)
print("Sum Rule Demonstration: d(f(x) + g(x))/dx = df(x)/dx + dg(x)/dx")
print("=" * 80)
print("\nGiven:")
print("  f(x) = x³")
print("  g(x) = x²")
print("  h(x) = f(x) + g(x) = x³ + x²")
print("\nDerivatives using Sum Rule:")
print("  f'(x) = 3x²")
print("  g'(x) = 2x")
print("  h'(x) = f'(x) + g'(x) = 3x² + 2x")
print("\n" + "=" * 80)
print("Verification at specific points:")
print("=" * 80)
header = "x        f(x)         g(x)         h(x)         f'(x)        g'(x)        h'(x)"
print(header)
print("-" * 80)

for x_val in [-2, -1, 0, 1, 2]:
    print(f"{x_val:<8} {f(x_val):<12.2f} {g(x_val):<12.2f} {f_plus_g(x_val):<12.2f} {f_prime(x_val):<12.2f} {g_prime(x_val):<12.2f} {f_plus_g_prime(x_val):<12.2f}")

print("\n" + "=" * 80)
print("Key Insight:")
print("=" * 80)
print("• The derivative of the sum (red curve) is exactly the sum of individual derivatives")
print("• Notice: h'(x) = f'(x) + g'(x) at every point")
print("• This makes computing derivatives of complex functions much easier!")
print("=" * 80)
```

## Geometric Representation

The sum rule states that the derivative of a sum of functions equals the sum of their derivatives. This is intuitive from a geometric perspective: if you have a curve that is the sum of two component functions, the slope at any point is the sum of the slopes of the individual components at that point. This makes the sum rule one of the most practical rules in calculus.
