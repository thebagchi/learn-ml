# Unconstrained Optimization

## Introduction

**Unconstrained optimization** is a mathematical process that finds the optimum value of a function without any restrictions on the variables.

### Key Concepts

- **Finding the optimum value** of a function means finding either maximum or minimum values
- **Unconstrained** means there are no restrictions (constraints) on the decision variables
- Let $f(x)$ be a function of a single variable $x$
- An **unconstrained optimization problem** deals with finding the optimal value $x^*$ for which $f(x)$ is optimal

### Problem Formulation

**Minimization Problem:**

$x^* = \min_{x \in \mathbb{R}} f(x)$

where $x^*$ is computed such that for all $x$ near $x^*$, we have $f(x^*) \leq f(x)$

**Maximization Problem:**

$x^* = \max_{x \in \mathbb{R}} f(x)$

where $x^*$ is computed such that for all $x$ near $x^*$, we have $f(x^*) \geq f(x)$

## Methods for Unconstrained Optimization

There are many methods for unconstrained optimization. One of the fundamental approaches is **gradient methods**.

### The Gradient Condition for Optimality

In gradient methods, the condition for optimality is:

$\frac{df(x)}{dx} = 0$

**Interpretation:**

- The gradient of the function at $x$ is given by the derivative of $f(x)$ with respect to $x$
- Derivative becomes zero means the slope of the curve/surface at $x$ becomes zero
- That point $x$ for which gradient is zero corresponds to either a minimum or maximum

### Finding the Optimal Value

The optimal value $x^*$ is obtained by:

1. Computing the derivative of $f(x)$ with respect to $x$
2. Making it equal to zero: $\frac{df(x)}{dx} = 0$
3. Solving for $x$ to get $x^*$

## Second-Order Condition: Determining Min vs Max

Once we find a point where the gradient is zero, we need to determine whether it's a minimum or maximum.

### Using the Second Derivative

To identify the nature of the optimum:

1. Compute the **second-order derivative** and observe whether it's a constant or an expression with $x$
2. If the outcome is an expression with function of $x$, substitute $x^*$ for $x$ to obtain the **second derivative value**

### Classification

- If the **second-order derivative value is positive**, then for $x^*$, $f(x^*)$ is a **minimum**
- If the **second-order derivative value is negative**, then for $x^*$, $f(x^*)$ is a **maximum**

### Mathematical Notation

$\text{Second derivative evaluated at } x^*: \quad \frac{d^2f(x)}{dx^2}\bigg|_{x=x^*}$

## Illustration: Single-Variable Optimization

### Example 1: Finding a Minimum

Let $f(x) = x^2 - 4x + 3$

**Optimization problem:** Find the value for $x^*$ for which the function $f(x) = x^2 - 4x + 3$ is minimum

$$x^* = \min_x f(x)$$

**Step 1: Apply First-Order Condition**

Set the gradient to zero:

$\nabla_x f(x) = \frac{df(x)}{dx} = \frac{d(x^2 - 4x + 3)}{dx} = 2x - 4 = 0$

$2x - 4 = 0 \Rightarrow x^* = 2$

This indicates the gradient is 0 at $x = 2$, meaning the function has a critical point there.

**Step 2: Apply Second-Order Condition**

Compute the second derivative:

$\frac{d^2f(x)}{dx^2} = \frac{d}{dx}\left(\frac{df(x)}{dx}\right) = \frac{d(2x-4)}{dx} = 2$

**Conclusion:** The second derivative is $+2$ (positive), so at $x^* = 2$, the function has a **minimum**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f_min(x):
    return x**2 - 4*x + 3

# First derivative
def df_min(x):
    return 2*x - 4

# Second derivative
def d2f_min(x):
    return 2

# Find optimal point
x_optimal = 2  # From solving df/dx = 0
f_optimal = f_min(x_optimal)

print(f"Minimum Problem: f(x) = x^2 - 4x + 3")
print(f"Optimal point: x* = {x_optimal}")
print(f"Optimal value: f(x*) = {f_optimal}")
print(f"First derivative at x*: df/dx|_x=x* = {df_min(x_optimal)}")
print(f"Second derivative: d2f/dx2 = {d2f_min(x_optimal)} (positive → minimum)")
print()

# Plot
x = np.linspace(-2, 6, 100)
y = f_min(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = x² - 4x + 3')
plt.plot(x_optimal, f_optimal, 'ro', markersize=10, label=f'Minimum at ({x_optimal}, {f_optimal})')
plt.axvline(x_optimal, color='r', linestyle='--', alpha=0.5)
plt.axhline(f_optimal, color='r', linestyle='--', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Unconstrained Optimization: Finding Minimum')
plt.legend()
plt.tight_layout()
plt.show()
```

### Example 2: Finding a Maximum

Let $f(x) = -x^2 - 4x + 3$

**Optimization problem:** Find the value for $x^*$ for which the function $f(x) = -x^2 - 4x + 3$ is maximum

$$x^* = \max_x f(x)$$

**Step 1: Apply First-Order Condition**

Set the gradient to zero:

$\nabla_x f(x) = \frac{df(x)}{dx} = \frac{d(-x^2 - 4x + 3)}{dx} = -2x - 4 = 0$

$-2x - 4 = 0 \Rightarrow x^* = -2$

This indicates the gradient is 0 at $x = -2$.

**Step 2: Apply Second-Order Condition**

Compute the second derivative:

$\frac{d^2f(x)}{dx^2} = \frac{d}{dx}\left(\frac{df(x)}{dx}\right) = \frac{d(-2x-4)}{dx} = -2$

**Conclusion:** The second derivative is $-2$ (negative), so at $x^* = -2$, the function has a **maximum**.

```python
# Define the function for maximum problem
def f_max(x):
    return -x**2 - 4*x + 3

# First derivative
def df_max(x):
    return -2*x - 4

# Second derivative
def d2f_max(x):
    return -2

# Find optimal point
x_optimal = -2  # From solving df/dx = 0
f_optimal = f_max(x_optimal)

print(f"Maximum Problem: f(x) = -x^2 - 4x + 3")
print(f"Optimal point: x* = {x_optimal}")
print(f"Optimal value: f(x*) = {f_optimal}")
print(f"First derivative at x*: df/dx|_x=x* = {df_max(x_optimal)}")
print(f"Second derivative: d2f/dx2 = {d2f_max(x_optimal)} (negative → maximum)")
print()

# Plot
x = np.linspace(-8, 4, 100)
y = f_max(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = -x² - 4x + 3')
plt.plot(x_optimal, f_optimal, 'ro', markersize=10, label=f'Maximum at ({x_optimal}, {f_optimal})')
plt.axvline(x_optimal, color='r', linestyle='--', alpha=0.5)
plt.axhline(f_optimal, color='r', linestyle='--', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Unconstrained Optimization: Finding Maximum')
plt.legend()
plt.tight_layout()
plt.show()
```

## Unconstrained Optimization: Function of Multiple Variables

We know that a function of multiple variables is nothing but a **function of vectors**.

### Problem Setup

- Let $\mathbf{x}$ be a vector of $d$ variables: $\mathbf{x} = [x_1, x_2, \ldots, x_d]^T$
- Unconstrained optimization problem deals with finding the optimum value $\mathbf{x}^*$ of a function $f(\mathbf{x})$ of $d$ variables

### First-Order Optimality Condition

The optimal values $\mathbf{x}^* = [x_1^*, x_2^*, \ldots, x_d^*]^T$ are obtained by computing the gradient of function $f(\mathbf{x})$ with respect to $\mathbf{x}$ and solving by equating to zero:

$\frac{d f(\mathbf{x})}{d\mathbf{x}} = \begin{bmatrix} \frac{\partial f(\mathbf{x})}{\partial x_1} \\ \frac{\partial f(\mathbf{x})}{\partial x_2} \\ \vdots \\ \frac{\partial f(\mathbf{x})}{\partial x_d} \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \\ \vdots \\ 0 \end{bmatrix} = \mathbf{0}$

Which gives us $d$ equations:

$\frac{\partial f(\mathbf{x})}{\partial x_1} = 0$

$\frac{\partial f(\mathbf{x})}{\partial x_2} = 0$

$\vdots$

$\frac{\partial f(\mathbf{x})}{\partial x_d} = 0$

**Important Note:** Since there are no constraints on the values of $\mathbf{x}^*$, we have $\mathbf{x}^* = [x_1^*, x_2^*, \ldots, x_d^*]^T$ can take any values. This is why it's called an **unconstrained optimization problem**.

## Second-Order Condition for Multiple Variables: The Hessian

For functions of multiple variables, the second-order condition uses the **Hessian matrix** (matrix of second partial derivatives).

### The Hessian Matrix

The Hessian matrix is defined as:

$\mathbf{H}(\mathbf{x}) = \begin{bmatrix} \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_1 \partial x_d} \\ \frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} & \cdots & \frac{\partial^2 f}{\partial x_2 \partial x_d} \\ \vdots & \vdots & \ddots & \vdots \\ \frac{\partial^2 f}{\partial x_d \partial x_1} & \frac{\partial^2 f}{\partial x_d \partial x_2} & \cdots & \frac{\partial^2 f}{\partial x_d^2} \end{bmatrix}$

### Classification Using Hessian

- If $\mathbf{H}(\mathbf{x}^*)$ is **positive definite** (all eigenvalues positive), then $\mathbf{x}^*$ is a **local minimum**
- If $\mathbf{H}(\mathbf{x}^*)$ is **negative definite** (all eigenvalues negative), then $\mathbf{x}^*$ is a **local maximum**
- If $\mathbf{H}(\mathbf{x}^*)$ is **indefinite** (mixed sign eigenvalues), then $\mathbf{x}^*$ is a **saddle point**

## Practical Example: Function of Two Variables

Let $f(x_1, x_2) = x_1^2 + 2x_2^2 - 4x_1 - 8x_2 + 5$

### Step 1: Compute Partial Derivatives (First-Order Condition)

$\frac{\partial f}{\partial x_1} = 2x_1 - 4 = 0 \Rightarrow x_1^* = 2$

$\frac{\partial f}{\partial x_2} = 4x_2 - 8 = 0 \Rightarrow x_2^* = 2$

So the critical point is $\mathbf{x}^* = [2, 2]^T$

### Step 2: Compute Hessian Matrix (Second-Order Condition)

$\mathbf{H} = \begin{bmatrix} \frac{\partial^2 f}{\partial x_1^2} & \frac{\partial^2 f}{\partial x_1 \partial x_2} \\ \frac{\partial^2 f}{\partial x_2 \partial x_1} & \frac{\partial^2 f}{\partial x_2^2} \end{bmatrix} = \begin{bmatrix} 2 & 0 \\ 0 & 4 \end{bmatrix}$

Both eigenvalues are positive (2 and 4), so the Hessian is **positive definite**.

**Conclusion:** $\mathbf{x}^* = [2, 2]^T$ is a **local minimum** with $f(\mathbf{x}^*) = -7$

```python
# Define the function of two variables
def f_2d(x1, x2):
    return x1**2 + 2*x2**2 - 4*x1 - 8*x2 + 5

# Partial derivatives
def df_dx1(x1, x2):
    return 2*x1 - 4

def df_dx2(x1, x2):
    return 4*x2 - 8

# Hessian matrix
def hessian_2d(x1, x2):
    return np.array([[2, 0], [0, 4]])

# Find optimal point
x1_opt = 2
x2_opt = 2
f_opt = f_2d(x1_opt, x2_opt)

print("Function of Two Variables Example")
print(f"f(x1, x2) = x1^2 + 2x2^2 - 4x1 - 8x2 + 5")
print()
print(f"Optimal point: x* = [{x1_opt}, {x2_opt}]")
print(f"Optimal value: f(x*) = {f_opt}")
print(f"∂f/∂x1 at x* = {df_dx1(x1_opt, x2_opt)}")
print(f"∂f/∂x2 at x* = {df_dx2(x1_opt, x2_opt)}")
print()

# Hessian analysis
H = hessian_2d(x1_opt, x2_opt)
eigenvalues = np.linalg.eigvalsh(H)
print(f"Hessian matrix:")
print(H)
print(f"Eigenvalues: {eigenvalues}")
print(f"All eigenvalues positive: {all(eigenvalues > 0)}")
print(f"→ Positive definite → LOCAL MINIMUM")
print()
```

```python
# Visualize the 2D function
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x1 = np.linspace(0, 4, 50)
x2 = np.linspace(0, 4, 50)
X1, X2 = np.meshgrid(x1, x2)
Z = f_2d(X1, X2)

# 3D plot
fig = plt.figure(figsize=(12, 5))

# Surface plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, Z, cmap='viridis', alpha=0.8)
ax1.scatter([x1_opt], [x2_opt], [f_opt], color='red', s=100, label='Minimum')
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('f(x1, x2)')
ax1.set_title('3D Surface Plot')
ax1.legend()

# Contour plot
ax2 = fig.add_subplot(122)
contours = ax2.contour(X1, X2, Z, levels=15, cmap='viridis')
ax2.clabel(contours, inline=True, fontsize=8)
ax2.plot(x1_opt, x2_opt, 'r*', markersize=15, label='Minimum')
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_title('Contour Plot')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Role of Optimization in Machine Learning

### Why Optimization Matters in ML

Most machine learning techniques involve either **maximization** or **minimization** of a function. In the context of machine learning, such a function is called an **objective function**.

### Types of Objective Functions

Depending on the different techniques and goals, the objective function may be:

- **Cost function** — in supervised learning problems
- **Error function** — measures prediction errors
- **Criteria function** — defines model performance
- **Likelihood function** — in probabilistic models
- **Loss function** — generic term for what we're optimizing

### The ML Optimization Pipeline

1. **Define the objective function** — formalize what you want to optimize
2. **Set up the optimization problem** — minimize cost/loss or maximize likelihood
3. **Apply optimization methods** — use gradient-based methods to find parameters
4. **Evaluate the solution** — check if the found parameters lead to good predictions

### Key Insight

Most ML algorithms are essentially **optimization algorithms** that find the best set of parameters (weights, biases, etc.) by minimizing a loss function or maximizing a likelihood function. Understanding optimization is fundamental to understanding how ML models learn from data.

## Summary

### Key Takeaways

1. **Unconstrained optimization** finds minimum or maximum of a function with no restrictions

2. **First-order condition** (gradient = 0) identifies critical points:
   - Single variable: $\frac{df}{dx} = 0$
   - Multiple variables: $\nabla f(\mathbf{x}) = \mathbf{0}$

3. **Second-order condition** determines the nature of critical points:
   - Single variable: Check sign of $\frac{d^2f}{dx^2}$ (positive → min, negative → max)
   - Multiple variables: Check eigenvalues of Hessian (positive definite → min, negative definite → max)

4. **Gradient methods** are fundamental algorithms that repeatedly move in the direction of negative gradient to find minima

5. **ML connection**: Most ML algorithms are optimization problems where we minimize a loss function or maximize a likelihood function

6. The entire field of optimization is essential to understanding how neural networks, regression models, and other ML techniques learn from data
