# Partial Derivatives: Derivatives of Functions with Multiple Variables

## Introduction to Functions of Multiple Variables

### From Single Variable to Multiple Variables

Until now, we've studied functions of **one variable**:
- $f(x)$ depends on a single input $x$
- The derivative $f'(x)$ tells us the rate of change with respect to $x$

But in real-world applications, functions often depend on **multiple variables**:
- Machine learning models depend on many input features
- Physical phenomena depend on multiple coordinates (x, y, z)
- Economic models depend on multiple factors

### Example: Function of Multiple Variables

Let $p$, $q$, and $r$ be three scalar variables.

**Definition:** A function of these three variables is written as:

$f(p, q, r)$

**Example:**

$f(p, q, r) = p^3 + 2q - 4r^2 + 10$

This is a single-output function that takes three inputs.

### Representing Multiple Variables as a Vector

Then:

$\mathbf{x} = \begin{bmatrix} x_1 \\ x_2 \\ x_3 \end{bmatrix}$

Now we can write:

$f(p, q, r) = f(x_1, x_2, x_3) = f(\mathbf{x})$

**Important:** A function of multiple variables is equivalent to a **function of a vector**:
- **Input:** Vector $\mathbf{x} \in \mathbb{R}^d$ (where $d$ is the number of variables)
- **Output:** Scalar $y = f(\mathbf{x})$

This is the standard representation in **machine learning and calculus**!

## Partial Derivatives: Definition and Intuition

### What is a Partial Derivative?

When we have a function of multiple variables, we want to know:
- **How does the function change when we vary one variable while keeping others fixed?**

This is exactly what a **partial derivative** measures.

### Formal Definition

The **partial derivative** of $f(\mathbf{x})$ with respect to variable $x_i$ is:

$\dfrac{\partial f(\mathbf{x})}{\partial x_i} = \lim_{h \to 0} \dfrac{f(x_1, ..., x_i + h, ..., x_d) - f(x_1, ..., x_i, ..., x_d)}{h}$

**Key Point:** We change only $x_i$ by a small amount $h$, while keeping all other variables constant.

### Computing Partial Derivatives: The Technique

**To compute $\frac{\partial f}{\partial x_i}$:**
1. Treat all other variables as **constants**
2. Differentiate with respect to $x_i$ using standard derivative rules
3. Result is a new function of all variables

**Example:** For $f(x_1, x_2) = 2x_1^2 + 3x_2^2 + 5x_1x_2 - x_1 + x_2 + 6$

$\dfrac{\partial f}{\partial x_1} = 4x_1 + 5x_2 - 1$

Partial derivative with respect to $x_2$ (treating $x_1$ as constant):

$\dfrac{\partial f}{\partial x_2} = 6x_2 + 5x_1 + 1$

## Partial Derivatives: From Theory to Practice

### How to Compute Partial Derivatives Systematically

For any function $f(x_1, x_2, ..., x_d)$:
1. To find $\frac{\partial f}{\partial x_i}$, treat **all other variables as constants**
2. Differentiate with respect to $x_i$ using standard derivative rules
3. Repeat for each variable

### Key Distinction: Partial vs. Total

- **Partial derivative $\frac{\partial f}{\partial x_i}$:** Rate of change when **only** $x_i$ varies (others fixed)
- **Total derivative:** How $f$ changes when **all** variables change together

### Next Step: The Gradient Vector

All partial derivatives can be combined into a single **gradient vector** $\nabla f$:

$\nabla f(\mathbf{x}) = \begin{bmatrix} \dfrac{\partial f}{\partial x_1} \\ \dfrac{\partial f}{\partial x_2} \\ \vdots \\ \dfrac{\partial f}{\partial x_d} \end{bmatrix}$

## Examples: Computing Partial Derivatives

```python
import numpy as np

# Example 1: f(x1, x2) = 2x1^2 + 3x2^2 + 5*x1*x2 - x1 + x2 + 6
print("="*100)
print("EXAMPLE 1: Computing Partial Derivatives")
print("="*100)
print("\nFunction: f(x₁, x₂) = 2x₁² + 3x₂² + 5x₁x₂ - x₁ + x₂ + 6")
print("\nStep 1: Find ∂f/∂x₁ (treat x₂ as constant)")
print("  ∂f/∂x₁ = 4x₁ + 5x₂ - 1")

print("\nStep 2: Find ∂f/∂x₂ (treat x₁ as constant)")
print("  ∂f/∂x₂ = 6x₂ + 5x₁ + 1")

print("\nStep 3: Evaluate at different points")
print(f"\n{'Point (x₁, x₂)':<20} {'∂f/∂x₁':<20} {'∂f/∂x₂':<20} {'Interpretation':<30}")
print("-" * 90)

def df_dx1(x1, x2):
    return 4*x1 + 5*x2 - 1

def df_dx2(x1, x2):
    return 6*x2 + 5*x1 + 1

points = [(0, 0), (1, 0), (0, 1), (1, 1), (-1, 2)]

for x1, x2 in points:
    df1 = df_dx1(x1, x2)
    df2 = df_dx2(x1, x2)
    interpretation = "Large change both directions" if abs(df1) > 5 or abs(df2) > 5 else "Moderate change"
    print(f"({x1:2}, {x2:2})          {df1:10.2f}           {df2:10.2f}          {interpretation}")

print("\n" + "="*100)
print("EXAMPLE 2: Three Variables")
print("="*100)
print("\nFunction: g(x, y, z) = x² + 2y² + 3z² - xy + xz")
print("\nPartial derivatives:")
print("  ∂g/∂x = 2x - y + z")
print("  ∂g/∂y = 4y - x")
print("  ∂g/∂z = 6z + x")

print("\nAt point (1, 2, -1):")
x, y, z = 1, 2, -1
dg_dx = 2*x - y + z
dg_dy = 4*y - x
dg_dz = 6*z + x
print(f"  ∂g/∂x = {dg_dx}")
print(f"  ∂g/∂y = {dg_dy}")
print(f"  ∂g/∂z = {dg_dz}")
print("\n  Interpretation:")
print(f"    - Function most sensitive to y (|∂g/∂y| = {abs(dg_dy)})")
print(f"    - Function slightly sensitive to x and z")
```

## Verification Exercise

Let's verify our partial derivative calculations with another example and look at how they behave at different points:

```python
# f(x1, x2) = 2x1^2 + 3x2^2 + 5*x1*x2 - x1 + x2 + 6
def f(x1, x2):
    return 2*x1**2 + 3*x2**2 + 5*x1*x2 - x1 + x2 + 6

def df_dx1(x1, x2):
    return 4*x1 + 5*x2 - 1

def df_dx2(x1, x2):
    return 6*x2 + 5*x1 + 1

print("\n" + "="*80)
print("VERIFICATION: Partial Derivatives at Various Points")
print("="*80)
print(f"\n{'Point':<15} {'f(x)':<15} {'∂f/∂x₁':<15} {'∂f/∂x₂':<15}")
print("-" * 80)

points = [(0, 0), (1, 1), (2, -1), (-1, 2)]

for x1, x2 in points:
    f_val = f(x1, x2)
    df1 = df_dx1(x1, x2)
    df2 = df_dx2(x1, x2)
    print(f"({x1:2}, {x2:2})       {f_val:8.2f}      {df1:8.2f}      {df2:8.2f}")

print("\nInterpretation:")
print("- When ∂f/∂x₁ is large and positive: f increases rapidly as x₁ increases")
print("- When ∂f/∂x₁ is negative: f decreases as x₁ increases (move opposite direction)")
print("- When ∂f/∂x₁ ≈ 0: f is relatively insensitive to x₁ at that point")
```

## Applications of Partial Derivatives

### 1. Machine Learning and Optimization
- **Sensitivity analysis:** Which input affects output most?
- **Feature importance:** Magnitude of partial derivatives shows sensitivity
- **Backpropagation:** Computes partial derivatives of loss w.r.t. all weights

### 2. Physics and Engineering
- **Heat equation:** Partial derivatives describe temperature change over time and space
- **Wave equation:** Relates spatial and temporal partial derivatives
- **Fluid dynamics:** Partial derivatives of velocity field describe flow

### 3. Economics and Finance
- **Elasticity:** Partial derivatives measure responsiveness to input changes
- **Marginal analysis:** How profit/revenue changes with each factor
- **Sensitivity testing:** Impact of parameter changes on outcomes

### 4. Multivariable Calculus Foundations
- **Critical points:** Solve system $\frac{\partial f}{\partial x_i} = 0$ for all $i$
- **Level surfaces:** Partial derivatives perpendicular to these surfaces
- **Implicit functions:** Find relationships between variables

---

## Summary: Partial Derivatives Essentials

| Concept | Description |
|---------|-------------|
| **Definition** | Rate of change with one variable fixed, others constant |
| **Notation** | $\frac{\partial f}{\partial x_i}$ or $f_{x_i}$ |
| **Computation** | Treat other variables as constants, differentiate normally |
| **Physical meaning** | Sensitivity of output to changes in variable $x_i$ |
| **At extrema** | All partial derivatives = 0 at critical points |
| **Comparison to single-variable** | Similar rules apply (power, product, chain, etc.) |
| **Next step** | Combine all partial derivatives into **gradient vector** ∇f |
