# Gradient Vector: Direction of Steepest Increase

## What is a Gradient Vector?

### Definition

Given a function $f(\mathbf{x}) = f(x_1, x_2, ..., x_d)$ of $d$ variables, the **gradient vector** is defined as:

$\nabla f(\mathbf{x}) = \begin{bmatrix} \dfrac{\partial f}{\partial x_1} \\ \dfrac{\partial f}{\partial x_2} \\ \vdots \\ \dfrac{\partial f}{\partial x_d} \end{bmatrix}$

**Key Properties:**
- **Vector-valued function:** Input is a point $\mathbf{x}$, output is a vector $\nabla f(\mathbf{x})$
- **Dimensionality:** If $f: \mathbb{R}^d \to \mathbb{R}$, then $\nabla f: \mathbb{R}^d \to \mathbb{R}^d$
- **Components:** Each component is a partial derivative

### Notation

**Different ways to write the gradient:**
- $\nabla f(\mathbf{x})$ (nabla symbol, most common)
- $\text{grad } f(\mathbf{x})$ (word form)
- $Df(\mathbf{x})$ (differential notation)
- $\frac{df}{d\mathbf{x}}$ (derivative with respect to vector)

All of these mean the same thing!

## Geometric Interpretation: The Gradient Points Uphill

### The Fundamental Theorem of Gradients

The gradient vector has a beautiful geometric meaning:

**$\nabla f(\mathbf{x})$ points in the direction of the steepest increase of $f$ at point $\mathbf{x}$**

### What This Means

Imagine you're standing on a hill represented by $f(\mathbf{x})$:
- **Direction:** Look at $\nabla f(\mathbf{x})$ - this is the direction you should walk to go **uphill the fastest**
- **Steepness:** The magnitude $||\nabla f(\mathbf{x})||$ tells you **how steep** the hill is in that direction
- **Opposite direction:** To go downhill fastest (toward valley/minimum), walk in direction of $-\nabla f(\mathbf{x})$
- **Perpendicular:** Directions perpendicular to $\nabla f$ correspond to **level paths** (constant $f$)

### Level Curves and Gradient

**Level curve (contour line):** Set of all points where $f(\mathbf{x}) = c$ for some constant $c$

**Key fact:** The gradient is **always perpendicular to level curves**

**Implication:** 
- Moving along a level curve keeps $f$ constant
- Moving perpendicular to level curves (in direction of $\nabla f$) changes $f$ fastest

## Computing the Gradient: Step by Step

### Algorithm to Compute Gradient

**For a function $f(x_1, x_2, ..., x_d)$:**

1. Compute partial derivative w.r.t. $x_1$: $\frac{\partial f}{\partial x_1}$
2. Compute partial derivative w.r.t. $x_2$: $\frac{\partial f}{\partial x_2}$
3. ... continue for all variables ...
4. Stack all partial derivatives into a column vector
5. Result is $\nabla f(\mathbf{x})$

### Example 1: Simple Quadratic

**Function:** $f(x, y) = x^2 + y^2$

**Partial derivatives:**

$\dfrac{\partial f}{\partial x} = 2x$

$\dfrac{\partial f}{\partial y} = 2y$

**Gradient:**

$\nabla f(x, y) = \begin{bmatrix} 2x \\ 2y \end{bmatrix}$

**At point $(1, 2)$:**

$\nabla f(1, 2) = \begin{bmatrix} 2 \\ 4 \end{bmatrix}$

This points in direction of $(1, 2)$ with magnitude $\sqrt{4 + 16} = \sqrt{20} \approx 4.47$

### Example 2: Mixed Terms

**Function:** $f(x, y) = x^2 + xy + y^2 - x + 2y$

**Partial derivatives:**

$\dfrac{\partial f}{\partial x} = 2x + y - 1$

$\dfrac{\partial f}{\partial y} = x + 2y + 2$

**Gradient:**

$\nabla f(x, y) = \begin{bmatrix} 2x + y - 1 \\ x + 2y + 2 \end{bmatrix}$

**At point $(1, 0)$:**

$\nabla f(1, 0) = \begin{bmatrix} 2(1) + 0 - 1 \\ 1 + 2(0) + 2 \end{bmatrix} = \begin{bmatrix} 1 \\ 3 \end{bmatrix}$

## Key Properties of the Gradient Vector

```python
import numpy as np
import matplotlib.pyplot as plt

# Example function: f(x, y) = x^2 + y^2
def f_quadratic(x, y):
    return x**2 + y**2

def grad_quadratic(x, y):
    """Gradient of f at point (x, y)"""
    return np.array([2*x, 2*y])

# Example function: f(x, y) = -x^2 - 0.5*y^2 + 3
def f_bowl(x, y):
    return -x**2 - 0.5*y**2 + 3

def grad_bowl(x, y):
    """Gradient of f at point (x, y)"""
    return np.array([-2*x, -y])

print("="*100)
print("GRADIENT VECTOR PROPERTIES")
print("="*100)

print("\n1. GRADIENT POINTS IN DIRECTION OF STEEPEST INCREASE")
print("-" * 100)
print("\nFunction: f(x, y) = x² + y²")
print("\nAt various points:")
print(f"\n{'Point (x,y)':<20} {'Gradient ∇f':<30} {'Magnitude ||∇f||':<20} {'Normalized direction':<30}")
print("-" * 100)

test_points = [(0, 0), (1, 0), (0, 1), (1, 1), (2, 1), (-1, 2)]

for x, y in test_points:
    grad = grad_quadratic(x, y)
    mag = np.linalg.norm(grad)
    if mag > 0:
        direction = grad / mag
        print(f"({x:2}, {y:2})           [{grad[0]:6.1f}, {grad[1]:6.1f}]      {mag:10.4f}         [{direction[0]:6.3f}, {direction[1]:6.3f}]")
    else:
        print(f"({x:2}, {y:2})           [{grad[0]:6.1f}, {grad[1]:6.1f}]      {mag:10.4f}         [  ZERO - NO DIRECTION  ]")

print("\n" + "="*100)
print("2. ZERO GRADIENT AT EXTREMA (MINIMA/MAXIMA)")
print("="*100)
print("\nAt minimum point (0, 0): ∇f(0, 0) = [0, 0]ᵀ")
print("  Interpretation: No direction of increase - we're at a minimum!")
print("\nAt any other point: ∇f ≠ 0")
print("  Interpretation: We can move in direction of ∇f to increase f")

print("\n" + "="*100)
print("3. MAGNITUDE INCREASES WITH STEEPNESS")
print("="*100)
print("\nFunction: f(x, y) = x² + y² (paraboloid shape)")
print("\nPoints along line y = x:")
print(f"\n{'Distance from origin':<25} {'Gradient magnitude':<20} {'Interpretation':<30}")
print("-" * 100)

for r in [0.5, 1.0, 1.5, 2.0, 2.5]:
    x = r / np.sqrt(2)
    y = r / np.sqrt(2)
    grad = grad_quadratic(x, y)
    mag = np.linalg.norm(grad)
    print(f"{r:<25.2f} {mag:<20.4f} {'Steep' if mag > 3 else 'Moderate' if mag > 1.5 else 'Gentle'}")

print("\nConclusion: Points farther from origin have LARGER gradients (steeper slopes)")

print("\n" + "="*100)
print("4. NEGATIVE GRADIENT POINTS DOWNHILL")
print("="*100)
print("\nTo MINIMIZE f: Move in direction of -∇f (negative gradient)")
print("To MAXIMIZE f: Move in direction of ∇f (positive gradient)")

print("\nExample: At point (2, 3)")
grad = grad_quadratic(2, 3)
print(f"  ∇f(2, 3) = {grad}")
print(f"  Moving direction: (2, 3) → go UPHILL, f increases")
print(f"  Moving direction: (2, 3) → go DOWNHILL, f decreases")
print(f"  Moving opposite (-∇f): (-2, -3) → go DOWNHILL toward minimum at (0, 0)")

print("\n" + "="*100)
```

## Visualization: Gradient Vector Field

```python
# Create comprehensive visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Create mesh grid
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)
Z_quad = f_quadratic(X, Y)
Z_bowl = f_bowl(X, Y)

# Subplot 1: Contour plot with gradient vectors (minimization)
ax = axes[0]
levels1 = np.linspace(0, 15, 15)
contour1 = ax.contour(X, Y, Z_quad, levels=levels1, colors='gray', alpha=0.5, linewidths=0.5)
ax.clabel(contour1, inline=True, fontsize=7)

# Create gradient field
x_grad = np.linspace(-2.5, 2.5, 15)
y_grad = np.linspace(-2.5, 2.5, 15)
X_grad, Y_grad = np.meshgrid(x_grad, y_grad)
U = 2*X_grad  # df/dx
V = 2*Y_grad  # df/dy

# Normalize for visualization
Grad_mag = np.sqrt(U**2 + V**2)
U_norm = U / (Grad_mag + 0.1)
V_norm = V / (Grad_mag + 0.1)

quiver1 = ax.quiver(X_grad, Y_grad, U_norm, V_norm, Grad_mag, cmap='hot', 
                    scale=25, width=0.005, pivot='middle')
ax.plot(0, 0, 'g*', markersize=20, label='Minimum (∇f = 0)', zorder=5)
ax.set_xlabel('x', fontsize=12, fontweight='bold')
ax.set_ylabel('y', fontsize=12, fontweight='bold')
ax.set_title('Gradient Field: f(x,y) = x² + y²\nArrows point to steepest increase', fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='upper right')
ax.grid(True, alpha=0.3)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
plt.colorbar(quiver1, ax=ax, label='||∇f||')

# Subplot 2: Contour plot for bowl (maximization)
ax = axes[1]
levels2 = np.linspace(-3, 3, 15)
contour2 = ax.contour(X, Y, Z_bowl, levels=levels2, colors='gray', alpha=0.5, linewidths=0.5)
ax.clabel(contour2, inline=True, fontsize=7)

# Gradient field for bowl
U_bowl = -2*X_grad
V_bowl = -Y_grad
Grad_mag_bowl = np.sqrt(U_bowl**2 + V_bowl**2)
U_norm_bowl = U_bowl / (Grad_mag_bowl + 0.1)
V_norm_bowl = V_bowl / (Grad_mag_bowl + 0.1)

quiver2 = ax.quiver(X_grad, Y_grad, U_norm_bowl, V_norm_bowl, Grad_mag_bowl, 
                    cmap='hot', scale=25, width=0.005, pivot='middle')
ax.plot(0, 0, 'r*', markersize=20, label='Maximum (∇f = 0)', zorder=5)
ax.set_xlabel('x', fontsize=12, fontweight='bold')
ax.set_ylabel('y', fontsize=12, fontweight='bold')
ax.set_title('Gradient Field: f(x,y) = -x² - 0.5y² + 3\nArrows point AWAY from maximum', fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='upper right')
ax.grid(True, alpha=0.3)
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
plt.colorbar(quiver2, ax=ax, label='||∇f||')

# Subplot 3: Gradient descent path
ax = axes[2]

# Plot contours
levels3 = np.linspace(0, 15, 15)
contour3 = ax.contour(X, Y, Z_quad, levels=levels3, colors='gray', alpha=0.4, linewidths=0.5)
ax.clabel(contour3, inline=True, fontsize=7)

# Gradient descent starting from point (3, 2.5)
def gradient_descent(x0, y0, learning_rate=0.1, num_steps=50):
    """Perform gradient descent"""
    path = [(x0, y0)]
    x, y = x0, y0
    
    for _ in range(num_steps):
        grad = grad_quadratic(x, y)
        x_new = x - learning_rate * grad[0]
        y_new = y - learning_rate * grad[1]
        
        if np.linalg.norm([x_new - x, y_new - y]) < 1e-6:
            break
        
        x, y = x_new, y_new
        path.append((x, y))
    
    return np.array(path)

# Run gradient descent from different starting points
colors_path = ['red', 'blue', 'green']
starts = [(3, 2.5), (-2.5, 3), (2.5, -2.5)]

for start, color in zip(starts, colors_path):
    path = gradient_descent(start[0], start[1], learning_rate=0.15, num_steps=30)
    ax.plot(path[:, 0], path[:, 1], 'o-', color=color, linewidth=2, markersize=4, 
           label=f'Start: {start}', alpha=0.7)
    ax.plot(start[0], start[1], 's', color=color, markersize=10, markeredgecolor='black', 
           markeredgewidth=1.5, zorder=5)

ax.plot(0, 0, 'y*', markersize=25, markeredgecolor='black', markeredgewidth=1.5, 
       label='Minimum', zorder=5)
ax.set_xlabel('x', fontsize=12, fontweight='bold')
ax.set_ylabel('y', fontsize=12, fontweight='bold')
ax.set_title('Gradient Descent: Following -∇f to Minimum', fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='upper right')
ax.grid(True, alpha=0.3)
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-3.5, 3.5)

plt.suptitle('Gradient Vectors: Direction of Steepest Change', fontsize=14, fontweight='bold', y=1.00)
plt.tight_layout()
plt.show()

print("\nVisualization shows:")
print("1. LEFT: Gradient field for bowl minimum - arrows point outward")
print("2. MIDDLE: Gradient field for inverted bowl maximum - arrows point inward")
print("3. RIGHT: Gradient descent paths converging to minimum by following -∇f")
```

## Applications: Where Gradients Matter

```python
print("="*100)
print("GRADIENT VECTOR APPLICATIONS")
print("="*100)

print("\n1. GRADIENT DESCENT (Machine Learning Core Algorithm)")
print("-" * 100)
print("""
Problem: Minimize loss function L(w) where w are neural network weights

Solution: Iteratively update weights in direction of -∇L(w)
  w_new = w_old - α * ∇L(w_old)
  
Where:
  α = learning rate (step size)
  ∇L(w_old) = gradient of loss at current weights
  -∇L(w_old) = direction of steepest DECREASE (opposite of ∇L)

Effect: Each iteration moves toward local minimum of loss function
Result: Eventually w converges to values that minimize error
""")

print("\n2. FEATURE IMPORTANCE / SENSITIVITY")
print("-" * 100)
print("""
Problem: Which input variable affects output most?

Solution: Look at components of ∇f
  ∇f(x₁, x₂, x₃, x₄) = [2.5, 0.1, -3.2, 0.3]ᵀ
  
Interpretation:
  - ∂f/∂x₁ = 2.5: Large sensitivity to x₁
  - ∂f/∂x₂ = 0.1: Small sensitivity to x₂  
  - ∂f/∂x₃ = -3.2: Large sensitivity to x₃ (inverse relationship)
  - ∂f/∂x₄ = 0.3: Medium sensitivity to x₄

Conclusion: x₃ and x₁ are most important, x₂ least important
""")

print("\n3. OPTIMAL CONTROL / NAVIGATION")
print("-" * 100)
print("""
Problem: Navigate from current state to best state

Solution: Always move in direction of ∇f (or -∇f for minimization)
  
Examples:
  - Robot navigation: ∇f points toward goal
  - Portfolio optimization: ∇f shows how to adjust holdings
  - Temperature control: ∇f shows adjustment needed
  - Parameter tuning: ∇f shows which parameters to adjust
""")

print("\n4. VISUALIZATION / UNDERSTANDING FUNCTION BEHAVIOR")
print("-" * 100)
print("""
Problem: Understand how complex function behaves

Solution: Plot gradient vector field
  - Shows local behavior everywhere
  - Arrows indicate flow direction
  - Length indicates steepness
  - Helps identify minima, maxima, saddle points
""")

print("\n" + "="*100)
print("PRACTICAL EXAMPLE: Linear Regression")
print("="*100)
print("""
Objective: Find best-fit line y = mx + b for data points

Loss function: L(m, b) = Σ(y_i - (mx_i + b))²

Gradient:
  ∂L/∂m = -2Σ x_i(y_i - (mx_i + b))
  ∂L/∂b = -2Σ(y_i - (mx_i + b))
  ∇L = [∂L/∂m, ∂L/∂b]ᵀ

Gradient Descent Update:
  m_new = m_old - α * ∂L/∂m
  b_new = b_old - α * ∂L/∂b
  
Result: Iteratively refine m and b until loss is minimized
""")

print("="*100)
```

## Summary: Gradient Vector Quick Reference

### Definition and Notation

$\nabla f(\mathbf{x}) = \begin{bmatrix} \dfrac{\partial f}{\partial x_1} \\ \dfrac{\partial f}{\partial x_2} \\ \vdots \\ \dfrac{\partial f}{\partial x_d} \end{bmatrix}$

### Key Properties

| Property | Meaning |
|----------|----------|
| **Direction** | Points in direction of steepest **increase** of $f$ |
| **Magnitude** | $\|\nabla f\|$ = rate of increase in that direction |
| **Perpendicularity** | $\nabla f$ ⊥ level curves (contours of constant $f$) |
| **At extrema** | $\nabla f = \mathbf{0}$ at local minima/maxima |
| **Optimization** | Move in direction of $-\nabla f$ to minimize $f$ |
| **Sensitivity** | Large $\left\|\frac{\partial f}{\partial x_i}\right\|$ = high sensitivity to $x_i$ |

### Gradient Descent Update Rule

$\mathbf{x}_{n+1} = \mathbf{x}_n - \alpha \nabla f(\mathbf{x}_n)$

Where:
- $\alpha$ = learning rate (step size)
- $-\nabla f$ = direction of steepest **decrease**
- Repeat until convergence

### Geometric Picture

Think of $f(\mathbf{x})$ as a landscape (height map):
- **$\nabla f$ points uphill** (direction of steepest ascent)
- **$-\nabla f$ points downhill** (direction of steepest descent)
- **Level curves** are paths where height stays constant
- **Gradient is perpendicular to level curves**
- **Following $-\nabla f$ reaches the valley (minimum) fastest**
