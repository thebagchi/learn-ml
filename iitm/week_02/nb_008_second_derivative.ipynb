# Differentiation: Second Derivative

## Key Concepts

- Derivative of a function $f(x)$ is also called as **first-order derivative** $f'(x)$

- **Second derivative** or **second-order derivative** of a function $f(x)$ is the derivative of the derivative of function $f(x)$

- It is denoted as:

  $\dfrac{d^2 f(x)}{dx^2} = \dfrac{d}{dx}\left(\dfrac{df(x)}{dx}\right) = f''(x)$

## Example

Let $f(x) = 2x^2 - 6x + 3$

  $\dfrac{df(x)}{dx} = \dfrac{d(2x^2-6x+3)}{dx} = 4x - 6$

**Second derivative:**

  $\dfrac{d^2f(x)}{dx^2} = \dfrac{d(4x-6)}{dx} = 2$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions for analysis
def f(x):
    return 2*x**2 - 6*x + 3

def f_prime(x):
    return 4*x - 6

def f_double_prime(x):
    return 2 * np.ones_like(x)

# Create x values
x = np.linspace(-2, 6, 300)
y = f(x)
y_prime = f_prime(x)
y_double_prime = f_double_prime(x)

# Create figure with three subplots
fig, axes = plt.subplots(3, 1, figsize=(14, 11))

# Plot 1: Original function
ax1 = axes[0]
ax1.plot(x, y, 'b-', linewidth=2.5, label='$f(x) = 2x^2 - 6x + 3$')
# Mark critical point
critical_x = 1.5  # where f'(x) = 0
critical_y = f(critical_x)
ax1.plot(critical_x, critical_y, 'r*', markersize=20, label=f'Minimum at x={critical_x}')
ax1.axvline(x=critical_x, color='red', linestyle='--', alpha=0.5)
ax1.grid(True, alpha=0.3)
ax1.set_ylabel('f(x)', fontsize=12, fontweight='bold')
ax1.set_title('Original Function: $f(x) = 2x^2 - 6x + 3$\n(Concave Up - Positive Second Derivative)', fontsize=13, fontweight='bold')
ax1.legend(fontsize=11, loc='upper right')
ax1.fill_between(x, y, alpha=0.2, color='blue')

# Plot 2: First derivative (slope)
ax2 = axes[1]
ax2.plot(x, y_prime, 'g-', linewidth=2.5, label="$f'(x) = 4x - 6$")
ax2.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
ax2.plot(critical_x, 0, 'r*', markersize=20, label=f"f'(x)=0 at x={critical_x}")
ax2.axvline(x=critical_x, color='red', linestyle='--', alpha=0.5)
# Shade regions to show increasing/decreasing
ax2.fill_between(x[x < critical_x], y_prime[x < critical_x], 0, alpha=0.3, color='red', label='Decreasing')
ax2.fill_between(x[x > critical_x], y_prime[x > critical_x], 0, alpha=0.3, color='green', label='Increasing')
ax2.grid(True, alpha=0.3)
ax2.set_ylabel("f'(x)", fontsize=12, fontweight='bold')
ax2.set_title('First Derivative: Rate of Change of f(x)', fontsize=13, fontweight='bold')
ax2.legend(fontsize=11, loc='upper left')
ax2.set_ylim(-15, 15)

# Plot 3: Second derivative (concavity)
ax3 = axes[2]
ax3.plot(x, y_double_prime, 'r-', linewidth=2.5, label="$f''(x) = 2$")
ax3.axhline(y=0, color='k', linestyle='-', linewidth=0.8)
ax3.fill_between(x, y_double_prime, 0, alpha=0.3, color='green', label='Concave up (positive)')
ax3.grid(True, alpha=0.3)
ax3.set_xlabel('x', fontsize=12, fontweight='bold')
ax3.set_ylabel("f''(x)", fontsize=12, fontweight='bold')
ax3.set_title('Second Derivative: Rate of Change of Slope (Concavity)\nPositive = Concave Up (∪)', fontsize=13, fontweight='bold')
ax3.legend(fontsize=11, loc='upper right')
ax3.set_ylim(-1, 5)

plt.tight_layout()
plt.show()

# Print analysis
print("=" * 90)
print("Second Derivative Analysis: f(x) = 2x² - 6x + 3")
print("=" * 90)
print("\nDerivatives:")
print(f"  f'(x)  = 4x - 6")
print(f"  f''(x) = 2 (constant)")
print("\n" + "=" * 90)
print("Analysis at different points:")
print("=" * 90)
print(f"{'x':<10} {'f(x)':<15} {'f''(x)':<15} {'f''(x)':<15} {'Concavity':<20} {'Meaning':<25}")
print(f"{'':10} {'':15} {'(slope)':<15} {'(curvature)':<15}")
print("-" * 90)

analysis_x = [-1, 0, 1, 1.5, 2, 3, 4]

for x_val in analysis_x:
    f_val = f(x_val)
    f_p_val = f_prime(x_val)
    f_pp_val = f_double_prime(x_val)
    
    if f_pp_val > 0:
        concavity = "Concave Up (∪)"
        meaning = "Curving upward"
    elif f_pp_val < 0:
        concavity = "Concave Down (∩)"
        meaning = "Curving downward"
    else:
        concavity = "Inflection"
        meaning = "Concavity changes"
    
    print(f"{x_val:<10} {f_val:<15.2f} {f_p_val:<15.2f} {f_pp_val:<15.2f} {concavity:<20} {meaning:<25}")

print("\n" + "=" * 90)
print("Second Derivative Test for Extrema:")
print("=" * 90)
print(f"\nAt x = 1.5 (where f'(1.5) = 0):")
print(f"  f(1.5) = {f(1.5):.2f}")
print(f"  f'(1.5) = {f_prime(1.5):.2f} (critical point)")
print(f"  f''(1.5) = {f_double_prime(1.5):.2f} > 0")
print(f"  → Since f''(1.5) > 0, x = 1.5 is a LOCAL MINIMUM")
print("\n" + "=" * 90)
print("Key Insights:")
print("=" * 90)
print("1. Second derivative measures how fast the slope is changing")
print("2. Positive f''(x) → Concave up (minimum point if f'(x)=0)")
print("3. Negative f''(x) → Concave down (maximum point if f'(x)=0)")
print("4. f''(x) = 0 → Possible inflection point")
print("5. Second Derivative Test: Use f''(x) at critical points to classify maxima/minima")
print("=" * 90)
```

## Geometric Representation and Applications

The second derivative represents the **rate of change of the slope** or the **curvature** of a function:

- **Positive second derivative**: Function is **concave up** (curves upward like ∪). The slope is increasing.
- **Negative second derivative**: Function is **concave down** (curves downward like ∩). The slope is decreasing.
- **Zero second derivative**: Possible **inflection point** where concavity changes.

**Applications:**
- Determining local maxima and minima (using Second Derivative Test)
- Identifying inflection points where concavity changes
- Measuring acceleration (second derivative of position with respect to time)
- Analyzing convexity in optimization problems
