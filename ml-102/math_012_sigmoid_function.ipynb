# Sigmoid Function

The sigmoid function is a fundamental activation function used in neural networks to introduce non-linearity and squash outputs to a range between 0 and 1.

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

**Key Properties:**
- **Domain**: All real numbers (-∞ to +∞)
- **Range**: (0, 1)
- **Shape**: S-shaped curve (hence "sigmoid")
- **Use**: Output layer for binary classification, hidden layer activations
- **Derivative**: $\sigma'(x) = \sigma(x)(1 - \sigma(x))$

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

print("Libraries imported successfully!")
```

```python
# Define sigmoid function (numerically stable version)
def sigmoid(x):
    # Clip input to prevent overflow in exp function
    x_clipped = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x_clipped))
```

```python
# Create data for wide range (-1 million to +1 million)
x_wide = np.linspace(-1e6, 1e6, 10000)
y_wide = sigmoid(x_wide)

# Create focused view for better visualization
x_focused = np.linspace(-10, 10, 1000)
y_focused = sigmoid(x_focused)

# Plot: Wide range + Focused view
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Wide range
axes[0].plot(x_wide, y_wide, color='blue', linewidth=2)
axes[0].axhline(y=0.5, color='red', linestyle='--', alpha=0.5, label='y = 0.5')
axes[0].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[0].axhline(y=1, color='gray', linestyle='-', alpha=0.3)
axes[0].axvline(x=0, color='gray', linestyle='-', alpha=0.3)
axes[0].set_xlabel('x (range: -1M to +1M)', fontsize=11)
axes[0].set_ylabel('σ(x)', fontsize=11)
axes[0].set_title('Sigmoid Function - Wide Range', fontsize=12, fontweight='bold')
axes[0].grid(True, alpha=0.3)
axes[0].legend()

# Focused range
axes[1].plot(x_focused, y_focused, color='blue', linewidth=2.5)
axes[1].axhline(y=0.5, color='red', linestyle='--', alpha=0.5, label='y = 0.5 (inflection point)')
axes[1].axvline(x=0, color='red', linestyle='--', alpha=0.5, label='x = 0')
axes[1].axhline(y=0, color='gray', linestyle='-', alpha=0.3)
axes[1].axhline(y=1, color='gray', linestyle='-', alpha=0.3)
axes[1].grid(True, alpha=0.3)
axes[1].set_xlabel('x', fontsize=11)
axes[1].set_ylabel('σ(x)', fontsize=11)
axes[1].set_title('Sigmoid Function - Focused View (-10 to +10)', fontsize=12, fontweight='bold')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# SymPy: Sigmoid function and its derivative
x = sp.Symbol('x')
sigma = sp.Function('sigma')

# Define sigmoid symbolically
sigmoid_symbolic = 1 / (1 + sp.exp(-x))

# Compute derivative
sigmoid_derivative = sp.diff(sigmoid_symbolic, x)

# Display sigmoid function
print("Sigmoid function:")
display(sp.Eq(sp.Symbol('sigma(x)'), sigmoid_symbolic))

# Display derivative
print("Derivative of sigmoid:")
display(sp.Eq(sp.Symbol("sigma'(x)"), sigmoid_derivative))

# Show σ'(x) = σ(x)(1 - σ(x)) expanded form
print("Alternative form:")
sigmoid_alt = sigmoid_symbolic * (1 - sigmoid_symbolic)
display(sp.Eq(sp.Symbol("sigma'(x)"), sigmoid_alt))

# Show σ'(x) = σ(x)(1 - σ(x)) in symbolic notation
print("Symbolic notation:")
display(sp.Eq(sp.Symbol("sigma'(x)"), sigma(x) * (1 - sigma(x))))
```
