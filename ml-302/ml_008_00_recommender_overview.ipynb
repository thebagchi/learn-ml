# Recommender Systems: Overview

## Learning Objectives

1. **Describe** the utility matrix and its sparsity challenge
2. **Contrast** content-based and collaborative filtering approaches
3. **Identify** the cold start problem and common solutions
4. **Explain** the overall pipeline from data to recommendation

## The Utility Matrix

The core data structure is the **utility matrix** $M \in \mathbb{R}^{m \times n}$:
- Rows = users ($m$ users)
- Columns = items ($n$ items)
- Entry $M_{ui}$ = preference of user $u$ for item $i$

**Forms of preference:**
| Type | Example | Issue |
|------|---------|-------|
| Explicit ratings | Star ratings (1–5) | Sparse; users rarely rate |
| Implicit feedback | Click, view, purchase | No negative signal |
| Binary | Like/dislike | Harder to model |

**Sparsity:** Netflix has 500K users × 17K movies; only 1% of entries are known.
Goal: fill in the unknown entries to generate recommendations.

## Three Paradigms

### 1. Content-Based Filtering
- Build item profiles from item features (genre, director, keywords)
- Build user profiles from items they've rated
- Recommend items similar to what the user has liked before

**Pro:** no cold start for new items (if features available)
**Con:** requires item features; limited to "more of the same"

### 2. Collaborative Filtering
- Use patterns of user-item interactions (no features needed)
- "Users who liked what you liked also liked X"

**Sub-types:**
- **User-user CF:** find similar users; average their ratings
- **Item-item CF:** find items with similar rating patterns

**Pro:** captures complex preferences; cross-domain discovery
**Con:** cold start for new users/items; sparsity

### 3. Latent Factor Models

Learn $k$-dimensional embeddings for users and items:
$$\hat{M}_{ui} = U_u \cdot V_i$$

Training: minimize error on observed ratings via SGD or ALS.

**Pro:** compact; handles sparsity well; scales to millions of users/items
**Con:** hard to interpret; harder to explain recommendations

### Hybrid Approaches
Combine content-based signals with collaborative filtering:
- Weighted average of scores from both systems
- Feature-augmented matrix factorization (LightFM)
- Netflix winner: Ensemble of 100+ models

## The Cold Start Problem

| Situation | Problem |
|-----------|---------|
| New user | No ratings to base recommendations on |
| New item | No ratings; collaborative filtering blind |
| New domain | No data at all |

**Solutions:**
- **New user:** ask for initial ratings; use demographic features; popular items first
- **New item:** use item content features until ratings accumulate
- **Both:** matrix factorization with side information ($U_u = A z_u$ where $z_u$ = user features)

## Evaluation

| Metric | When | Formula |
|--------|------|---------|
| RMSE | Explicit ratings | $\sqrt{\frac{1}{|\Omega|}\sum_{ui}(M_{ui}-\hat{M}_{ui})^2}$ |
| Precision@k | Top-k recommendations | Fraction of top-k that are relevant |
| Recall@k | Top-k recommendations | Fraction of relevant items in top-k |
| NDCG | Ranking quality | Normalized discounted cumulative gain |
