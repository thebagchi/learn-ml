## 1 — Counting

| Rule | Formula | When to use |
|------|---------|-------------|
| **Product rule** | $n_1 \times n_2 \times \cdots \times n_k$ | Sequential **independent** choices |
| **Sum rule** | $n_1 + n_2 + \cdots + n_k$ | **Mutually exclusive** alternatives |
| **Factorial** | $n! = 1 \cdot 2 \cdots n \quad (0! = 1)$ | Total orderings of $n$ objects |

---
## 2 — Permutations & Combinations

| | Order matters | Order doesn't matter |
|---|---|---|
| **Without repetition** | $P(n,r) = \dfrac{n!}{(n-r)!}$ | $C(n,r) = \dbinom{n}{r} = \dfrac{n!}{r!(n-r)!}$ |
| **With repetition** | $n^r$ | $\dbinom{n+r-1}{r}$ (stars & bars) |

**Key identities:**

$$P(n,r) = C(n,r) \cdot r! \qquad \binom{n}{r} = \binom{n}{n-r}$$

**Stars & bars** — distribute $n$ identical objects into $k$ distinct bins:

$$\binom{n+k-1}{k-1}$$

### 2.1 — Combinations with Repetition & Distributing into Buckets

These two problems look different but use the **same formula** (stars & bars).

---

#### A. Selecting $r$ items from $n$ types (repetition allowed)

> How many ways can you choose $r$ items when there are $n$ types to pick from, and you **may pick the same type more than once**?

$$\binom{n + r - 1}{r} = \frac{(n + r - 1)!}{r!\,(n - 1)!}$$

**Example:** Choose 4 scoops from 3 ice-cream flavours (vanilla, chocolate, strawberry).

$$\binom{3 + 4 - 1}{4} = \binom{6}{4} = 15$$

---

#### B. Distributing $n$ identical objects into $r$ distinct buckets

> How many ways can you place $n$ identical items into $r$ labelled bins (each bin may hold any number)?

$$\binom{n + r - 1}{r - 1} = \frac{(n + r - 1)!}{(r-1)!\,n!}$$

**Example:** Distribute 5 identical balls into 3 labelled boxes.

$$\binom{5 + 3 - 1}{3 - 1} = \binom{7}{2} = 21$$

---

#### Why the same formula?

Both reduce to placing **stars** (items) and **bars** (dividers):

| Problem | Stars | Bars | Formula |
|---------|-------|------|---------|
| Select $r$ from $n$ types | $r$ chosen items | $n-1$ dividers between types | $\binom{n+r-1}{r}$ |
| Distribute $n$ into $r$ buckets | $n$ objects | $r-1$ dividers between buckets | $\binom{n+r-1}{r-1}$ |

Note: $\binom{n+r-1}{r} = \binom{n+r-1}{n-1}$ — these are the **same number** viewed from different angles.

```python
from math import comb

# --- A. Selecting r items from n types (repetition allowed) ---
n_types, r_picks = 3, 4  # 3 flavours, choose 4 scoops
ways_select = comb(n_types + r_picks - 1, r_picks)
print("A. Selecting r items from n types (repetition allowed)")
print(f"   {r_picks} scoops from {n_types} flavours: C({n_types}+{r_picks}-1, {r_picks})"
      f" = C({n_types + r_picks - 1}, {r_picks}) = {ways_select}")

# --- B. Distributing n identical objects into r distinct buckets ---
n_objects, r_buckets = 5, 3  # 5 balls into 3 boxes
ways_dist = comb(n_objects + r_buckets - 1, r_buckets - 1)
print(f"\nB. Distributing n identical objects into r distinct buckets")
print(f"   {n_objects} balls into {r_buckets} boxes: C({n_objects}+{r_buckets}-1, {r_buckets}-1)"
      f" = C({n_objects + r_buckets - 1}, {r_buckets - 1}) = {ways_dist}")

# --- Show they are the same formula ---
print(f"\nBoth use stars-and-bars:")
print(f"   C(n+r-1, r) = C(n+r-1, n-1)  —  symmetric form of the same binomial coefficient")
```

## 3 — Probability Fundamentals

| Term | Meaning |
|------|---------|
| **Sample space** $S$ | Set of all possible outcomes |
| **Event** $E$ | A subset $E \subseteq S$ |
| **Complement** $E^C$ | Everything in $S$ not in $E$ |

### Kolmogorov axioms

1. $0 \leq P(E) \leq 1$
2. $P(S) = 1$
3. If $E_1, E_2, \ldots$ are **mutually exclusive**: $\;P(E_1 \cup E_2 \cup \cdots) = P(E_1) + P(E_2) + \cdots$

### Equally likely outcomes

$$P(E) = \frac{|E|}{|S|}$$

### Complement rule

$$P(E^C) = 1 - P(E)$$

---
## 4 — OR  (Union)

**Mutually exclusive** ($E \cap F = \emptyset$):

$$P(E \cup F) = P(E) + P(F)$$

**General (inclusion-exclusion):**

$$P(E \cup F) = P(E) + P(F) - P(E \cap F)$$

**Three events:**

$$P(E \cup F \cup G) = P(E) + P(F) + P(G) - P(E \cap F) - P(E \cap G) - P(F \cap G) + P(E \cap F \cap G)$$

---
## 5 — Conditional Probability

$$P(E|F) = \frac{P(E \cap F)}{P(F)} \qquad P(F) > 0$$

### Chain rule

$$P(E \cap F) = P(E|F)\,P(F) = P(F|E)\,P(E)$$

---
## 6 — AND  (Intersection)

**Dependent events** (general):

$$P(E \cap F) = P(E|F)\,P(F)$$

**Independent events:**

$$P(E \cap F) = P(E)\,P(F)$$

**$n$ independent events:**

$$P(E_1 \cap E_2 \cap \cdots \cap E_n) = \prod_{i=1}^{n} P(E_i)$$

---
## 7 — Independence

$$E \perp\!\!\!\perp F \;\iff\; P(E|F) = P(E) \;\iff\; P(E \cap F) = P(E)\,P(F)$$

**Properties (if $E \perp\!\!\!\perp F$):**

- Independence is **symmetric**: $E \perp\!\!\!\perp F \iff F \perp\!\!\!\perp E$
- Extends to complements: $E \perp\!\!\!\perp F^C$, $\;E^C \perp\!\!\!\perp F$, $\;E^C \perp\!\!\!\perp F^C$

**Conditional independence:**

$$P(E|F,G) = P(E|G) \qquad \text{($E$ independent of $F$ given $G$)}$$

---
## 8 — Law of Total Probability (LOTP)

**Two-event partition:**

$$P(E) = P(E|F)\,P(F) + P(E|F^C)\,P(F^C)$$

**General** ($B_1, B_2, \ldots, B_n$ mutually exclusive & exhaustive):

$$P(E) = \sum_{i=1}^{n} P(E|B_i)\,P(B_i)$$

---
## 9 — Bayes' Theorem

$$\boxed{\;P(B|E) = \frac{P(E|B)\,P(B)}{P(E)}\;}$$

**Expanded with LOTP:**

$$P(B|E) = \frac{P(E|B)\,P(B)}{P(E|B)\,P(B) + P(E|B^C)\,P(B^C)}$$

| Term | Name | Meaning |
|------|------|---------|
| $P(B)$ | **Prior** | Initial belief before evidence |
| $P(E\|B)$ | **Likelihood** | How probable the evidence is if $B$ is true |
| $P(B\|E)$ | **Posterior** | Updated belief after observing $E$ |
| $P(E)$ | **Normaliser** | Total probability of the evidence |

---
## 10 — Log Probabilities

$$\log P(E) \in (-\infty,\; 0]$$

**Key property — products become sums:**

$$\log\!\left(\prod_i P(E_i)\right) = \sum_i \log P(E_i)$$

Use log probabilities to avoid **numerical underflow** when multiplying many small probabilities.

---
## 11 — Many Flips / Binomial

Coin with $P(\text{heads}) = p$, flipped $n$ times.

| Event | Formula |
|-------|---------|
| All heads | $p^n$ |
| All tails | $(1-p)^n$ |
| Exactly $k$ heads | $\dbinom{n}{k}\,p^{\,k}\,(1-p)^{n-k}$ |
| At least 1 head | $1 - (1-p)^n$ |
