### Problem 11: Eye Color Genetics

The color of a person's eyes is determined by a pair of eye-color genes, as follows:

If both of the eye-color genes are blue-eyed genes, then the person will have blue eyes.
If one or more of the genes is a brown-eyed gene, then the person will have brown eyes.

A newborn child independently receives one eye-color gene from each of its parents, and the gene it receives from a parent is equally likely to be either of the two eye-color genes of that parent. Suppose William and both of his parents have brown eyes, but William's sister (Claire) has blue eyes. (We assume that blue and brown are the only eye-color genes.)

*Note: Since William's sister has blue eyes, the parents must carry both blue and brown genes.*

a. What is the probability that William possesses a blue-eyed gene?  
b. Suppose that William's wife has blue eyes. What is the probability that their first child will have blue eyes?  
c. Still assuming that William's wife has blue eyes, if their first child had brown eyes, what is the probability that their next child will also have brown eyes?

````python
# since Williams sister has blue eyes thus parents carry both Blue and Brown gene

# Using combinations: each parent has 2 genes, chooses 1 to pass.
# Number of ways per parent: math.comb(2, 1) = 2
# Total ways for both parents: 2 * 2 = 4

ways_per_parent = math.comb(2, 1)
total_ways = ways_per_parent * ways_per_parent
print(f"Total ways genes can be passed: {total_ways}")

# For William possessing a Blue-eyed gene (at least one Blue allele):
# Number of ways with exactly 1 Blue gene: math.comb(2,1)
# Number of ways with exactly 2 Blue genes: math.comb(2,2)
# Ways with only Brown genes (0 Blue genes): math.comb(2,0)

total_ways_blue_gene = math.comb(2, 1) + math.comb(2, 2)  # 2 + 1 = 3
total_ways_only_blue_gene = math.comb(2, 2)  # 1 (both Blue)
total_ways_brown_gene = math.comb(2, 1) + math.comb(2, 2)  # 2 + 1 = 3 (at least one Brown: BlueBrown, BrownBlue, BrownBrown)
total_ways_only_brown_gene = math.comb(2, 0)  # 1 (no Blue genes, i.e., both Brown)
total_ways_brown_and_blue_gene = math.comb(2, 1)  # 2 (exactly one Blue and one Brown: BlueBrown, BrownBlue)

print(f"Ways with Blue gene: {total_ways_blue_gene}")
print(f"Ways with only Blue genes: {total_ways_only_blue_gene}")
print(f"Ways with Brown gene: {total_ways_brown_gene}")
print(f"Ways with only Brown genes: {total_ways_only_brown_gene}")
print(f"Ways with both Brown and Blue genes: {total_ways_brown_and_blue_gene}")

# Probability that William possesses a Blue-eyed gene
prob_blue_gene = total_ways_brown_and_blue_gene / total_ways_brown_gene
print(f"Probability of possessing Blue gene: {prob_blue_gene}")

# Part b: Probability that their first Child will have Blue eyes
# William's wife has Blue eyes, so she has genotype BlueBlue (two Blue genes)
# William has Brown eyes, and from part (a):
#   - With probability 2/3, William has genotype BlueBrown (one Blue, one Brown)
#   - With probability 1/3, William has genotype BrownBrown (two Brown)

# If William is BlueBrown (prob 2/3):
#   - Child gets Blue or Brown from William with equal probability (1/2 each)
#   - Child gets Blue from wife (always)
#   - Child genotypes: BlueBlue (prob 1/2) or BrownBlue (prob 1/2)
#   - P(Child has Blue eyes | William is BlueBrown) = 1/2

# If William is BrownBrown (prob 1/3):
#   - Child gets Brown from William (always)
#   - Child gets Blue from wife (always)
#   - Child genotype: BrownBlue (Brown eyes)
#   - P(Child has Blue eyes | William is BrownBrown) = 0

prob_william_blue_gene = prob_blue_gene  # 2/3
prob_william_blue_gene_c = 1 - prob_william_blue_gene  # 1/3

# P(Child has Blue eyes | William is BlueBrown)
prob_child_blue_given_blue_brown = 1/2

# P(Child has Blue eyes | William is BrownBrown)
prob_child_blue_given_brown_brown = 0

# P(Child has Blue eyes) = P(William BlueBrown) * P(Child Blue | BlueBrown) + P(William BrownBrown) * P(Child Blue | BrownBrown)
prob_child_blue_eyes = (prob_william_blue_gene * prob_child_blue_given_blue_brown + 
                        prob_william_blue_gene_c * prob_child_blue_given_brown_brown)

print(f"\nPart b: Probability that first Child will have Blue eyes")
print(f"P(Child has Blue eyes): {prob_child_blue_eyes}")

# Part c: If their first child had Brown eyes, what is the probability that their next child will also have Brown eyes?
# Given that the first child has Brown eyes:
# - Child got Blue from wife (always)
# - Child has Brown eyes, so must have at least one Brown gene
# - Child must have gotten Brown gene from William
# This gives us information about William's genotype

# Possible first child genotype: BrownBlue (has Brown eyes)
# P(child 1 BrownBlue | William BlueBrown, wife BlueBlue) = 1/2 (William gives Brown with prob 1/2)
# P(child 1 BrownBlue | William BrownBrown, wife BlueBlue) = 1 (William always gives Brown)

prob_child_1_brown_given_blue_brown = 1/2
prob_child_1_brown_given_brown_brown = 1

# P(child 1 has Brown eyes) using law of total probability
prob_child_1_brown = (prob_william_blue_gene * prob_child_1_brown_given_blue_brown + 
                      prob_william_blue_gene_c * prob_child_1_brown_given_brown_brown)

# P(William BlueBrown | child 1 has Brown eyes) using Bayes' theorem
prob_william_blue_brown_given_child_1_brown = (prob_child_1_brown_given_blue_brown * prob_william_blue_gene) / prob_child_1_brown

# P(William BrownBrown | child 1 has Brown eyes) using Bayes' theorem
prob_william_brown_brown_given_child_1_brown = (prob_child_1_brown_given_brown_brown * prob_william_blue_gene_c) / prob_child_1_brown

# For Child 2, wife still has BlueBlue
# If William is BlueBrown (updated prob): P(Child 2 Brown eyes) = 1/2
# If William is BrownBrown (updated prob): P(Child 2 Brown eyes) = 1

prob_child_2_brown_given_blue_brown = 1/2
prob_child_2_brown_given_brown_brown = 1

# P(Child 2 has Brown eyes | Child 1 has Brown eyes)
prob_child_2_brown_eyes = (prob_william_blue_brown_given_child_1_brown * prob_child_2_brown_given_blue_brown + 
                           prob_william_brown_brown_given_child_1_brown * prob_child_2_brown_given_brown_brown)

print(f"\nPart c: If Child 1 had Brown eyes, probability Child 2 also has Brown eyes")
print(f"P(Child 1 Brown eyes): {prob_child_1_brown}")
print(f"P(William BlueBrown | Child 1 Brown): {prob_william_blue_brown_given_child_1_brown}")
print(f"P(William BrownBrown | Child 1 Brown): {prob_william_brown_brown_given_child_1_brown}")
print(f"P(Child 2 Brown eyes | Child 1 Brown): {prob_child_2_brown_eyes}")
````
