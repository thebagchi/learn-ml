# Variance

**Variance** measures the **spread** of a random variable around its mean.

$$\text{Var}(X) = E[(X - \mu)^2] = E[X^2] - (E[X])^2$$

where $\mu = E[X]$.

**Standard deviation:** $\text{Std}(X) = \sigma = \sqrt{\text{Var}(X)}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Libraries loaded!")
```

---
## 1. Why Variance Matters — Three Graders

Imagine three graders each assigning scores for a paper worth 70/100:
- **Grader A**: Wide spread (high variance)
- **Grader B**: Tight spread (low variance)
- **Grader C**: Biased mean (wrong expectation)

```python
# Three graders with different spreads
np.random.seed(42)

fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

# Grader A: correct mean, high variance
grader_a = stats.norm(loc=70, scale=15)
# Grader B: correct mean, low variance
grader_b = stats.norm(loc=70, scale=5)
# Grader C: wrong mean, medium variance
grader_c = stats.norm(loc=55, scale=8)

x = np.linspace(20, 100, 300)
graders = [
    ("Grader A: High Spread", grader_a, '#e74c3c'),
    ("Grader B: Low Spread", grader_b, '#2ecc71'),
    ("Grader C: Wrong Mean", grader_c, '#f39c12'),
]

for ax, (name, rv, color) in zip(axes, graders):
    ax.fill_between(x, rv.pdf(x), alpha=0.4, color=color)
    ax.plot(x, rv.pdf(x), color=color, linewidth=2)
    ax.axvline(rv.mean(), color='black', linewidth=2, linestyle='--', label=f'μ = {rv.mean():.0f}')
    ax.axvline(70, color='gray', linewidth=1, linestyle=':', alpha=0.5)
    ax.set_title(f'{name}\nVar = {rv.var():.0f}, σ = {rv.std():.1f}',
                 fontsize=11, fontweight='bold')
    ax.set_xlabel('Score', fontsize=10)
    ax.set_ylabel('Density', fontsize=10)
    ax.legend(fontsize=9)
    ax.set_xlim(20, 100)

plt.suptitle('Same Paper (true score = 70) — Different Graders', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("Grader B is best: correct mean AND low variance")
```

---
## 2. Computing Variance

$$\text{Var}(X) = E[X^2] - (E[X])^2$$

This form is easier to compute than the definition $E[(X - \mu)^2]$.

```python
# Variance of a single fair die
# X ∈ {1, 2, 3, 4, 5, 6}, each with probability 1/6
E_X = sum(x * (1/6) for x in range(1, 7))
E_X2 = sum(x**2 * (1/6) for x in range(1, 7))
Var_X = E_X2 - E_X**2

print("Fair die:")
print(f"  E[X]    = {E_X:.4f}")
print(f"  E[X²]   = {E_X2:.4f}")
print(f"  Var(X)  = E[X²] - E[X]² = {E_X2:.4f} - {E_X**2:.4f} = {Var_X:.4f}")
print(f"  Std(X)  = √Var(X) = {np.sqrt(Var_X):.4f}")
print(f"  Theory: Var = 35/12 = {35/12:.4f} ✓")
```

---
## 3. Properties of Variance

| Property | Formula |
|----------|---------|
| **Constant** | $\text{Var}(a) = 0$ |
| **Scaling** | $\text{Var}(aX) = a^2 \text{Var}(X)$ |
| **Shift** | $\text{Var}(X + b) = \text{Var}(X)$ |
| **Linear** | $\text{Var}(aX + b) = a^2 \text{Var}(X)$ |
| **Independent sum** | $\text{Var}(X + Y) = \text{Var}(X) + \text{Var}(Y)$ if $X \perp Y$ |

```python
# Demonstrate properties with simulation
np.random.seed(42)
n = 100000
X = np.random.randint(1, 7, size=n)  # fair die

a, b = 3, 5
print(f"Var(X)     = {X.var():.4f}  (theory: {35/12:.4f})")
print(f"Var(aX)    = {(a*X).var():.4f}  (theory: a²·Var(X) = {a**2 * 35/12:.4f})")
print(f"Var(X + b) = {(X + b).var():.4f}  (theory: Var(X) = {35/12:.4f})")
print(f"Var(aX+b)  = {(a*X + b).var():.4f}  (theory: a²·Var(X) = {a**2 * 35/12:.4f})")

# Independent sum
Y = np.random.randint(1, 7, size=n)
print(f"\nVar(X + Y) = {(X + Y).var():.4f}  (theory: Var(X) + Var(Y) = {2*35/12:.4f})")
```

## 4. Comparing Distributions by Variance

```python
fig, ax = plt.subplots(figsize=(12, 5))

params = [(20, 0.1), (20, 0.3), (20, 0.5), (20, 0.9)]
colors = ['#3498db', '#2ecc71', '#e74c3c', '#9b59b6']
x = np.arange(0, 21)

for (n, p), color in zip(params, colors):
    rv = stats.binom(n, p)
    ax.plot(x, rv.pmf(x), 'o-', color=color, linewidth=2, markersize=5,
            label=f'Bin({n}, {p}): μ={rv.mean():.1f}, σ²={rv.var():.1f}')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('P(X = x)', fontsize=12)
ax.set_title('Binomial Distributions — Same n, Different Variance', fontsize=13, fontweight='bold')
ax.legend(fontsize=10)
plt.tight_layout()
plt.show()

print("Note: Variance is maximized at p = 0.5 for a fixed n")
```

---
## 5. Python: Computing Variance

```python
# scipy.stats
rv = stats.binom(n=20, p=0.5)
print(f"scipy.stats:  Var = {rv.var():.2f}, Std = {rv.std():.4f}")

# From samples
samples = rv.rvs(size=100000, random_state=42)
print(f"From samples: Var = {samples.var():.2f}, Std = {samples.std():.4f}")

# numpy directly
print(f"np.var:       Var = {np.var(samples):.2f}")
print(f"np.std:       Std = {np.std(samples):.4f}")
```
