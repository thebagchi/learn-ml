# Expectation

The **expectation** (or **expected value**) of a random variable $X$ is the weighted average of all possible values, weighted by their probabilities.

$$E[X] = \sum_{x} x \cdot P(X = x)$$

Also called: **mean**, **weighted average**, **center of mass**, **first moment**.

Notation: $E[X]$, $\mu$, $\mu_X$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Libraries loaded!")
```

---
## 1. Computing Expectation — Sum of Two Dice

```python
# E[X] for sum of two dice
def pmf_sum_two_dice(y):
    if y < 2 or y > 12:
        return 0
    return (y - 1) / 36 if y <= 7 else (13 - y) / 36

E_X = sum(y * pmf_sum_two_dice(y) for y in range(2, 13))

print("Computing E[X] for sum of two dice:")
print(f"  E[X] = Σ x·P(X=x)")
terms = []
for y in range(2, 13):
    p = pmf_sum_two_dice(y)
    terms.append(f"{y}·({p:.4f})")
    print(f"       + {y:2d} × {p:.4f} = {y * p:.4f}")
print(f"  E[X] = {E_X}")

# Visualize
fig, ax = plt.subplots(figsize=(10, 5))
x_vals = np.arange(2, 13)
pmf_vals = np.array([pmf_sum_two_dice(y) for y in x_vals])
ax.bar(x_vals, pmf_vals, color='#3498db', edgecolor='black', linewidth=1.2, width=0.6, alpha=0.7)
ax.axvline(E_X, color='#e74c3c', linewidth=3, linestyle='--', label=f'E[X] = {E_X}')
ax.set_xlabel('x (sum)', fontsize=12)
ax.set_ylabel('P(X = x)', fontsize=12)
ax.set_title('Expectation = Center of Mass of the PMF', fontsize=13, fontweight='bold')
ax.legend(fontsize=12)
ax.set_xticks(x_vals)
plt.tight_layout()
plt.show()
```

---
## 2. Properties of Expectation

| Property | Formula | Notes |
|----------|---------|-------|
| **Linearity** | $E[aX + b] = aE[X] + b$ | $a, b$ are constants |
| **Sum of RVs** | $E[X + Y] = E[X] + E[Y]$ | Always true, even if dependent! |
| **Constant** | $E[a] = a$ | A constant doesn't change |
| **LOTUS** | $E[g(X)] = \sum_x g(x) P(X = x)$ | No need to find the distribution of $g(X)$ |

```python
# Demonstrating properties
np.random.seed(42)
n = 100000

# Single die: E[X] = 3.5
die = np.random.randint(1, 7, size=n)
print(f"E[Die] = {die.mean():.4f}  (theory: 3.5)")

# Linearity: E[2X + 3]
print(f"E[2X + 3] = {(2 * die + 3).mean():.4f}  (theory: {2*3.5 + 3})")

# Sum: E[X + Y] = E[X] + E[Y]
die2 = np.random.randint(1, 7, size=n)
print(f"E[X + Y] = {(die + die2).mean():.4f}  (theory: {3.5 + 3.5})")

# LOTUS: E[X²]
print(f"E[X²] = {(die**2).mean():.4f}  (theory: {sum(x**2 / 6 for x in range(1, 7)):.4f})")
```

---
## 3. LOTUS — Law of the Unconscious Statistician

$$E[g(X)] = \sum_x g(x) \cdot P(X = x)$$

You can compute the expectation of **any function** of $X$ without finding the distribution of $g(X)$.

```python
# LOTUS example: E[X²] for sum of two dice
E_X2 = sum(y**2 * pmf_sum_two_dice(y) for y in range(2, 13))
E_X = sum(y * pmf_sum_two_dice(y) for y in range(2, 13))

print("LOTUS: E[X²] for sum of two dice")
print(f"  E[X²] = Σ x² · P(X=x) = {E_X2:.4f}")
print(f"  E[X]  = {E_X:.4f}")
print(f"  E[X]² = {E_X**2:.4f}")
print(f"  Note: E[X²] ≠ E[X]²  ({E_X2:.4f} ≠ {E_X**2:.4f})")
print(f"  This difference gives us Var(X) = E[X²] - E[X]² = {E_X2 - E_X**2:.4f}")
```

## 4. Visualizing Expectation — Comparing Distributions

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

distributions = [
    ("Binom(20, 0.3)", stats.binom(20, 0.3), range(0, 21)),
    ("Binom(20, 0.5)", stats.binom(20, 0.5), range(0, 21)),
    ("Binom(20, 0.8)", stats.binom(20, 0.8), range(0, 21)),
]

colors = ['#e74c3c', '#2ecc71', '#3498db']
for ax, (name, rv, x_range), color in zip(axes, distributions, colors):
    x = np.array(list(x_range))
    pmf = rv.pmf(x)
    ax.bar(x, pmf, color=color, edgecolor='black', linewidth=0.8, width=0.7, alpha=0.7)
    mean = rv.mean()
    ax.axvline(mean, color='black', linewidth=2.5, linestyle='--', label=f'E[X] = {mean:.1f}')
    ax.set_title(name, fontsize=13, fontweight='bold')
    ax.set_xlabel('x', fontsize=11)
    ax.set_ylabel('P(X=x)', fontsize=11)
    ax.legend(fontsize=11)

plt.suptitle('Expectation shifts with the parameter p', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

---
## 5. Python: Computing Expectation

```python
# Method 1: From scratch with a PMF
def expectation_from_pmf(pmf_func, support):
    return sum(x * pmf_func(x) for x in support)

E = expectation_from_pmf(pmf_sum_two_dice, range(2, 13))
print(f"Method 1 (manual):     E[sum of two dice] = {E}")

# Method 2: Using scipy.stats
rv = stats.binom(n=10, p=0.6)
print(f"Method 2 (scipy):      E[Binom(10, 0.6)] = {rv.mean()}")
print(f"                       Var = {rv.var()}")

# Method 3: Simulation
np.random.seed(0)
samples = rv.rvs(size=100000)
print(f"Method 3 (simulation): E[Binom(10, 0.6)] ≈ {samples.mean():.4f}")
```
