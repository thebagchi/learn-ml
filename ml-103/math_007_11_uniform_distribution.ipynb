# Uniform Distribution

The **Uniform** distribution assigns equal probability to all values in an interval $[a, b]$.

| Property | Value |
|----------|-------|
| **Notation** | $X \sim \text{Uni}(a, b)$ |
| **Description** | Equal probability across $[a, b]$ |
| **Parameters** | $a$ = lower bound, $b$ = upper bound |
| **Support** | $[a, b]$ |
| **PDF** | $f(x) = \frac{1}{b - a}$ for $a \leq x \leq b$ |
| **CDF** | $F(x) = \frac{x - a}{b - a}$ for $a \leq x \leq b$ |
| **Expectation** | $E[X] = \frac{a + b}{2}$ |
| **Variance** | $\text{Var}(X) = \frac{(b - a)^2}{12}$ |

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Libraries loaded!")
```

---
## 1. PDF and CDF Visualization

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 9))

configs = [(0, 1, '#e74c3c'), (2, 5, '#2ecc71'), (-3, 3, '#3498db')]

for col, (a, b, color) in enumerate(configs):
    rv = stats.uniform(loc=a, scale=b-a)
    x = np.linspace(a - 1, b + 1, 300)
    
    # PDF
    ax = axes[0, col]
    ax.plot(x, rv.pdf(x), color=color, linewidth=2.5)
    ax.fill_between(x, rv.pdf(x), alpha=0.3, color=color)
    ax.set_title(f'PDF: Uni({a}, {b})', fontsize=12, fontweight='bold')
    ax.set_ylabel('f(x)', fontsize=11)
    ax.set_xlabel('x', fontsize=11)
    ax.annotate(f'height = 1/({b}-{a}) = {1/(b-a):.3f}',
                xy=((a+b)/2, 1/(b-a)), fontsize=10, ha='center',
                xytext=((a+b)/2, 1/(b-a) + 0.15),
                arrowprops=dict(arrowstyle='->', color=color))
    
    # CDF
    ax = axes[1, col]
    ax.plot(x, rv.cdf(x), color=color, linewidth=2.5)
    ax.set_title(f'CDF: Uni({a}, {b})', fontsize=12, fontweight='bold')
    ax.set_ylabel('F(x)', fontsize=11)
    ax.set_xlabel('x', fontsize=11)
    ax.axhline(0, color='gray', alpha=0.3, linestyle=':')
    ax.axhline(1, color='gray', alpha=0.3, linestyle=':')

plt.tight_layout()
plt.show()
```

---
## 2. Probability Computations

For $X \sim \text{Uni}(a, b)$:

$$P(c \leq X \leq d) = \frac{d - c}{b - a}$$

The probability is proportional to the **length of the interval**.

```python
a, b = 0, 10
rv = stats.uniform(loc=a, scale=b-a)

print(f"X ~ Uni({a}, {b}):")
print(f"  P(2 ≤ X ≤ 5) = (5-2)/(10-0)  = {(5-2)/(b-a):.4f}  (scipy: {rv.cdf(5) - rv.cdf(2):.4f})")
print(f"  P(X ≤ 3)     = (3-0)/(10-0)  = {(3-0)/(b-a):.4f}  (scipy: {rv.cdf(3):.4f})")
print(f"  P(X > 7)     = (10-7)/(10-0) = {(10-7)/(b-a):.4f}  (scipy: {1 - rv.cdf(7):.4f})")
print(f"  P(X = 5)     = 0 (continuous!) (scipy: -)")
print(f"\n  E[X] = (0+10)/2       = {(a+b)/2}  (scipy: {rv.mean()})")
print(f"  Var(X) = (10-0)²/12   = {(b-a)**2/12:.4f}  (scipy: {rv.var():.4f})")

# Visual
fig, ax = plt.subplots(figsize=(10, 5))
x = np.linspace(-1, 11, 300)
ax.plot(x, rv.pdf(x), 'b-', linewidth=2.5)
ax.fill_between(x, rv.pdf(x), alpha=0.15, color='blue')

# Shade P(2 ≤ X ≤ 5)
mask = (x >= 2) & (x <= 5)
ax.fill_between(x[mask], rv.pdf(x[mask]), alpha=0.6, color='orange',
               label=f'P(2 ≤ X ≤ 5) = {rv.cdf(5) - rv.cdf(2):.2f}')
ax.axvline(rv.mean(), color='red', linewidth=2, linestyle='--', label=f'E[X] = {rv.mean():.1f}')
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('f(x)', fontsize=12)
ax.set_title(f'Uniform({a}, {b})', fontsize=13, fontweight='bold')
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

## 3. Generating Other Distributions from Uniform

The **inverse transform method**: if $U \sim \text{Uni}(0, 1)$ and $F$ is a CDF, then $X = F^{-1}(U)$ has CDF $F$.

```python
# Inverse transform: generate Exponential from Uniform
np.random.seed(42)
n = 50000

# Step 1: Generate Uniform(0,1)
U = np.random.uniform(0, 1, size=n)

# Step 2: Apply inverse CDF of Exp(lambda=2)
lam = 2
# CDF of Exp: F(x) = 1 - e^{-λx}, so F^{-1}(u) = -ln(1-u)/λ
X_exp = -np.log(1 - U) / lam

fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

# Uniform samples
axes[0].hist(U, bins=50, density=True, color='#3498db', edgecolor='white', alpha=0.7)
axes[0].set_title('Step 1: U ~ Uni(0,1)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('u')

# Arrow
axes[1].text(0.5, 0.5, 'F⁻¹(u) = −ln(1−u)/λ', fontsize=14,
             ha='center', va='center', transform=axes[1].transAxes,
             fontweight='bold', color='red')
axes[1].text(0.5, 0.3, '(inverse CDF)', fontsize=11,
             ha='center', va='center', transform=axes[1].transAxes,
             style='italic')
axes[1].axis('off')

# Resulting Exponential
x_plot = np.linspace(0, 4, 300)
axes[2].hist(X_exp, bins=80, density=True, color='#e74c3c', edgecolor='white', alpha=0.7,
             label='Generated samples')
axes[2].plot(x_plot, stats.expon.pdf(x_plot, scale=1/lam), 'k-', linewidth=2.5,
             label=f'True Exp(λ={lam}) PDF')
axes[2].set_title('Step 2: X ~ Exp(λ=2)', fontsize=12, fontweight='bold')
axes[2].set_xlabel('x')
axes[2].legend(fontsize=10)

plt.tight_layout()
plt.show()
print("Inverse transform: Uniform → any distribution with known inverse CDF")
```

## 4. Python: scipy.stats.uniform

```python
from scipy.stats import uniform

# Note: scipy parameterization is uniform(loc=a, scale=b-a)
a, b = 3, 8
rv = uniform(loc=a, scale=b - a)

print(f"Uniform({a}, {b}):")
print(f"  E[X]   = {rv.mean()}")
print(f"  Var(X) = {rv.var():.4f}")
print(f"  P(X ≤ 5) = {rv.cdf(5):.4f}")
print(f"  P(4 < X < 6) = {rv.cdf(6) - rv.cdf(4):.4f}")
print(f"  Samples: {rv.rvs(size=8, random_state=42).round(3)}")
```

## Summary

| Property | Formula |
|----------|---------|
| PDF | $f(x) = \frac{1}{b-a}$ for $x \in [a, b]$ |
| CDF | $F(x) = \frac{x-a}{b-a}$ |
| $E[X]$ | $\frac{a+b}{2}$ |
| $\text{Var}(X)$ | $\frac{(b-a)^2}{12}$ |
| Python | `scipy.stats.uniform(loc=a, scale=b-a)` |
