# Random Variables & Distributions — Formula Reference

Quick reference for all distributions covered in Part 2.

---

## 1. Core Concepts

| Concept | Discrete | Continuous |
|---------|----------|------------|
| Likelihood | PMF: $P(X=x)$ | PDF: $f(x)$ |
| Probability | $P(X=x) \geq 0$ | $P(X=x) = 0$ |
| Range prob. | $\sum_{x=a}^{b} P(X=x)$ | $\int_a^b f(x)\,dx$ |
| Normalization | $\sum_{\text{all } x} P(X=x) = 1$ | $\int_{-\infty}^{\infty} f(x)\,dx = 1$ |
| CDF | $F(x) = P(X \leq x) = \sum_{k \leq x} P(X=k)$ | $F(x) = \int_{-\infty}^{x} f(t)\,dt$ |
| Expectation | $E[X] = \sum_x x \cdot P(X=x)$ | $E[X] = \int x \cdot f(x)\,dx$ |
| Variance | $\text{Var}(X) = E[X^2] - (E[X])^2$ | $\text{Var}(X) = E[X^2] - (E[X])^2$ |

### Properties of Expectation
$$E[aX + b] = aE[X] + b \quad \quad E[X + Y] = E[X] + E[Y]$$
$$E[g(X)] = \sum_x g(x)P(X=x) \quad \text{(LOTUS)}$$

### Properties of Variance
$$\text{Var}(aX + b) = a^2\text{Var}(X) \quad \quad \text{Var}(X+Y) = \text{Var}(X) + \text{Var}(Y) \text{ if independent}$$

## 2. Discrete Distributions

| Distribution | PMF $P(X=k)$ | $E[X]$ | $\text{Var}(X)$ | Support |
|---|---|---|---|---|
| **Bernoulli** $\text{Bern}(p)$ | $p^k(1-p)^{1-k}$ | $p$ | $p(1-p)$ | $\{0,1\}$ |
| **Binomial** $\text{Bin}(n,p)$ | $\binom{n}{k}p^k(1-p)^{n-k}$ | $np$ | $np(1-p)$ | $\{0,\ldots,n\}$ |
| **Poisson** $\text{Poi}(\lambda)$ | $\frac{\lambda^k e^{-\lambda}}{k!}$ | $\lambda$ | $\lambda$ | $\{0,1,2,\ldots\}$ |
| **Geometric** $\text{Geo}(p)$ | $(1-p)^{k-1}p$ | $1/p$ | $(1-p)/p^2$ | $\{1,2,3,\ldots\}$ |
| **Neg. Binomial** $\text{NegBin}(r,p)$ | $\binom{k-1}{r-1}p^r(1-p)^{k-r}$ | $r/p$ | $r(1-p)/p^2$ | $\{r,r+1,\ldots\}$ |
| **Categorical** $\text{Cat}(p_1,\ldots,p_k)$ | $P(X=c_i) = p_i$ | — | — | $\{c_1,\ldots,c_k\}$ |

---

## 3. Continuous Distributions

| Distribution | PDF $f(x)$ | $E[X]$ | $\text{Var}(X)$ | Support |
|---|---|---|---|---|
| **Uniform** $\text{Uni}(a,b)$ | $\frac{1}{b-a}$ | $\frac{a+b}{2}$ | $\frac{(b-a)^2}{12}$ | $[a,b]$ |
| **Exponential** $\text{Exp}(\lambda)$ | $\lambda e^{-\lambda x}$ | $1/\lambda$ | $1/\lambda^2$ | $[0,\infty)$ |
| **Normal** $N(\mu,\sigma^2)$ | $\frac{1}{\sigma\sqrt{2\pi}}e^{-(x-\mu)^2/2\sigma^2}$ | $\mu$ | $\sigma^2$ | $(-\infty,\infty)$ |

---

## 4. Key Relationships

| Relationship | Formula |
|---|---|
| Binomial = sum of Bernoullis | $\text{Bin}(n,p) = \sum_{i=1}^{n} \text{Bern}(p)$ |
| NegBin = sum of Geometrics | $\text{NegBin}(r,p) = \sum_{i=1}^{r} \text{Geo}(p)$ |
| Poisson limit | $\text{Bin}(n,p) \xrightarrow{n\to\infty, p\to 0} \text{Poi}(np)$ |
| Normal approx (CLT) | $\text{Bin}(n,p) \approx N(np, np(1-p))$ for large $n$ |
| Exponential ↔ Poisson | Inter-arrival times of $\text{Poi}(\lambda)$ are $\text{Exp}(\lambda)$ |
| Memoryless (discrete) | Geometric only |
| Memoryless (continuous) | Exponential only |

---

## 5. Approximation Rules

| Approximation | Use When | Formula |
|---|---|---|
| Normal → Binomial | $np \geq 5$ and $n(1-p) \geq 5$ | $\text{Bin}(n,p) \approx N(np, np(1-p))$ |
| Poisson → Binomial | $n \geq 20$ and $p \leq 0.05$ | $\text{Bin}(n,p) \approx \text{Poi}(np)$ |
| Continuity correction | Normal approx of discrete | $P(X \leq k) \approx \Phi\left(\frac{k+0.5-\mu}{\sigma}\right)$ |

### 68-95-99.7 Rule (Normal)
- $P(\mu - \sigma < X < \mu + \sigma) \approx 68.27\%$
- $P(\mu - 2\sigma < X < \mu + 2\sigma) \approx 95.45\%$
- $P(\mu - 3\sigma < X < \mu + 3\sigma) \approx 99.73\%$

### Z-Score
$$Z = \frac{X - \mu}{\sigma} \sim N(0, 1) \qquad P(X \leq x) = \Phi\left(\frac{x - \mu}{\sigma}\right)$$

## 6. Python Quick Reference

```python
from scipy import stats
import numpy as np

# ─── Discrete ────────────────────────────────────────────
# Bernoulli
bern  = stats.bernoulli(p=0.7)

# Binomial
binom = stats.binom(n=10, p=0.3)

# Poisson
pois  = stats.poisson(mu=5)

# Geometric
geom  = stats.geom(p=0.2)

# Negative Binomial (scipy: # failures before r successes)
nbinom = stats.nbinom(n=3, p=0.4)

# ─── Continuous ──────────────────────────────────────────
# Uniform
uni   = stats.uniform(loc=2, scale=5)  # Uni(2, 7)

# Exponential
exp   = stats.expon(scale=1/3)         # Exp(λ=3), scale = 1/λ

# Normal
norm  = stats.norm(loc=0, scale=1)     # N(0, 1)

# ─── Common Methods ─────────────────────────────────────
rv = stats.binom(n=20, p=0.4)
print("Binomial(20, 0.4):")
print(f"  rv.mean()    = {rv.mean()}")
print(f"  rv.var()     = {rv.var()}")
print(f"  rv.std()     = {rv.std():.4f}")
print(f"  rv.pmf(8)    = {rv.pmf(8):.6f}     ← P(X=8)")
print(f"  rv.cdf(8)    = {rv.cdf(8):.6f}     ← P(X≤8)")
print(f"  rv.sf(8)     = {rv.sf(8):.6f}     ← P(X>8) = 1-CDF")
print(f"  rv.ppf(0.5)  = {rv.ppf(0.5):.0f}              ← median (inverse CDF)")
print(f"  rv.rvs(5)    = {rv.rvs(size=5, random_state=42)}   ← random samples")
```

```python
# Quick comparison chart
print("Distribution Summary Table")
print("=" * 80)
print(f"{'Distribution':<25s} {'E[X]':>8s} {'Var(X)':>10s} {'scipy name':>15s}")
print("-" * 80)

dists = [
    ("Bernoulli(0.7)",     stats.bernoulli(0.7),     "bernoulli"),
    ("Binomial(20, 0.3)",  stats.binom(20, 0.3),     "binom"),
    ("Poisson(5)",         stats.poisson(5),          "poisson"),
    ("Geometric(0.2)",     stats.geom(0.2),           "geom"),
    ("NegBin(3, 0.4)",     stats.nbinom(3, 0.4),     "nbinom"),
    ("Uniform(0, 10)",     stats.uniform(0, 10),      "uniform"),
    ("Exponential(λ=2)",   stats.expon(scale=0.5),    "expon"),
    ("Normal(5, 4)",       stats.norm(5, 2),          "norm"),
]

for name, rv, scipy_name in dists:
    print(f"{name:<25s} {rv.mean():>8.3f} {rv.var():>10.4f} {scipy_name:>15s}")
```

## 7. Visualization Quick Reference

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 4, figsize=(20, 9))

# Row 1: Discrete distributions
discrete = [
    ("Bernoulli(0.6)", stats.bernoulli(0.6), range(0, 2)),
    ("Binomial(15, 0.4)", stats.binom(15, 0.4), range(0, 16)),
    ("Poisson(4)", stats.poisson(4), range(0, 15)),
    ("Geometric(0.3)", stats.geom(0.3), range(1, 15)),
]

for ax, (name, rv, x_range) in zip(axes[0], discrete):
    x = np.array(list(x_range))
    ax.bar(x, rv.pmf(x), color='#3498db', edgecolor='black', linewidth=0.8, alpha=0.8)
    ax.axvline(rv.mean(), color='red', linewidth=2, linestyle='--', alpha=0.8)
    ax.set_title(name, fontsize=11, fontweight='bold')
    ax.set_xlabel('k')
    ax.set_ylabel('P(X=k)')

# Row 2: Continuous distributions
continuous = [
    ("Uniform(0, 5)", stats.uniform(0, 5), (-1, 6)),
    ("Exponential(λ=1)", stats.expon(scale=1), (0, 6)),
    ("Normal(0, 1)", stats.norm(0, 1), (-4, 4)),
    ("Normal(3, 4)", stats.norm(3, 2), (-4, 10)),
]

for ax, (name, rv, (lo, hi)) in zip(axes[1], continuous):
    x = np.linspace(lo, hi, 300)
    ax.plot(x, rv.pdf(x), 'b-', linewidth=2.5)
    ax.fill_between(x, rv.pdf(x), alpha=0.3, color='blue')
    ax.axvline(rv.mean(), color='red', linewidth=2, linestyle='--', alpha=0.8)
    ax.set_title(name, fontsize=11, fontweight='bold')
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')

plt.suptitle('All Distributions at a Glance (red dashed = mean)',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
