# Probability Mass Functions (PMF)

The **Probability Mass Function** (PMF) is the most important property of a discrete random variable.
It maps each possible value to its probability.

$$P(X = x) \quad \text{or shorthand} \quad p(x)$$

**Rules a valid PMF must satisfy:**

1. $P(X = x) \geq 0$ for all $x$
2. $\displaystyle\sum_{\text{all } x} P(X = x) = 1$

```python
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction

print("Libraries loaded!")
```

---
## 1. PMF as a Graph — Single Die vs Sum of Two Dice

```python
# PMF of a single fair die
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Single die
ax = axes[0]
x = np.arange(1, 7)
pmf = np.ones(6) / 6
ax.bar(x, pmf, color='#3498db', edgecolor='black', linewidth=1.5, width=0.6)
ax.set_xlabel('x (die face)', fontsize=12)
ax.set_ylabel('P(X = x)', fontsize=12)
ax.set_title('PMF: Single Fair Die', fontsize=13, fontweight='bold')
ax.set_ylim(0, 0.3)
ax.set_xticks(x)
for xi, pi in zip(x, pmf):
    ax.text(xi, pi + 0.005, f'1/6', ha='center', fontsize=9)
ax.axhline(y=0, color='black', linewidth=0.5)

# Sum of two dice
ax = axes[1]
x2 = np.arange(2, 13)
counts = np.array([1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1])
pmf2 = counts / 36
bars = ax.bar(x2, pmf2, color='#e74c3c', edgecolor='black', linewidth=1.5, width=0.6)
ax.set_xlabel('y (sum)', fontsize=12)
ax.set_ylabel('P(Y = y)', fontsize=12)
ax.set_title('PMF: Sum of Two Dice', fontsize=13, fontweight='bold')
ax.set_xticks(x2)
ax.set_ylim(0, 0.22)
for xi, pi, c in zip(x2, pmf2, counts):
    ax.text(xi, pi + 0.003, f'{c}/36', ha='center', fontsize=8)
ax.axhline(y=0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()

print("✓ Left: Uniform PMF (all outcomes equally likely)")
print("✓ Right: Triangular PMF (7 is most likely sum)")
```

---
## 2. PMF as a Function

We can express a PMF as a mathematical formula or Python function.

**Single die:** $P(X = x) = \frac{1}{6}$ for $x \in \{1, 2, 3, 4, 5, 6\}$

**Sum of two dice:**
$$P(Y = y) = \begin{cases} \frac{y-1}{36} & \text{if } 2 \leq y \leq 7 \\ \frac{13-y}{36} & \text{if } 8 \leq y \leq 12 \end{cases}$$

```python
def pmf_single_die(x):
    """PMF for a single fair die."""
    if 1 <= x <= 6:
        return Fraction(1, 6)
    return Fraction(0)

def pmf_sum_two_dice(y):
    """PMF for the sum of two fair dice."""
    if y < 2 or y > 12:
        return Fraction(0)
    if y <= 7:
        return Fraction(y - 1, 36)
    else:
        return Fraction(13 - y, 36)

# Verify both PMFs sum to 1
single_sum = sum(pmf_single_die(x) for x in range(1, 7))
double_sum = sum(pmf_sum_two_dice(y) for y in range(2, 13))

print("PMF values for single die:")
for x in range(1, 7):
    print(f"  P(X = {x}) = {pmf_single_die(x)} = {float(pmf_single_die(x)):.4f}")
print(f"  Sum = {single_sum} ✓\n")

print("PMF values for sum of two dice:")
for y in range(2, 13):
    p = pmf_sum_two_dice(y)
    print(f"  P(Y = {y:2d}) = {str(p):>4s} = {float(p):.4f}")
print(f"  Sum = {double_sum} ✓")
```

## 3. From Data to Histogram to PMF

Given a dataset of observations, you can **approximate** the PMF:
1. **Histogram**: Count how often each value appears
2. **Normalize**: Divide each count by total number of observations
3. The result approximates the true PMF

```python
# Simulate rolling two dice 10,000 times
np.random.seed(42)
n_samples = 10000
die1 = np.random.randint(1, 7, size=n_samples)
die2 = np.random.randint(1, 7, size=n_samples)
sums = die1 + die2

fig, axes = plt.subplots(1, 3, figsize=(16, 4.5))

# Step 1: Raw data histogram (counts)
ax = axes[0]
ax.hist(sums, bins=np.arange(1.5, 13.5, 1), color='#95a5a6', edgecolor='black', linewidth=1.2)
ax.set_xlabel('Sum', fontsize=11)
ax.set_ylabel('Count', fontsize=11)
ax.set_title('Step 1: Histogram (counts)', fontsize=12, fontweight='bold')
ax.set_xticks(range(2, 13))

# Step 2: Normalized histogram (approximate PMF)
ax = axes[1]
ax.hist(sums, bins=np.arange(1.5, 13.5, 1), density=True, color='#e74c3c', edgecolor='black',
        linewidth=1.2, alpha=0.7, label='Empirical')
ax.set_xlabel('Sum', fontsize=11)
ax.set_ylabel('Probability', fontsize=11)
ax.set_title('Step 2: Normalized (approx PMF)', fontsize=12, fontweight='bold')
ax.set_xticks(range(2, 13))

# Step 3: Overlay with true PMF
ax = axes[2]
y_vals = np.arange(2, 13)
true_pmf = np.array([float(pmf_sum_two_dice(y)) for y in y_vals])
empirical = np.array([np.mean(sums == y) for y in y_vals])

ax.bar(y_vals - 0.2, empirical, 0.4, color='#e74c3c', edgecolor='black', alpha=0.7, label=f'Empirical (n={n_samples:,})')
ax.bar(y_vals + 0.2, true_pmf, 0.4, color='#3498db', edgecolor='black', alpha=0.7, label='True PMF')
ax.set_xlabel('Sum', fontsize=11)
ax.set_ylabel('Probability', fontsize=11)
ax.set_title('Step 3: Empirical vs True PMF', fontsize=12, fontweight='bold')
ax.set_xticks(y_vals)
ax.legend(fontsize=10)

plt.tight_layout()
plt.show()

# Show numerical comparison
print(f"{'Sum':>4s} | {'Empirical':>10s} | {'True':>10s} | {'Difference':>10s}")
print("-" * 44)
for y in y_vals:
    emp = np.mean(sums == y)
    true = float(pmf_sum_two_dice(y))
    print(f"{y:4d} | {emp:10.4f} | {true:10.4f} | {abs(emp - true):10.4f}")
```

## 4. Python: Using scipy.stats PMF

`scipy.stats` provides `.pmf()` for all discrete distributions.

```python
from scipy import stats

# Binomial PMF
X = stats.binom(n=10, p=0.5)
x_vals = np.arange(0, 11)

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(x_vals, X.pmf(x_vals), color='#2ecc71', edgecolor='black', linewidth=1.2, width=0.6)
ax.set_xlabel('k (number of heads)', fontsize=12)
ax.set_ylabel('P(X = k)', fontsize=12)
ax.set_title('PMF of Binomial(n=10, p=0.5)', fontsize=13, fontweight='bold')
ax.set_xticks(x_vals)

# Annotate the mode
mode_idx = np.argmax(X.pmf(x_vals))
bars[mode_idx].set_color('#e74c3c')
ax.annotate('Mode', xy=(x_vals[mode_idx], X.pmf(x_vals[mode_idx])),
            xytext=(x_vals[mode_idx] + 1.5, X.pmf(x_vals[mode_idx]) + 0.02),
            arrowprops=dict(arrowstyle='->', color='red', lw=2),
            fontsize=12, color='red', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"P(X = 5) = {X.pmf(5):.4f}")
print(f"P(X ≤ 5) = {X.cdf(5):.4f}  (CDF)")
print(f"Sum of all PMF values = {sum(X.pmf(x_vals)):.6f}")
print(f"Mean = {X.mean()}, Variance = {X.var()}")
```
