# Bernoulli Distribution

The **Bernoulli** distribution is the simplest parametric random variable — a single yes/no trial.

| Property | Value |
|----------|-------|
| **Notation** | $X \sim \text{Bern}(p)$ |
| **Description** | 1 with probability $p$, 0 otherwise |
| **Parameter** | $p$ = probability of success |
| **Support** | $\{0, 1\}$ |
| **PMF** | $P(X = 1) = p, \quad P(X = 0) = 1 - p$ |
| **Expectation** | $E[X] = p$ |
| **Variance** | $\text{Var}(X) = p(1 - p)$ |

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Libraries loaded!")
```

---
## 1. PMF Visualization

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

p_values = [0.2, 0.5, 0.7, 0.9]
for ax, p in zip(axes, p_values):
    ax.bar([0, 1], [1 - p, p], color=['#e74c3c', '#2ecc71'], edgecolor='black',
           linewidth=1.5, width=0.5)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['0 (fail)', '1 (success)'], fontsize=9)
    ax.set_ylim(0, 1.1)
    ax.set_title(f'Bern(p={p})', fontsize=12, fontweight='bold')
    ax.set_ylabel('Probability')
    for x_val, prob in [(0, 1-p), (1, p)]:
        ax.text(x_val, prob + 0.03, f'{prob:.2f}', ha='center', fontsize=11, fontweight='bold')

plt.suptitle('Bernoulli PMF for Different p Values', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

---
## 2. Expectation and Variance Derivation

**Expectation:**
$$E[X] = 0 \cdot (1-p) + 1 \cdot p = p$$

**Variance:**
$$E[X^2] = 0^2 \cdot (1-p) + 1^2 \cdot p = p$$
$$\text{Var}(X) = E[X^2] - (E[X])^2 = p - p^2 = p(1-p)$$

```python
# Verify expectation and variance for different p values
print(f"{'p':>5s} | {'E[X]':>6s} | {'Var(X)':>8s} | {'Std(X)':>8s} | {'Sim E[X]':>9s} | {'Sim Var':>9s}")
print("-" * 60)
np.random.seed(42)

for p in [0.1, 0.3, 0.5, 0.7, 0.9]:
    rv = stats.bernoulli(p)
    samples = rv.rvs(size=100000)
    print(f"{p:5.1f} | {rv.mean():6.3f} | {rv.var():8.4f} | {rv.std():8.4f} | "
          f"{samples.mean():9.4f} | {samples.var():9.4f}")

# Variance is maximized at p = 0.5
p_range = np.linspace(0, 1, 100)
var_vals = p_range * (1 - p_range)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(p_range, var_vals, 'b-', linewidth=2.5)
ax.axvline(0.5, color='red', linestyle='--', linewidth=1.5, label='max at p=0.5')
ax.set_xlabel('p', fontsize=12)
ax.set_ylabel('Var(X) = p(1-p)', fontsize=12)
ax.set_title('Bernoulli Variance vs p', fontsize=13, fontweight='bold')
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()
print("Maximum variance at p = 0.5: Var = 0.25")
```

## 3. Indicator Random Variables

An **indicator** variable $I_A$ is a Bernoulli that equals 1 when event $A$ occurs:

$$I_A = \begin{cases} 1 & \text{if } A \text{ occurs} \\ 0 & \text{otherwise} \end{cases}$$

Key property: $E[I_A] = P(A)$

```python
# Indicator variable example: P(die roll > 4)
np.random.seed(42)
rolls = np.random.randint(1, 7, size=10000)
indicator = (rolls > 4).astype(int)  # 1 if roll > 4, else 0

print("Indicator: I = 1 if die roll > 4")
print(f"  P(roll > 4) = {2/6:.4f}  (theory)")
print(f"  E[I]        = {indicator.mean():.4f}  (simulation)")
print(f"  Var(I)      = {indicator.var():.4f}  (theory: {(2/6)*(4/6):.4f})")
```

## 4. Python: Working with Bernoulli

```python
from scipy.stats import bernoulli

p = 0.7
rv = bernoulli(p)

print(f"Bernoulli(p={p}):")
print(f"  PMF: P(X=0) = {rv.pmf(0):.2f}, P(X=1) = {rv.pmf(1):.2f}")
print(f"  CDF: P(X≤0) = {rv.cdf(0):.2f}, P(X≤1) = {rv.cdf(1):.2f}")
print(f"  Mean = {rv.mean():.2f}, Var = {rv.var():.4f}")
print(f"  Samples: {rv.rvs(size=20, random_state=42)}")
```
