# Poisson Distribution

The **Poisson** distribution models the number of events in a fixed interval (time, space, etc.) when events occur independently at a constant average rate.

| Property | Value |
|----------|-------|
| **Notation** | $X \sim \text{Poi}(\lambda)$ |
| **Description** | Number of events in a fixed interval |
| **Parameter** | $\lambda$ = average rate (events per interval) |
| **Support** | $\{0, 1, 2, \ldots\}$ (unbounded!) |
| **PMF** | $P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$ |
| **Expectation** | $E[X] = \lambda$ |
| **Variance** | $\text{Var}(X) = \lambda$ |

**Key insight:** Mean = Variance = $\lambda$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from math import factorial, exp

print("Libraries loaded!")
```

---
## 1. PMF Visualization — Different Rates

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

lambdas = [1, 5, 15]
colors = ['#e74c3c', '#2ecc71', '#3498db']

for ax, lam, color in zip(axes, lambdas, colors):
    rv = stats.poisson(lam)
    x = np.arange(0, int(lam + 4 * np.sqrt(lam)) + 1)
    pmf = rv.pmf(x)
    
    ax.bar(x, pmf, color=color, edgecolor='black', linewidth=0.8, alpha=0.8)
    ax.axvline(lam, color='black', linewidth=2, linestyle='--',
               label=f'E[X] = λ = {lam}')
    ax.set_title(f'Poi(λ={lam})', fontsize=13, fontweight='bold')
    ax.set_xlabel('k (events)')
    ax.set_ylabel('P(X = k)')
    ax.legend(fontsize=10)

plt.suptitle('Poisson PMF — More Events = More Spread', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
print("Notice: as λ grows, the distribution becomes more symmetric (approaches Normal)")
```

---
## 2. From Binomial to Poisson

When $n$ is large and $p$ is small (rare events), with $\lambda = np$:

$$\text{Bin}(n, p) \approx \text{Poi}(\lambda = np)$$

This is called the **Poisson limit theorem**.

```python
# Demonstrate Binomial → Poisson convergence
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

lam = 5  # target lambda
configs = [(10, 0.5), (50, 0.1), (500, 0.01)]  # n, p pairs all with np=5

for ax, (n, p) in zip(axes, configs):
    rv_bin = stats.binom(n, p)
    rv_poi = stats.poisson(lam)
    x = np.arange(0, 15)
    
    ax.bar(x - 0.2, rv_bin.pmf(x), 0.4, color='#3498db', edgecolor='black',
           alpha=0.7, label=f'Bin({n}, {p})')
    ax.bar(x + 0.2, rv_poi.pmf(x), 0.4, color='#e74c3c', edgecolor='black',
           alpha=0.7, label=f'Poi({lam})')
    ax.set_title(f'n={n}, p={p}, np={n*p}', fontsize=12, fontweight='bold')
    ax.set_xlabel('k')
    ax.set_ylabel('P(X = k)')
    ax.legend(fontsize=9)

plt.suptitle('Binomial → Poisson as n↑ and p↓ (np = λ = 5)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 3. Practical Examples

```python
from scipy.stats import poisson

# Example 1: Server requests
lam_requests = 10  # avg 10 requests/minute
X = poisson(lam_requests)
print("Server: avg 10 requests/minute")
print(f"  P(exactly 10) = {X.pmf(10):.4f}")
print(f"  P(at most 5)  = {X.cdf(5):.4f}")
print(f"  P(more than 15) = {1 - X.cdf(15):.4f}")

# Example 2: Typos per page
lam_typos = 2  # avg 2 typos/page
Y = poisson(lam_typos)
print(f"\nTypos: avg 2 per page")
print(f"  P(no typos)     = {Y.pmf(0):.4f}")
print(f"  P(at most 1)    = {Y.cdf(1):.4f}")
print(f"  P(3 or more)    = {1 - Y.cdf(2):.4f}")

# Example 3: Changing the interval
# If 10 requests/min, then 20 requests/2min
Z = poisson(20)
print(f"\nServer over 2 minutes (λ=20):")
print(f"  P(exactly 20) = {Z.pmf(20):.4f}")
print(f"  P(15 to 25)   = {Z.cdf(25) - Z.cdf(14):.4f}")
```

---
## 4. Simulation

```python
np.random.seed(42)
lam = 7
samples = stats.poisson.rvs(lam, size=100000)

fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(0, 20)

# Empirical vs theory
emp = [np.mean(samples == k) for k in x]
theory = stats.poisson.pmf(x, lam)

ax.bar(x - 0.2, emp, 0.4, color='#e74c3c', alpha=0.7, label='Simulation (n=100K)', edgecolor='black')
ax.bar(x + 0.2, theory, 0.4, color='#3498db', alpha=0.7, label='Theory', edgecolor='black')
ax.set_xlabel('k', fontsize=12)
ax.set_ylabel('P(X = k)', fontsize=12)
ax.set_title(f'Poisson(λ={lam}): Simulation vs Theory', fontsize=13, fontweight='bold')
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()

print(f"Theory:     E[X] = {lam}, Var(X) = {lam}")
print(f"Simulation: E[X] = {samples.mean():.4f}, Var(X) = {samples.var():.4f}")
```

## Summary

| Property | Formula |
|----------|---------|
| PMF | $P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$ |
| $E[X]$ | $\lambda$ |
| $\text{Var}(X)$ | $\lambda$ |
| Limit | $\text{Bin}(n,p) \to \text{Poi}(np)$ as $n\to\infty, p\to 0$ |
| Python | `scipy.stats.poisson(mu=λ)` |
