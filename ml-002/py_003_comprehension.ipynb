## 1. List Comprehension

A **list comprehension** builds a list by applying a transformation to each item in an iterable, with an optional filter.

**General syntax**
```
[expression  for item in iterable  if condition]
```

**Equivalent for-loop**
```python
result = []
for item in iterable:
    if condition:          # optional
        result.append(expression)
```

The `if condition` clause is optional — omit it to include every element.

````python
# Transform: square each number
squares = [x**2 for x in range(1, 6)]
print("Squares:", squares)

# Transform strings
words = ["hello", "world", "python"]
upper = [w.upper() for w in words]
print("Upper:", upper)

# Extract a field from a list of dicts
people = [{"name": "Alice", "age": 30}, {"name": "Bob", "age": 25}]
names = [p["name"] for p in people]
print("Names:", names)
````

````python
# Filter: keep only even numbers
evens = [x for x in range(1, 11) if x % 2 == 0]
print("Evens:", evens)

# Filter + transform in one step
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]
print("Even squares:", even_squares)

# Filter strings by length
words = ["cat", "elephant", "dog", "hippopotamus", "rat"]
long_words = [w for w in words if len(w) > 4]
print("Long words:", long_words)
````

````python
# Flatten a 2D list — two for clauses, left to right
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row]
print("Flat:", flat)

# Cartesian product
pairs = [(x, y) for x in [1, 2, 3] for y in ["a", "b"]]
print("Pairs:", pairs)

# Build a multiplication table as a list of lists
table = [[i * j for j in range(1, 5)] for i in range(1, 5)]
for row in table:
    print(row)
````

## 2. Dictionary Comprehension

A **dictionary comprehension** builds a `dict` by producing key-value pairs from an iterable.

**General syntax**
```
{key_expr: value_expr  for item in iterable  if condition}
```

**Equivalent for-loop**
```python
result = {}
for item in iterable:
    if condition:          # optional
        result[key_expr] = value_expr
```

````python
# Invert a dictionary
original = {"a": 1, "b": 2, "c": 3}
inverted = {v: k for k, v in original.items()}
print("Inverted:", inverted)

# Build a dict from two parallel lists
keys   = ["name", "age", "city"]
values = ["Alice", 30, "NYC"]
person = {k: v for k, v in zip(keys, values)}
print("Person:", person)

# Transform values — apply a 10 % discount
prices     = {"apple": 1.0, "banana": 0.5, "cherry": 2.0}
discounted = {item: round(price * 0.9, 2) for item, price in prices.items()}
print("Discounted:", discounted)
````

````python
# Filter: keep only items above a price threshold
prices    = {"apple": 1.0, "banana": 0.5, "cherry": 2.0, "date": 3.0}
expensive = {k: v for k, v in prices.items() if v > 1.0}
print("Expensive:", expensive)

# Conditional value expression — categorise each number
parity = {n: ("even" if n % 2 == 0 else "odd") for n in range(1, 9)}
print("Parity:", parity)

# Map words to their lengths, filtering short words
words      = ["cat", "elephant", "dog", "hippopotamus", "rat"]
word_lens  = {w: len(w) for w in words if len(w) > 3}
print("Word lengths:", word_lens)
````

## 3. Set Comprehension

A **set comprehension** builds a `set` — like a list comprehension but with curly braces and automatic deduplication.

**Syntax**
```
{expression  for item in iterable  if condition}
```

> Distinguish from a dict comprehension by the absence of the `:` separator.

````python
# Duplicate inputs collapse into unique outputs
nums           = [1, -1, 2, -2, 3, 3, -3]
unique_squares = {x**2 for x in nums}
print("Unique squares:", unique_squares)

# Unique first characters
words         = ["apple", "banana", "avocado", "blueberry", "cherry"]
first_letters = {w[0] for w in words}
print("First letters:", first_letters)

# Find common elements using set comprehension + membership test
a      = [1, 2, 3, 4, 5]
b      = [3, 4, 5, 6, 7]
common = {x for x in a if x in b}
print("Common:", common)
````

## 4. Advantages

| Advantage | Detail |
|-----------|--------|
| **Concise** | One line replaces 4–5 lines of loop boilerplate |
| **Faster** | CPython optimises comprehensions with `LIST_APPEND` bytecode, avoiding attribute lookup for `.append` |
| **Readable** | For simple transforms, intent is clear at a glance |
| **Functional style** | Maps naturally to `map` / `filter` mental model |
| **Scope-safe** | Loop variable does not leak into the enclosing scope (unlike a regular `for` loop in Python 2) |
| **Composable** | Can be nested or chained to build pipelines |

````python
import timeit

n = 100_000

loop_time = timeit.timeit(
    "result = []\nfor x in range(n):\n    result.append(x**2)",
    globals={"n": n}, number=100
)
comp_time = timeit.timeit(
    "[x**2 for x in range(n)]",
    globals={"n": n}, number=100
)

print(f"For-loop:            {loop_time:.3f} s")
print(f"List comprehension:  {comp_time:.3f} s")
print(f"Speedup:             {loop_time / comp_time:.2f}x faster")
````

````python
# Same logic, two styles — comprehension is shorter and reads like English
data = [3, -1, 4, -1, 5, -9, 2, 6]

# For-loop version
positives_loop = []
for x in data:
    if x > 0:
        positives_loop.append(x * 2)

# Comprehension version
positives_comp = [x * 2 for x in data if x > 0]

print("Loop result:", positives_loop)
print("Comp result:", positives_comp)
assert positives_loop == positives_comp
````

## 5. Disadvantages

| Disadvantage | Detail |
|-------------|--------|
| **Readability degrades with complexity** | Two or more nested loops, multiple conditions, or long expressions make comprehensions harder to read than a loop |
| **Eager evaluation** | The entire result list is built in memory immediately; use a generator expression `(...)` when you only need to iterate once |
| **No early exit** | Cannot `break` mid-comprehension; you must consume (or filter) all items |
| **Debugging difficulty** | A single-expression comprehension is harder to step through with a debugger than a loop |
| **Side effects are an anti-pattern** | Using a comprehension purely to call a function for its side effects (e.g., printing) wastes memory and signals wrong intent |
| **Not always faster** | When the bottleneck is the expression itself (heavy computation), the speedup from avoiding `.append` is negligible |

````python
import sys

# 1. Readability — complex nesting is hard to parse
matrix = [[1, 2], [3, 4], [5, 6]]

# Comprehension: compact but requires mental parsing
flat_odd_comp = [x for row in matrix for x in row if x % 2 != 0]

# Loop: longer but immediately clear
flat_odd_loop = []
for row in matrix:
    for x in row:
        if x % 2 != 0:
            flat_odd_loop.append(x)

print("Odd values:", flat_odd_comp)
assert flat_odd_comp == flat_odd_loop

# 2. Memory: list comprehension allocates the whole list at once
list_comp = [x**2 for x in range(10_000)]
gen_expr  = (x**2 for x in range(10_000))   # generator expression
print(f"\nList comp size:      {sys.getsizeof(list_comp):>8,} bytes")
print(f"Generator expr size: {sys.getsizeof(gen_expr):>8,} bytes")

# 3. Side effects — anti-pattern
data = [1, 2, 3]
# Bad (works, but wastes a list and is misleading):
# _ = [print(x) for x in data]

# Correct: use a loop when you only want side effects
print()
for x in data:
    print("item:", x)
````

## 6. Summary

| Type | Syntax | Result type | Deduplicates |
|------|--------|-------------|:------------:|
| List comprehension | `[expr for x in it if cond]` | `list` | No |
| Dict comprehension | `{k: v for x in it if cond}` | `dict` | By key |
| Set comprehension | `{expr for x in it if cond}` | `set` | Yes |
| Generator expression | `(expr for x in it if cond)` | `generator` | No (lazy) |

### When to use a comprehension

| Situation | Recommendation |
|-----------|----------------|
| Simple transform or filter | **Use a comprehension** — concise and fast |
| Multiple nested loops or conditions | **Use a for-loop** — easier to read and debug |
| Only iterating once over a large sequence | **Use a generator expression** — saves memory |
| Need to break or continue mid-iteration | **Use a for-loop** — comprehensions have no break |
| Need to call a function for its side effects | **Use a for-loop** — comprehensions imply a return value |
