# Dot Product

This notebook defines a custom function to compute the dot product of two vectors and verifies the result using NumPy.

```python
import numpy as np
```

## Custom Dot Product Function

```python
def dot_product(v1, v2):
    if len(v1) != len(v2):
        raise ValueError('Vectors must have the same length.')

    total = 0
    for i in range(len(v1)):
        total += v1[i] * v2[i]

    return total
```

## Verify Using NumPy

```python
import time

vec1 = np.array([1, 3, -5], dtype=float)
vec2 = np.array([4, -2, -1], dtype=float)

res_1 = dot_product(vec1, vec2)
res_2 = np.dot(vec1, vec2)

# Benchmark both approaches over many runs for more stable timing
runs = 100000

time_1 = time.perf_counter()
for _ in range(runs):
    dot_product(vec1, vec2)
time_1 = time.perf_counter() - time_1

time_2 = time.perf_counter()
for _ in range(runs):
    np.dot(vec1, vec2)
time_2 = time.perf_counter() - time_2

print(f'Vector 1: {vec1}')
print(f'Vector 2: {vec2}')
print(f'Custom dot product: {res_1}')
print(f'NumPy dot product: {res_2}')
print(f'Match: {np.isclose(res_1, res_2)}')
print()
print(f'Custom function time over {runs} runs: {time_1:.6f} s')
print(f'NumPy dot time over {runs} runs: {time_2:.6f} s')
```
