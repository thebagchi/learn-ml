# Mathematics Basics Starter
This notebook introduces basic mathematics concepts using NumPy.

### Creating one dimentional array

```python
import numpy as np

# Example: create an array and compute its mean
arr = np.array([1,2,3,4,5])
print('Array:', arr)
print('Mean:', arr.mean())
print('Transpose:', arr.transpose())
print('Size: ', arr.size)
print('Shape: ', arr.shape)
print('Dimension: ', arr.ndim)

# Additional useful prints for arr
print('Data type:', arr.dtype)
print('Item size (bytes per element):', arr.itemsize)
print('Total bytes:', arr.nbytes)
print('Sum:', arr.sum())
print('Standard deviation:', arr.std())
print('Variance:', arr.var())
print('Minimum:', arr.min())
print('Maximum:', arr.max())
print('Cumulative sum:', arr.cumsum())
print('Cumulative product:', arr.cumprod())
print('Sorted array:', np.sort(arr))
print('Unique elements:', np.unique(arr))
print('Flattened array:', arr.flatten())

# get rows and columns from arr.shape
if len(arr.shape) == 1:
    rows = arr.shape[0]
    cols = 1  # assume 1 column for 1D array
else:
    rows, cols = arr.shape

print("Rows:", rows, "Cols:", cols)

# use them in reshape
arr = arr.reshape(rows, cols)
print("Reshaped array:")
print(arr)
print("New shape:", arr.shape)
```

### Other ways of creating one dimensional array

```python
arr_1 = np.array([1,2,3])
print(arr_1)

arr_2 = np.arange(1,4,1)
print(arr_2)

arr_3 = np.arange(3)
print(arr_3)

arr_4 = np.linspace(1, 4, 3, dtype=int)
print(arr_4)

arr_5 = np.ones(3, dtype=int)
print(arr_5)

arr_6 = np.zeros(3, dtype=int)
print(arr_6)

arr_7 = np.empty(3, dtype=int)
print(arr_7)

arr_8 = np.random.rand(3)
print(arr_8)

# Random integers instead of floats
arr_9 = np.random.randint(0, 10, 3)  # random ints from 0 to 9
print(arr_9)
```

### Creating multidimentional array (matrix)

```python
# 2D array examples
mat = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array from list:")
print(mat)
print("Shape: ", mat.shape)
print("Transpose: ")
print(mat.transpose())
print('Size: ', mat.size)
print('Shape: ', mat.shape)
print('Dimension: ', mat.ndim)

# get rows and columns from mat.shape
if len(mat.shape) == 1:
    rows = mat.shape[0]
    cols = 1  # assume 1 column for 1D array
else:
    rows, cols = mat.shape

print("Rows:", rows, "Cols:", cols)

# use them in reshape
mat = mat.reshape(cols, rows)
print("Reshaped matrix:")
print(mat)
print("New shape:", mat.shape)
```

### Other ways of creating multidimensional array (matrix)

```python
mat_1 = np.zeros((2, 3), dtype=int)
print("2D array of zeros:")
print(mat_1)

mat_2 = np.ones((2, 3), dtype=int)
print("2D array of ones:")
print(mat_2)

mat_3 = np.random.randint(0, 10, (2, 3))
print("2D array of random integers:")
print(mat_3)

mat_4 = np.eye(3, dtype=int)  # Identity matrix
print("Identity matrix:")
print(mat_4)


print("Reshaping array:")
# 1-D array 
arr = np.array([1, 2, 3, 4, 5, 6])

# Multidimensional array using reshape()
mat = np.reshape(
        arr, # the array to be reshaped
        (2,3) # dimensions of the new array
    )
# Print the new 2-D array with two rows and three columns
print(mat)
```
