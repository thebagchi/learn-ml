````python
import numpy as np
import matplotlib.pyplot as plt
````

# Solving Linear Equations

### Plotting lines

````python
# Example: Plotting the equation y = 2x + 1
x = np.linspace(-5, 5, 100)
y = 2 * x + 1

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 2x + 1', color='blue')
plt.title('Plot of Linear Equation')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
````

## Representing equations as matrix

This represents the system:
$$ -x_{1} + 3x_{2} = 7 $$
$$ 3x_{1} + 2x_{2} = 1 $$

**Coefficient Matrix:**
```
[ -1    3 ]
[  3    2 ]
```

**Output Matrix:**
```
[ 7 ]
[ 1 ]
```

````python
coefficient_matrix = np.array([
    [-1, 3],
    [3, 2]
])
print(coefficient_matrix)

output_matrix = np.array([7,1])
print(output_matrix)

solutions = np.linalg.solve(coefficient_matrix, output_matrix)
print("The solution to these equations are: ")
print(solutions)

system_matrix = np.hstack((coefficient_matrix, output_matrix.reshape(2,1)))
print("System: ")
print(system_matrix)
````

### Plot the equations (visual verification)

````python
# Plotting the system of equations to find intersection
x = np.linspace(-2, 4, 100)

# Equation 1: -x + 3y = 7 ⇒ y = (7 + x)/3
y1 = (7 + x) / 3

# Equation 2: 3x + 2y = 1 ⇒ y = (1 - 3x)/2
y2 = (1 - 3 * x) / 2

plt.figure(figsize=(10, 8))
plt.plot(x, y1, label='$-x + 3y = 7$', color='blue', linewidth=2)
plt.plot(x, y2, label='$3x + 2y = 1$', color='red', linewidth=2)

# Use the solution from previous cell
x_sol, y_sol = solutions

plt.plot(x_sol, y_sol, 'ko', markersize=8, label=f'Solution: ({x_sol:.2f}, {y_sol:.2f})')
plt.axvline(x=x_sol, color='black', linestyle='--', alpha=0.5)
plt.axhline(y=y_sol, color='black', linestyle='--', alpha=0.5)

plt.xlabel('x₁')
plt.ylabel('x₂')
plt.title('System of Linear Equations - Graphical Solution')
plt.grid(True, alpha=0.3)
plt.legend()
plt.axis('equal')
plt.show()

print(f"Solution: x₁ = {x_sol:.4f}, x₂ = {y_sol:.4f}")
````
