The **Long Short-Term Memory** (LSTM) and **Gated Recurrent Unit** (GRU) solve the vanishing gradient problem by introducing learned gates that control information flow through a **cell state** (LSTM) or directly through the hidden state (GRU).

## Learning Objectives

1. Write the GRU update and reset gate equations
2. Write all four LSTM gate equations and the cell-state update
3. Explain how gates prevent vanishing gradients
4. Compare GRU and LSTM parameter counts and use cases

## GRU Equations

$$\tilde{c}^{\langle t \rangle} = \tanh(W_c [c^{\langle t-1 \rangle}, x^{\langle t \rangle}] + b_c)$$
$$\Gamma_u = \sigma(W_u [c^{\langle t-1 \rangle}, x^{\langle t \rangle}] + b_u) \quad \text{(update gate)}$$
$$\Gamma_r = \sigma(W_r [c^{\langle t-1 \rangle}, x^{\langle t \rangle}] + b_r) \quad \text{(reset gate)}$$
$$c^{\langle t \rangle} = \Gamma_u \odot \tilde{c}^{\langle t \rangle} + (1 - \Gamma_u) \odot c^{\langle t-1 \rangle}$$

## LSTM Equations

$$\Gamma_f = \sigma(W_f [a^{\langle t-1 \rangle}, x^{\langle t \rangle}] + b_f) \quad \text{(forget gate)}$$
$$\Gamma_i = \sigma(W_i [a^{\langle t-1 \rangle}, x^{\langle t \rangle}] + b_i) \quad \text{(input gate)}$$
$$\Gamma_o = \sigma(W_o [a^{\langle t-1 \rangle}, x^{\langle t \rangle}] + b_o) \quad \text{(output gate)}$$
$$\tilde{c}^{\langle t \rangle} = \tanh(W_c [a^{\langle t-1 \rangle}, x^{\langle t \rangle}] + b_c)$$
$$c^{\langle t \rangle} = \Gamma_f \odot c^{\langle t-1 \rangle} + \Gamma_i \odot \tilde{c}^{\langle t \rangle}$$
$$a^{\langle t \rangle} = \Gamma_o \odot \tanh(c^{\langle t \rangle})$$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'figure.facecolor': '#f5f5f7', 'axes.facecolor': '#ffffff',
    'axes.edgecolor': '#c8ccd4', 'axes.labelcolor': '#1a1d27',
    'axes.titlecolor': '#1a1d27', 'xtick.color': '#2a2e3a',
    'ytick.color': '#2a2e3a', 'grid.color': '#e0e3ea',
    'grid.linestyle': '--', 'grid.alpha': 0.5,
    'text.color': '#1a1d27', 'font.family': 'DejaVu Sans',
    'axes.titlesize': 12, 'axes.labelsize': 11,
    'legend.facecolor': '#ffffff', 'legend.edgecolor': '#c8ccd4',
    'figure.dpi': 110,
})
P = ['#5b9bd5', '#e05c5c', '#f4b942', '#7ecba1', '#56b6c2', '#c678dd']

sigmoid = lambda z: 1 / (1 + np.exp(-z))

class LSTMCell:
    def __init__(self, n_x, n_a, seed=0):
        np.random.seed(seed)
        s = 0.05
        concat = n_a + n_x
        self.Wf = np.random.randn(n_a, concat) * s;  self.bf = np.zeros((n_a, 1))
        self.Wi = np.random.randn(n_a, concat) * s;  self.bi = np.zeros((n_a, 1))
        self.Wo = np.random.randn(n_a, concat) * s;  self.bo = np.zeros((n_a, 1))
        self.Wc = np.random.randn(n_a, concat) * s;  self.bc = np.zeros((n_a, 1))

    def step(self, x_t, a_prev, c_prev):
        z = np.vstack([a_prev, x_t])
        f = sigmoid(self.Wf @ z + self.bf)
        i = sigmoid(self.Wi @ z + self.bi)
        o = sigmoid(self.Wo @ z + self.bo)
        ct = np.tanh(self.Wc @ z + self.bc)
        c  = f * c_prev + i * ct
        a  = o * np.tanh(c)
        return a, c, {'f': f, 'i': i, 'o': o}

class GRUCell:
    def __init__(self, n_x, n_a, seed=0):
        np.random.seed(seed)
        s = 0.05
        concat = n_a + n_x
        self.Wu = np.random.randn(n_a, concat) * s;  self.bu = np.zeros((n_a, 1))
        self.Wr = np.random.randn(n_a, concat) * s;  self.br = np.zeros((n_a, 1))
        self.Wc = np.random.randn(n_a, concat) * s;  self.bc = np.zeros((n_a, 1))

    def step(self, x_t, c_prev):
        z = np.vstack([c_prev, x_t])
        u = sigmoid(self.Wu @ z + self.bu)
        r = sigmoid(self.Wr @ z + self.br)
        ct = np.tanh(self.Wc @ np.vstack([r * c_prev, x_t]) + self.bc)
        c  = u * ct + (1 - u) * c_prev
        return c, {'u': u, 'r': r}

# ---- Simulate a sequence ----
np.random.seed(1)
T, n_x, n_a = 20, 5, 16
lstm = LSTMCell(n_x, n_a)
gru  = GRUCell(n_x, n_a)

a_lstm = np.zeros((n_a, 1))
c_lstm = np.zeros((n_a, 1))
c_gru  = np.zeros((n_a, 1))

lstm_gates = {'f': [], 'i': [], 'o': []}
gru_gates  = {'u': [], 'r': []}
lstm_cells, gru_cells = [], []

for t in range(T):
    x = np.random.randn(n_x, 1)
    a_lstm, c_lstm, lg = lstm.step(x, a_lstm, c_lstm)
    c_gru,  gg         = gru.step(x, c_gru)
    lstm_cells.append(c_lstm[:4, 0])
    gru_cells.append(c_gru[:4, 0])
    for k in lstm_gates: lstm_gates[k].append(lg[k][:4, 0].mean())
    for k in gru_gates:   gru_gates[k].append(gg[k][:4, 0].mean())

# ---- Plots ----
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('LSTM vs GRU — Gate Activations & Cell States', fontsize=13, fontweight='bold')

# LSTM gates
ax = axes[0, 0]
for key, color in zip(['f','i','o'], [P[0], P[3], P[5]]):
    ax.plot(lstm_gates[key], color=color, lw=2, label=f'{key} gate')
ax.set_title('LSTM gate activations (mean over units)'); ax.set_xlabel('t')
ax.set_ylabel('σ output'); ax.legend(); ax.grid(True); ax.set_ylim(0, 1)

# GRU gates
ax = axes[0, 1]
for key, color in zip(['u','r'], [P[2], P[4]]):
    ax.plot(gru_gates[key], color=color, lw=2, label=f'{key} gate')
ax.set_title('GRU gate activations (mean over units)'); ax.set_xlabel('t')
ax.set_ylabel('σ output'); ax.legend(); ax.grid(True); ax.set_ylim(0, 1)

# LSTM cell state
ax = axes[1, 0]
for j in range(4):
    ax.plot([s[j] for s in lstm_cells], color=P[j], lw=1.5, label=f'unit {j}')
ax.set_title('LSTM cell state c⁽ᵗ⁾ (first 4 units)'); ax.set_xlabel('t')
ax.set_ylabel('cell state'); ax.legend(fontsize=8); ax.grid(True)

# GRU hidden state
ax = axes[1, 1]
for j in range(4):
    ax.plot([s[j] for s in gru_cells], color=P[j], lw=1.5, label=f'unit {j}')
ax.set_title('GRU hidden state c⁽ᵗ⁾ (first 4 units)'); ax.set_xlabel('t')
ax.set_ylabel('hidden state'); ax.legend(fontsize=8); ax.grid(True)

plt.tight_layout()
plt.show()

# ---- Parameter count comparison ----
def lstm_params(n_x, n_a): return 4 * n_a * (n_a + n_x + 1)
def gru_params(n_x, n_a):  return 3 * n_a * (n_a + n_x + 1)
def rnn_params(n_x, n_a):  return n_a * (n_a + n_x + 1)

sizes = [32, 64, 128, 256]
fig, ax = plt.subplots(figsize=(9, 4))
w = 0.25
for j, (label, fn, color) in enumerate([('RNN', rnn_params, P[0]),
                                          ('GRU', gru_params, P[2]),
                                          ('LSTM', lstm_params, P[1])]):
    counts = [fn(n_x, h) for h in sizes]
    x = np.arange(len(sizes)) + j * w
    ax.bar(x, counts, width=w, color=color, alpha=0.85, label=label)
ax.set_xticks(np.arange(len(sizes)) + w)
ax.set_xticklabels([f'n_a={h}' for h in sizes])
ax.set_ylabel('Parameter count')
ax.set_title(f'RNN vs GRU vs LSTM — Parameters  (n_x={n_x})')
ax.legend(); ax.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
