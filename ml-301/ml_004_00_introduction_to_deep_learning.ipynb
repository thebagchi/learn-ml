Deep learning is a subset of machine learning that uses multi-layered **neural networks** to learn representations directly from raw data — images, audio, text, etc.

| Property | Value |
|---|---|
| **Paradigm** | Supervised, unsupervised, reinforcement |
| **Core building block** | Artificial neuron (weighted sum + activation) |
| **Key advantage** | Learns feature hierarchies automatically |
| **Key challenge** | Requires large data, compute, careful tuning |

> **Key insight (Andrew Ng):** AI is the new electricity. Just as electricity transformed every industry 100 years ago, deep learning is doing the same today.

## What is a Neural Network?

The simplest neural network is **logistic regression** viewed as a single neuron:

$$\hat{y} = \sigma(w^T x + b), \qquad \sigma(z) = \frac{1}{1+e^{-z}}$$

Stack many such neurons in layers and you get a **deep neural network**:

Each hidden neuron computes: $a = g(w^T a^{prev} + b)$ where $g$ is an activation function.

### Why "Deep"?

| Depth | Architecture | Example use |
|---|---|---|
| 0 hidden layers | Logistic regression | Spam filter |
| 1 hidden layer | Shallow NN | Simple classification |
| 2–5 hidden layers | Deep NN | Image/speech recognition |
| 100+ layers | Very deep NN | ResNet, GPT |

### 2. Compute
GPUs and TPUs allow parallelising the billions of matrix multiplications inside deep networks.

### 3. Algorithmic Improvements
- Sigmoid → **ReLU** (solves vanishing gradients, trains 10× faster)
- Better optimisers: momentum, RMSprop, Adam
- Batch normalisation, dropout, residual connections


```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'figure.facecolor': '#f5f5f7', 'axes.facecolor': '#ffffff',
    'axes.edgecolor': '#c8ccd4', 'axes.labelcolor': '#1a1d27',
    'axes.titlecolor': '#1a1d27', 'xtick.color': '#2a2e3a',
    'ytick.color': '#2a2e3a', 'grid.color': '#e0e3ea',
    'grid.linestyle': '--', 'grid.alpha': 0.5,
    'text.color': '#1a1d27', 'font.family': 'DejaVu Sans',
    'axes.titlesize': 13, 'axes.labelsize': 11,
    'legend.facecolor': '#ffffff', 'legend.edgecolor': '#c8ccd4',
    'figure.dpi': 110,
})
P = ['#5b9bd5', '#e05c5c', '#f4b942', '#7ecba1', '#56b6c2', '#c678dd']

# --- Performance vs data scale ---
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('Why Deep Learning Has Taken Off', fontsize=14, fontweight='bold')

ax = axes[0]
m = np.linspace(0, 1, 300)

def lr_curve(m, cap):
    return cap * (1 - np.exp(-5 * m))

ax.plot(m, lr_curve(m, 0.65), color=P[2], lw=2, label='Logistic Regression / SVM')
ax.plot(m, lr_curve(m, 0.72), color=P[0], lw=2, label='Small Neural Network')
ax.plot(m, lr_curve(m, 0.85), color=P[1], lw=2.5, label='Large Neural Network')
ax.set_xlabel('Amount of labelled data (m)')
ax.set_ylabel('Performance')
ax.set_title('Scale drives deep learning')
ax.set_xticks([]); ax.set_yticks([])
ax.legend(loc='lower right', fontsize=9)
ax.grid(True)
ax.annotate('Plateau', xy=(0.7, 0.66), fontsize=9, color=P[2])
ax.annotate('Still improving!', xy=(0.6, 0.82), fontsize=9, color=P[1])

# --- Supervised learning applications ---
ax = axes[1]
applications = [
    'Image\nClassification', 'Speech\nRecognition', 'Machine\nTranslation',
    'Object\nDetection', 'Ad Click\nPrediction', 'Autonomous\nDriving'
]
architectures = ['CNN', 'RNN/LSTM', 'Seq2Seq', 'CNN+\nYOLO', 'Std NN', 'CNN+\nSensor']
colors = [P[0], P[1], P[2], P[3], P[4], P[5]]

y = np.arange(len(applications))
bars = ax.barh(y, [0.9, 0.85, 0.8, 0.88, 0.75, 0.92], color=colors, alpha=0.85, height=0.6)
ax.set_yticks(y)
ax.set_yticklabels(applications, fontsize=9)
ax.set_xlabel('Relative human-level performance match')
ax.set_title('Supervised DL Applications')
ax.set_xlim(0, 1.1)
ax.grid(True, axis='x')
for bar, arch in zip(bars, architectures):
    ax.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2,
            arch, va='center', fontsize=8, color='#555')

plt.tight_layout()
plt.show()
```

## Structured vs Unstructured Data

| Type | Examples | Notes |
|---|---|---|
| **Structured** | House prices, ad clicks, tabular records | Each feature has a clear meaning |
| **Unstructured** | Images, audio, text | Pixels / waveforms / tokens — humans excel naturally |

Deep learning has particularly **revolutionised unstructured data** tasks — humans had dominated these tasks for decades.

## The Iterative Cycle

Deep learning development is highly iterative:

$$\text{Idea} \xrightarrow{\text{code}} \text{Experiment} \xrightarrow{\text{results}} \text{Idea} \xrightarrow{\cdots}$$

Fast iteration (quick training) is critical. This is why GPUs, good frameworks (PyTorch, TensorFlow), and cloud compute matter so much.
