## Mathematical Foundations

### Terminology and Notation

**Nabla ($\nabla$):** The symbol $\nabla$ (pronounced "nabla" or "del") is the gradient operator used to denote the gradient of a function.

**Gradient Vector:** The **gradient vector** (also called the **gradient**) is the vector of all first-order partial derivatives of a function with respect to its variables. It points in the direction of steepest ascent.

### Gradient and Jacobian Matrix

The **gradient is a special case of the Jacobian matrix**. Understanding their relationship is crucial for optimization:

**Jacobian Matrix:** For a vector-valued function $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^m$, the Jacobian is the matrix of all first-order partial derivatives:

$$J = \begin{bmatrix} \frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\ \frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\ \vdots & \vdots & \ddots & \vdots \\ \frac{\partial f_m}{\partial x_1} & \frac{\partial f_m}{\partial x_2} & \cdots & \frac{\partial f_m}{\partial x_n} \end{bmatrix}$$

**Gradient Vector:** The gradient vector (denoted using nabla $\nabla$) is the Jacobian of a **scalar-valued function** $J: \mathbb{R}^n \to \mathbb{R}$ (when $m = 1$):

$$\nabla J(\theta) = \begin{bmatrix} \frac{\partial J}{\partial \theta_1} \\ \frac{\partial J}{\partial \theta_2} \\ \vdots \\ \frac{\partial J}{\partial \theta_n} \end{bmatrix}$$

In optimization, we work with scalar cost functions, so **the gradient vector is the Jacobian specialized for single-output functions**. It is a column vector (or row vector by convention) containing all first-order partial derivatives, and it points in the direction of steepest increase in the function value.


## 1. Batch Gradient Descent

### Key Mathematical Concepts

**Gradient Vector ($\nabla J(\theta)$):** The vector of partial derivatives of the cost function with respect to each parameter. The nabla symbol ($\nabla$) denotes the gradient operator.
$$\nabla J(\theta) = \begin{bmatrix} \frac{\partial J}{\partial \theta_0} \\ \frac{\partial J}{\partial \theta_1} \\ \vdots \\ \frac{\partial J}{\partial \theta_n} \end{bmatrix}$$

**Update Rule:** Parameters are updated based on the gradient vector computed over the entire dataset:
$$\theta := \theta - \alpha \nabla J(\theta)$$

Where $\alpha$ is the learning rate.

### Essential Concepts
- **Cost Function:** $J(\theta)$ - typically Mean Squared Error (MSE) or Cross-Entropy
- **Learning Rate ($\alpha$):** Controls step size; too large causes divergence, too small causes slow convergence
- **Convergence:** Process repeats until gradient vector approaches zero or max iterations reached
- **Dataset Size:** Uses all $m$ training examples per iteration

### When to Use
- Small to medium-sized datasets
- When you want guaranteed convergence (with proper learning rate)
- Offline learning scenarios


## 2. Stochastic Gradient Descent (SGD)

### Key Mathematical Concepts

**Per-Sample Gradient Vector:** Updates based on the gradient vector computed from a single training example:
$$\theta := \theta - \alpha \nabla J(\theta; x^{(i)}, y^{(i)})$$

Where $(x^{(i)}, y^{(i)})$ is a single training example, and $\nabla J(\theta; x^{(i)}, y^{(i)})$ is the gradient vector for that example.

### Essential Concepts
- **Noisy Updates:** Gradient vector estimates are noisy (based on single samples)
- **Oscillation:** Parameter updates oscillate around the minimum instead of smoothly converging
- **Variance:** High variance in gradient estimates requires careful learning rate scheduling
- **Update Frequency:** More frequent updates per epoch (one update per sample)

### Key Characteristics
- **Epoch:** One complete pass through the entire dataset (containing $m$ updates)
- **Noise Benefits:** Noise can help escape local minima (beneficial in non-convex optimization)
- **Learning Rate Schedule:** Often requires decreasing learning rate over time: $\alpha_t = \frac{\alpha_0}{1 + t}$

### When to Use
- Very large datasets
- Online learning (streaming data)
- Non-convex optimization problems
- Memory-constrained environments


## 3. Mini-Batch Gradient Descent

### Key Mathematical Concepts

**Batch Gradient Vector:** Updates based on the gradient vector computed over a subset (batch) of training data:
$$\theta := \theta - \alpha \frac{1}{b} \sum_{i=1}^{b} \nabla J(\theta; x^{(i)}, y^{(i)})$$

Where $b$ is the batch size (typically 32, 64, 128, etc.), and $\nabla J$ represents the gradient vector.

### Essential Concepts
- **Batch Size ($b$):** Critical hyperparameter balancing stability and speed
- **Gradient Vector Estimation:** Average gradient vector over $b$ samples reduces variance compared to SGD
- **Computational Efficiency:** Leverages vectorized operations for faster computation than pure SGD
- **Variance Reduction:** Lower variance than SGD, higher variance than batch GD

### Mini-Batch Properties
- **Stability:** More stable convergence than SGD due to gradient vector averaging
- **Updates per Epoch:** $\frac{m}{b}$ updates per epoch (between SGD and Batch GD)
- **Hardware Utilization:** Optimal balance for GPU/TPU processing
- **Number of Iterations:** Epoch = $\frac{m}{b}$ iterations

### When to Use
- Large datasets (most practical scenarios)
- Deep learning with modern hardware
- When you need a balance between stability and efficiency


## 4. Normal Equation

### Key Mathematical Concepts

**Closed-Form Solution:** Directly compute optimal parameters without iteration:
$$\theta = (X^T X)^{-1} X^T y$$

Where:
- $X$ is the design matrix (size: $m \times (n+1)$, with column of 1s for intercept)
- $y$ is the target vector (size: $m \times 1$)
- $\theta$ is the parameter vector

### Essential Concepts
- **Derivation Basis:** Derived by setting gradient vector to zero: $\nabla J(\theta) = 0$
- **Matrix Inversion:** Requires $(X^T X)$ to be invertible (full rank)
- **Computational Complexity:** $O(n^3)$ due to matrix inversion (where $n$ is number of features)
- **No Hyperparameters:** No learning rate or iteration count needed

### Essential Conditions
- **Invertibility:** $(X^T X)$ must be invertible (full rank)
- **Applicability:** Primarily for linear regression problems
- **Feature Scaling:** Not required (unlike gradient descent)

### Handling Non-Invertible Matrices: Pseudoinverse

**When $(X^T X)$ is singular or nearly singular**, the **Moore-Penrose pseudoinverse** $X^+$ provides a practical solution:

$$\theta = X^+ y$$

Where $X^+ = (X^T X)^{-1} X^T$ is replaced by the pseudoinverse.

**Key Benefits:**
- **Always computable:** Works even when $(X^T X)$ is singular or ill-conditioned
- **Least-squares solution:** Minimizes $\|X\theta - y\|^2$ even when exact solution doesn't exist
- **Numerical stability:** More robust to rounding errors and near-singular matrices
- **Unique solution:** Provides minimum-norm solution when multiple solutions exist

**When Singularity Occurs:**
- Redundant features (linearly dependent columns in $X$)
- More features than training examples ($n > m$)
- Features with perfect multicollinearity

### When to Use
- Small to medium datasets ($n < 10,000$ features)
- When you need exact solution (for linear regression)
- When interpretability is important
- Offline learning scenarios

### When NOT to Use
- Very large feature sets (matrix inversion becomes prohibitively expensive)
- Streaming/online learning
- Non-linear models
- When numerical stability is critical and pseudoinverse cannot be computed reliably


## 5. Locally Weighted Linear Regression (LWLR)

### Key Mathematical Concepts

**Weighted Cost Function:** Fit a separate local model for every query point $x$ by assigning a weight $w^{(i)}$ to each training example:
$$J(\theta) = \sum_{i=1}^{m} w^{(i)}\bigl(y^{(i)} - \theta^T x^{(i)}\bigr)^2$$

**Gaussian Kernel (weight function):**
$$w^{(i)} = \exp\!\left(-\frac{\bigl(x^{(i)} - x\bigr)^2}{2\tau^2}\right)$$

Where $\tau$ is the **bandwidth** — controls how quickly weights fall off with distance.

**Closed-Form Solution:** Same as the Normal Equation but with weight matrix $W = \text{diag}(w^{(1)}, \ldots, w^{(m)})$:
$$\theta = \left(X^T W X\right)^{-1} X^T W y$$

### Essential Concepts
- **Weighted Least Squares:** Extension of OLS where each residual is scaled by a sample weight $w^{(i)}$
- **Diagonal Weight Matrix ($W$):** $m \times m$ matrix with kernel weights on the diagonal; encodes per-sample importance
- **Gaussian / RBF Kernel:** Exponential function mapping distance to similarity; produces values in $(0, 1]$
- **Bandwidth ($\tau$):** Controls the effective neighbourhood size — the key bias-variance trade-off parameter
- **Matrix Inversion / Solve:** $(X^TWX)^{-1}$ must be computed per query; requires the matrix to be full rank
- **Bias-Variance Trade-off:** Small $\tau$ → low bias, high variance (overfit); large $\tau$ → high bias, low variance (underfit)
- **Non-parametric Learning:** No fixed parameter vector; model complexity grows with the training set
- **Computational Cost:** $O(mn^2 + n^3)$ per prediction — matrix solve dominates for large feature count

### When to Use
- Non-linear relationships where a single linear model is insufficient
- Small to medium datasets

### When NOT to Use
- Large datasets (per-query cost scales with $m$)
- High-dimensional inputs

## 5. Variants of Linear Regression

Beyond standard Linear Regression, several important variants address different problem characteristics and constraints:

| Variant | Objective Function | Key Feature | Use Case |
|---------|-------------------|-------------|----------|
| **Ridge Regression (L2)** | $J(\theta) = \frac{1}{2m}\|X\theta - y\|^2 + \lambda\|\theta\|^2$ | Adds L2 penalty to shrink coefficients | Handles multicollinearity; prevents overfitting |
| **Lasso Regression (L1)** | $J(\theta) = \frac{1}{2m}\|X\theta - y\|^2 + \lambda\|\theta\|_1$ | Adds L1 penalty; performs feature selection | Automatic feature selection; sparse solutions |
| **Elastic Net** | $J(\theta) = \frac{1}{2m}\|X\theta - y\|^2 + \lambda_1\|\theta\|_1 + \lambda_2\|\theta\|^2$ | Combines L1 and L2 penalties | Balance between Ridge and Lasso benefits |
| **Polynomial Regression** | $J(\theta) = \frac{1}{2m}\|X'\theta - y\|^2$ (with polynomial features) | Fits polynomial curves to data | Non-linear relationships; higher-degree patterns |
| **Weighted Least Squares** | $J(\theta) = \frac{1}{2m}(X\theta - y)^T W(X\theta - y)$ | Different weights for different samples | Heteroscedasticity; some samples more reliable |
| **Robust Regression** | $J(\theta) = \sum_{i=1}^{m} \rho(y^{(i)} - \theta^T x^{(i)})$ | Uses robust loss functions (Huber, absolute value) | Outlier-resistant fitting |
| **Quantile Regression** | Minimizes weighted sum of absolute errors | Fits specific quantiles (median, quartiles) | Capturing distribution properties beyond mean |
| **Logistic Regression** | $J(\theta) = -\frac{1}{m}\sum_{i=1}^{m}[y^{(i)}\log(\hat{y}^{(i)}) + (1-y^{(i)})\log(1-\hat{y}^{(i)})]$ | Sigmoid activation for binary classification | Binary/multi-class classification problems |

### Key Regularization Concepts

**Regularization Parameter ($\lambda$):** Controls the trade-off between fitting the training data and model complexity:
- $\lambda = 0$: Standard Linear Regression (no penalty)
- $\lambda \to \infty$: Strong penalty, coefficients shrink toward zero

**L1 Penalty:** $\|\theta\|_1 = \sum_{j=1}^{n}|\theta_j|$
- Promotes sparsity (some coefficients exactly zero)
- Useful for automatic feature selection

**L2 Penalty:** $\|\theta\|^2 = \sum_{j=1}^{n}\theta_j^2$
- Shrinks all coefficients proportionally
- Helps with multicollinearity

## Parametric vs Non-Parametric Learning

### Parametric Learning

A **parametric** model summarises the data into a **fixed, finite set of parameters** $\theta$. Once trained, the training data can be discarded — all information is captured in $\theta$.

- Model complexity is **fixed** regardless of dataset size
- Fast at prediction time (just evaluate $\theta^T x$)
- Strong assumptions about the functional form of the data

| Algorithm | Parameters |
|---|---|
| Linear Regression | Weights $\theta \in \mathbb{R}^{n+1}$ |
| Logistic Regression | Weights $\theta \in \mathbb{R}^{n+1}$ |
| Linear SVM | Weight vector $\mathbf{w}$, bias $b$ |
| Naive Bayes | Class priors + feature likelihoods |
| Neural Networks | All layer weights and biases |

---

### Non-Parametric Learning

A **non-parametric** model makes **no fixed assumption** about the functional form. The model complexity grows with the data — the training set itself is retained at prediction time.

- More flexible; can fit complex, irregular patterns
- Slower at prediction time (depends on training data)
- Requires more data to generalise well

| Algorithm | What is stored |
|---|---|
| Locally Weighted Linear Regression (LWLR) | Entire training set; $\theta$ recomputed per query |
| k-Nearest Neighbours (k-NN) | Entire training set |
| Kernel SVM | Support vectors |
| Gaussian Process Regression | Training inputs + kernel parameters |
| Decision Trees (deep, unpruned) | Full tree structure grows with data |

---

### Key Differences

| Aspect | Parametric | Non-Parametric |
|---|---|---|
| **Model size** | Fixed ($\theta$ only) | Grows with training data |
| **Training data needed after fit** | No | Yes (instance-based) |
| **Assumptions** | Strong (e.g. linearity) | Weak / none |
| **Flexibility** | Lower | Higher |
| **Prediction speed** | Fast | Slower (scales with $m$) |
| **Risk** | Underfitting if assumptions wrong | Overfitting with small datasets |

## 6. Classification and Logistic Regression

### Key Mathematical Concepts

**Sigmoid (Logistic) Function:** Squashes any real value into $(0, 1)$, interpreted as a probability:
$$h_\theta(x) = \sigma(\theta^T x) = \frac{1}{1 + e^{-\theta^T x}}$$

**Decision Boundary:** The threshold separating predicted classes:
$$\hat{y} = \begin{cases} 1 & \text{if } h_\theta(x) \ge 0.5 \\ 0 & \text{otherwise} \end{cases}$$

**Cost Function (Log-Loss / Binary Cross-Entropy):** Derived from maximum likelihood estimation; penalises confident wrong predictions heavily:
$$J(\theta) = -\frac{1}{m}\sum_{i=1}^{m}\left[y^{(i)}\log h_\theta(x^{(i)}) + (1 - y^{(i)})\log\bigl(1 - h_\theta(x^{(i)})\bigr)\right]$$

**Gradient Update Rule:** Same form as linear regression but $h_\theta(x)$ is the sigmoid output:
$$\theta_j := \theta_j - \alpha \frac{1}{m}\sum_{i=1}^{m}\bigl(h_\theta(x^{(i)}) - y^{(i)}\bigr)x_j^{(i)}$$

### Essential Concepts
- **Sigmoid Function:** Differentiable, monotonic; maps $\mathbb{R} \to (0,1)$; output is the predicted probability $P(y=1 \mid x;\theta)$
- **Log-Loss:** Convex cost function; ensures gradient descent finds a global minimum
- **Maximum Likelihood Estimation (MLE):** Log-loss is derived by maximising the likelihood of observed labels under the Bernoulli distribution
- **Decision Boundary:** Linear in feature space ($\theta^T x = 0$ is the boundary); can be made non-linear with polynomial features
- **Softmax (Multi-class):** Generalises sigmoid to $K$ classes: $P(y=k \mid x) = \frac{e^{\theta_k^T x}}{\sum_{j=1}^K e^{\theta_j^T x}}$
- **Gradient:** $\nabla J(\theta) = \frac{1}{m} X^T (h - y)$ — identical in form to linear regression
- **Regularisation:** L1 / L2 penalty added to $J(\theta)$ to prevent overfitting

### When to Use
- Binary or multi-class classification
- When a probabilistic output is needed
- Linearly (or near-linearly) separable classes

### When NOT to Use
- Complex non-linear decision boundaries without feature engineering
- When classes are highly imbalanced without adjustment

## 7. Batch Gradient Ascent

### Key Mathematical Concepts

**Update Rule:** Instead of minimising a cost, gradient ascent **maximises** an objective $\ell(\theta)$ by stepping in the direction of the gradient:
$$\theta := \theta + \alpha \nabla \ell(\theta)$$

The only difference from gradient descent is the **`+`** sign.

**Log-Likelihood for Logistic Regression:** The canonical use case — maximise the log-likelihood of the training labels:
$$\ell(\theta) = \sum_{i=1}^{m}\left[y^{(i)}\log h_\theta(x^{(i)}) + (1 - y^{(i)})\log\bigl(1 - h_\theta(x^{(i)})\bigr)\right]$$

**Gradient of Log-Likelihood:**
$$\nabla \ell(\theta) = X^T\bigl(y - h_\theta(X)\bigr)$$

### Essential Concepts
- **Ascent vs Descent:** Gradient descent minimises $J(\theta)$; gradient ascent maximises $\ell(\theta)$; equivalent since $J = -\ell$
- **Log-Likelihood:** Objective derived from MLE — measures how well $\theta$ explains the observed labels
- **Same Convergence Properties:** Learning rate, convergence criteria, and batch behaviour are identical to gradient descent
- **Dataset Size:** Uses all $m$ training examples per update (batch)

### When to Use
- Maximising a likelihood or posterior (MLE / MAP estimation)
- Logistic regression trained via maximum likelihood
- Any probabilistic model where the objective is a log-likelihood

### When NOT to Use
- When the objective is a cost / loss to be minimised (use gradient descent instead)

## 8. Newton's Method

### Key Mathematical Concepts

**Update Rule:** Uses both the gradient **and** the Hessian (second-order information) to take a more informed step toward the optimum:
$$\theta := \theta - H^{-1} \nabla_\theta \ell(\theta)$$

**Hessian Matrix:** Matrix of all second-order partial derivatives of the objective:
$$H_{jk} = \frac{\partial^2 \ell(\theta)}{\partial \theta_j \,\partial \theta_k}$$

For logistic regression the Hessian is:
$$H = X^T D X$$

where $D$ is a diagonal matrix with $D_{ii} = h_\theta(x^{(i)})(1 - h_\theta(x^{(i)}))$.

### Essential Concepts
- **Second-Order Optimisation:** Gradient descent uses only $\nabla \ell$ (first derivative); Newton's method also uses $H$ (second derivative) to account for curvature
- **Hessian Matrix ($H$):** Captures the curvature of the objective; must be computed and inverted at every iteration — $O(n^3)$ cost
- **No Learning Rate:** Step size is determined automatically by $H^{-1}$; no $\alpha$ hyperparameter needed
- **Quadratic Convergence:** Converges much faster than gradient descent near the optimum (quadratic vs linear rate)
- **Newton-Raphson:** The scalar version; Newton's method generalises it to vector-valued $\theta$
- **Positive Definiteness:** Requires $H$ to be negative definite (for maximisation) to guarantee a descent/ascent step

### When to Use
- Small to medium feature sets ($n$ not too large — Hessian inversion is $O(n^3)$)
- When fast convergence is needed and few iterations are preferred
- Logistic regression and GLMs (Hessian has a closed form)

### When NOT to Use
- Large number of features (Hessian inversion becomes prohibitively expensive)
- Non-convex objectives (may converge to saddle points or diverge)
- Deep learning (use quasi-Newton methods like L-BFGS instead)

## 9. Digression: The Perceptron

### Historical Context

The **Perceptron** (Rosenblatt, 1958) is the classical **threshold-based linear classifier** that predates and influenced modern neural networks. While rarely used directly in modern ML, it teaches fundamental concepts about how algorithms learn decision boundaries.

### Key Mathematical Concepts

**Prediction Rule:** Output is a simple threshold on the dot product:
$$\hat{y}^{(i)} = \begin{cases} 1 & \text{if } \theta^T x^{(i)} \geq 0 \\ 0 & \text{otherwise} \end{cases}$$

**Update Rule:** Update parameters **only when a misclassification occurs**:
$$\theta := \theta + \eta \cdot x^{(i)} \quad \text{if } \hat{y}^{(i)} \neq y^{(i)}$$

Where $\eta$ is the learning rate and $(x^{(i)}, y^{(i)})$ is a misclassified example.

**No Explicit Cost Function:** Unlike logistic regression or gradient methods, the perceptron has **no global objective function** — it updates only on errors.

### Essential Concepts

- **Binary Classification Only:** Outputs 0 or 1 (no probability estimates)
- **Hard Threshold:** Deterministic decision boundary; no soft probabilities
- **Convergence Guarantee (Linearly Separable Data):** If data is linearly separable, the perceptron learning algorithm **guaranteed to converge in finite time**
- **Failed Convergence (Non-Separable Data):** Oscillates indefinitely if classes overlap
- **No Global Objective:** Minimises misclassification errors incrementally; no loss function to optimize
- **Learning Rate ($\eta$):** Typically set to 1; doesn't need tuning for linearly separable data
- **Sparse Updates:** Only updates on misclassified examples (efficient on well-separated data)

### Perceptron vs Logistic Regression

| Aspect | Perceptron | Logistic Regression |
|--------|-----------|-------------------|
| **Output** | Hard binary prediction {0, 1} | Probability in (0, 1) |
| **Decision Boundary** | Linear (hard threshold) | Linear (soft sigmoid) |
| **Cost Function** | None (error-driven) | Binary cross-entropy |
| **Convergence** | Finite (if separable) | Asymptotic (always) |
| **Non-Separable Data** | Fails to converge | Converges to best fit |
| **Gradient-Based** | No (heuristic updates) | Yes (gradient descent) |
| **Probabilistic** | No | Yes (MLE framework) |
| **Modern Use** | Rarely (mainly historical) | Very common |

### Key Insights

1. **Linearly Separable Case:** If classes can be perfectly separated by a line/hyperplane, the perceptron learns it in finite updates
2. **Non-Separable Case:** Algorithm never terminates — it oscillates trying to satisfy contradictory constraints
3. **Why It Matters Historically:** Foundation for neural networks; showed that simple threshold units can learn
4. **Why It's Outdated:** No probabilistic interpretation, fails on non-separable data, slower convergence than modern methods on large data

### When to Use
- Historical / educational value
- Extremely large data where you only correct misclassifications
- When computational resources are severely limited

### When NOT to Use
- Data is not linearly separable (will never converge)
- You need probability estimates
- You need global optimization with a well-defined objective
- Standard classifiers like logistic regression or SVM are available

## 10. Generalized Linear Models (GLMs)

### Overview

**Generalized Linear Models (GLMs)** are a flexible framework that unifies linear regression, logistic regression, and many other statistical models. GLMs extend ordinary linear regression by allowing the response variable to follow non-normal distributions and connecting predictors to the response through a **link function**.

### Key Mathematical Concepts

**Three Components of a GLM:**

1. **Random Component (Exponential Family):** The response $y$ follows a distribution from the exponential family:
   $$p(y | \eta) = b(y) \exp(\eta^T T(y) - a(\eta))$$
   
   Examples: Normal (linear regression), Bernoulli (logistic regression), Poisson (count regression)

2. **Systematic Component (Linear Predictor):** A linear combination of features:
   $$\eta = \theta^T x = \theta_0 + \theta_1 x_1 + \cdots + \theta_p x_p$$

3. **Link Function:** Connects the linear predictor $\eta$ to the mean $\mu = E[y|x]$:
   $$\eta = g(\mu)$$
   
   Where $g$ is the link function. The inverse link (response function) is:
   $$\mu = g^{-1}(\eta)$$

**Common GLM Families:**

| Family | Response Distribution | Link Function | Inverse Link | Example |
|--------|----------------------|---------------|--------------|---------|
| **Gaussian** | Normal | Identity: $g(\mu) = \mu$ | $\mu = \eta$ | Linear regression |
| **Binomial** | Bernoulli (binary) | Logit: $g(\mu) = \log(\frac{\mu}{1-\mu})$ | $\mu = \sigma(\eta) = \frac{1}{1+e^{-\eta}}$ | Logistic regression |
| **Poisson** | Poisson (counts) | Log: $g(\mu) = \log(\mu)$ | $\mu = e^\eta$ | Count regression |
| **Gamma** | Gamma | Inverse: $g(\mu) = 1/\mu$ | $\mu = 1/\eta$ | Continuous positive data |
| **Categorical** | Categorical (multi-class) | Multinomial logit | Softmax: $\mu_k = \frac{e^{\eta_k}}{\sum_j e^{\eta_j}}$ | Multi-class classification |

### Estimation: Maximum Likelihood Estimation (MLE)

GLMs are typically estimated via **Maximum Likelihood Estimation**. The likelihood under the exponential family is:

$$\ell(\theta) = \sum_{i=1}^m \left[\eta_i^T T(y^{(i)}) - a(\eta_i)\right] + \text{const}$$

Where $\eta_i = \theta^T x^{(i)}$.

The gradient of the log-likelihood has an elegant form:

$$\nabla_\theta \ell(\theta) = X^T (y - \mu)$$

Where $\mu = g^{-1}(X\theta)$ is the vector of predicted means. This is the same form as in linear regression! The key difference is that the link function $g^{-1}$ transforms the linear predictor to the appropriate scale for the response distribution.

### Optimization

GLMs are optimized using one of the methods covered earlier:

- **Gradient Descent / Ascent:** For large datasets; simple and scalable
- **Newton's Method:** Faster convergence for small-medium feature sets; uses Hessian (Fisher information matrix for GLMs)
- **Iteratively Reweighted Least Squares (IRLS):** A specialized Newton-type method for GLMs; efficient and numerically stable

### Essential Concepts

- **Unified Framework:** Linear regression, logistic regression, and many others are special cases of GLMs
- **Exponential Family:** Provides the theoretical foundation; ensures consistent probability models and well-defined MLEs
- **Link Function:** The key innovation; allows non-linear relationships between predictors and response while maintaining linear structure in $\eta$
- **Variance Function:** Related to the link through the mean-variance relationship of the exponential family; determines heteroscedasticity
- **Overdispersion:** When data variability exceeds the theoretical variance of the assumed distribution; indicates model misfit or need for quasi-likelihood approaches
- **Canonical Link:** The link function that corresponds directly to the natural parameter; yields simplifications in inference and computation
- **Deviance:** A measure of model goodness-of-fit; generalizes the residual sum of squares from linear regression
- **Likelihood Ratio Tests:** Used to compare nested GLMs; asymptotically $\chi^2$ distributed

### When to Use

- **Linear Regression:** Response is continuous, approximately normally distributed
- **Logistic Regression:** Binary classification; need probability estimates
- **Poisson Regression:** Response is count data (non-negative integers) with potential overdispersion
- **Multinomial Logit:** Multi-class classification (generalisation of logistic regression)
- **Gamma Regression:** Continuous positive responses (e.g., waiting times, costs)
- **Robust Modelling:** When relationship between predictors and response is non-linear and can be captured by a known transformation

### When NOT to Use

- Highly non-linear relationships not captured by any standard link function → use non-linear regression or machine learning methods
- Very large feature spaces ($p$ very large) → use regularized methods (ridge, lasso, elastic net GLMs)
- Time series data with dependencies → use time series models (ARIMA, state-space)
- Complex interactions and non-additive effects → use kernel methods, trees, or deep learning
- Unknown distributional assumptions → use robust or distribution-free methods

### GLMs in the Supervised Learning Pipeline

**GLMs bridge classical statistics and machine learning:**

- **From Statistics:** Probabilistic framework, hypothesis testing, inference via MLEs
- **From ML:** Scalable optimization, regularization, cross-validation for hyperparameter tuning
- **Interpretability:** Coefficients have direct probabilistic interpretation (log-odds in logistic regression, log-rate in Poisson)
- **Computational Efficiency:** Closed-form solutions for some special cases; Newton's method typically converges in few iterations
- **Foundation for Deep Learning:** Neural networks with single hidden layer using a GLM output layer are conceptually similar to GLMs with learned feature transformations

### Relationships to Previous Methods

| Method | GLM Connection |
|--------|---|
| **Linear Regression** | GLM with Gaussian family, identity link |
| **Logistic Regression** | GLM with Binomial family, logit link |
| **Batch Gradient Descent** | Optimization algorithm for GLM parameter estimation |
| **Gradient Ascent** | Used to maximize GLM log-likelihood |
| **Newton's Method** | Specialized for GLM via Fisher scoring (Newton-type method) |
| **Perceptron** | Heuristic predecessor; logistic regression is probabilistic successor |

## 11. Generative Learning Algorithms

### Overview

**Generative Learning** models the joint probability distribution $p(x, y)$ by first modeling the class-conditional distribution $p(x|y)$ (how features are distributed given the class) and the class prior $p(y)$, then using Bayes' rule to compute the posterior $p(y|x)$ for prediction.

$$p(y|x) = \frac{p(x|y)p(y)}{p(x)} = \frac{p(x|y)p(y)}{\sum_{y'} p(x|y')p(y')}$$

Generative models "generate" new data points by sampling from the learned distributions. They answer: **"What is the process that generates both features and labels?"**

### Key Characteristics

| Aspect | Generative Models |
|--------|-------------------|
| **Learns** | Joint distribution $p(x, y)$ or $p(x \mid y)$ and $p(y)$ |
| **Prediction** | Uses Bayes' rule: $p(y \mid x) = \frac{p(x \mid y)p(y)}{p(x)}$ |
| **Data Generation** | Can generate synthetic samples by sampling $y \sim p(y)$, then $x \sim p(x \mid y)$ |
| **Assumptions** | Requires distributional assumptions (e.g., Gaussian, Bernoulli) |
| **Sample Efficiency** | Often more data-efficient; leverages distributional structure |
| **Interpretability** | Interpretable class-conditional distributions |
| **Computational Cost** | Can be expensive if feature space is high-dimensional |

### Common Generative Models

**1. Gaussian Discriminant Analysis (GDA)**
- Assumes $p(x \mid y=k) \sim \mathcal{N}(\mu_k, \Sigma_k)$ (multivariate Gaussian per class)
- Simple decision boundary (linear or quadratic)
- Closed-form solution for parameter estimation

**2. Naive Bayes**
- Assumes conditional independence: $p(x \mid y) = \prod_{j=1}^{p} p(x_j \mid y)$
- Simple and efficient; works well for discrete and continuous features
- Robust despite independence assumption violation
- Used extensively in text classification, spam detection

**3. Hidden Markov Models (HMMs)**
- Models temporal sequences with latent states
- Assumption: current state depends only on previous state (Markov property)
- Used in speech recognition, natural language processing

**4. Topic Models (Latent Dirichlet Allocation)**
- Unsupervised generative model for text documents
- Treats documents as mixtures of latent topics
- Can be extended to supervised settings (e.g., sLDA)

### Mathematical Framework

For K-class classification with $m$ training samples:

**Parameter Estimation via Maximum Likelihood:**

1. **Estimate class priors:**
   $$\phi_k = p(y=k) = \frac{\text{# samples with class } k}{m}$$

2. **Estimate class-conditional distributions** (e.g., Gaussian for GDA):
   $$\mu_k = \frac{\sum_{i: y^{(i)}=k} x^{(i)}}{\text{# samples with class } k}$$
   $$\Sigma_k = \frac{\sum_{i: y^{(i)}=k} (x^{(i)} - \mu_k)(x^{(i)} - \mu_k)^T}{\text{# samples with class } k}$$

3. **Prediction via Bayes' rule:**
   $$\hat{y} = \arg\max_k p(y=k \mid x) = \arg\max_k \frac{p(x \mid y=k)p(y=k)}{p(x)} = \arg\max_k p(x \mid y=k)p(y=k)$$

### Strengths
- **Interpretable:** Class-conditional distributions reveal feature characteristics per class
- **Data efficient:** Captures distributional structure; good with limited data
- **Principled:** Based on probability theory
- **Generative:** Can synthesize new samples, analyze feature importance
- **Handles imbalance:** Naturally incorporates class priors $p(y)$
- **Efficient learning:** Often closed-form solutions exist

### Limitations
- **Distributional assumptions:** Real data may violate assumptions (e.g., non-Gaussian features)
- **High-dimensional curse:** Estimating covariance matrices becomes difficult in high dimensions ($p > n$)
- **Feature independence:** Naive Bayes assumes unrealistic conditional independence
- **Computational cost:** Computing $p(x \mid y)$ can be expensive for complex distributions
- **Inflexible boundaries:** May be restricted to simple decision surfaces

### When to Use Generative Models
- **Limited labeled data:** Generative models leverage distributional structure
- **Need to generate samples:** Domain-specific applications like image synthesis, data augmentation
- **Class imbalance:** Prior $p(y)$ naturally handles imbalanced datasets
- **Interpretability required:** Understanding per-class feature distributions is important
- **Unlabeled data available:** Can use semi-supervised learning to leverage unlabeled data

### Examples from This Course
- **Linear Regression:** Can be viewed as generative with Gaussian likelihood $p(y \mid x) = \mathcal{N}(\theta^T x, \sigma^2)$
- **Logistic Regression:** Binary generative interpretation with Bernoulli $p(y \mid x) = \sigma(\theta^T x)$
- **Softmax Regression:** K-class generative with categorical $p(y \mid x) = \text{softmax}(\Theta x)$

## 12. Discriminative Learning Algorithms

### Overview

**Discriminative Learning** directly models the conditional probability distribution $p(y|x)$ without explicitly modeling the feature distribution $p(x|y)$ or the joint $p(x,y)$. Instead of asking "How are features generated given the class?", discriminative models ask: **"What is the decision boundary that best separates the classes?"**

Discriminative models focus on learning the boundary between classes, making them often more efficient for prediction tasks.

### Key Characteristics

| Aspect | Discriminative Models |
|--------|----------------------|
| **Learns** | Conditional distribution $p(y \mid x)$ directly |
| **Prediction** | Directly computes $p(y \mid x)$ or outputs decision function |
| **Data Generation** | Cannot generate new feature samples; only classify existing data |
| **Assumptions** | Fewer assumptions; flexible function class |
| **Sample Efficiency** | Often requires more labeled data for same performance |
| **Interpretability** | Decision boundaries are explicit; less insight into feature structure |
| **Computational Cost** | Often computationally efficient during training and prediction |
| **Flexibility** | Can capture complex non-linear decision boundaries |

### Common Discriminative Models

**1. Logistic Regression**
- Linear discriminant with sigmoid link function
- Outputs $p(y=1 \mid x) = \sigma(\theta^T x)$
- Efficient, interpretable, probabilistic

**2. Support Vector Machines (SVMs)**
- Learns maximum-margin separator
- Outputs decision function $f(x) = \theta^T \phi(x) + b$ (non-probabilistic)
- Kernel methods enable non-linear boundaries
- Efficient for both small and large datasets

**3. Decision Trees & Random Forests**
- Hierarchical feature-based splitting
- Outputs class probability via leaf node frequencies
- Non-parametric, handles interactions automatically

**4. Neural Networks**
- Universal function approximators
- Learns compositional feature representations
- Can model arbitrarily complex decision boundaries
- Outputs $p(y \mid x)$ via softmax layer

**5. Perceptron & Linear Classifiers**
- Simple linear decision boundary
- Outputs binary classification via sign($\theta^T x$)
- Convergent on linearly separable data

### Mathematical Framework

For K-class classification:

**Direct Optimization of Posterior (Maximum Likelihood):**

$$\ell(\theta) = \sum_{i=1}^{m} \log p(y^{(i)} \mid x^{(i)}, \theta)$$

Maximize this directly using optimization algorithms:
- **Gradient Descent:** Scalable for large datasets
- **Newton's Method:** Faster convergence; requires Hessian computation
- **Stochastic Optimization:** Online learning; memory efficient

**Common Objective Functions:**

1. **Log-Loss (Cross-Entropy)** - for probabilistic outputs
   $$J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \sum_{k=1}^{K} \mathbb{1}(y^{(i)}=k) \log p(y=k \mid x^{(i)}, \theta)$$

2. **Hinge Loss** - for SVMs
   $$J(\theta) = \sum_{i=1}^{m} \max(0, 1 - y^{(i)}(\theta^T \phi(x^{(i)})))$$

3. **0-1 Loss** - for classification accuracy (non-convex, rarely used directly)
   $$J(\theta) = \sum_{i=1}^{m} \mathbb{1}(\text{sign}(\theta^T x^{(i)}) \neq y^{(i)})$$

**Decision Rule:**

$$\hat{y} = \arg\max_k p(y=k \mid x, \theta)$$

or for non-probabilistic outputs:

$$\hat{y} = \text{sign}(f(x, \theta))$$

### Strengths
- **Efficient training:** Fewer parameters; faster convergence
- **Better for prediction:** Optimized directly for $p(y \mid x)$; often superior predictive performance
- **Flexible:** No distributional assumptions; can model complex boundaries
- **Scalable:** Gradient-based optimization works well with large datasets
- **Non-parametric variants:** Some discriminative models (e.g., SVMs, trees) are non-parametric
- **Optimal for supervised learning:** Directly minimizes classification error

### Limitations
- **Limited interpretability:** Less insight into feature distributions or class structure
- **Cannot generate data:** No way to synthesize new samples
- **Requires labeled data:** Cannot leverage unlabeled data as easily as generative models
- **Class imbalance:** Requires explicit handling (weighted loss, resampling)
- **Data efficiency:** Often needs more labeled data than generative models
- **Overfitting risk:** Flexible function classes can overfit without regularization

### When to Use Discriminative Models
- **Abundant labeled data:** Can afford to collect sufficient training examples
- **High accuracy required:** Discriminative models often achieve better test performance
- **Large feature space:** Efficient with high-dimensional data ($p >> m$)
- **Limited resources:** Faster inference and training
- **Complex boundaries:** Need to capture non-linear decision surfaces
- **Real-time constraints:** Efficient prediction is critical
- **Well-defined task:** Clear distinction between classes

### Examples from This Course
- **Logistic Regression:** Binary and multi-class classification
- **Softmax Regression:** Directly models $p(y \mid x)$ for K classes
- **Generalized Linear Models:** Framework for discriminative modeling
- **All optimization methods:** Gradient descent, Newton's method - designed for discriminative learning

## Generative vs. Discriminative: Comparison

| Criterion | Generative | Discriminative |
|-----------|-----------|----------------|
| **Models** | $p(x \mid y)$, $p(y)$, and $p(x, y)$ | $p(y \mid x)$ directly |
| **Answer** | "How is data generated?" | "What separates the classes?" |
| **Training Objective** | Maximize likelihood of joint distribution | Maximize likelihood of posterior |
| **Decision Boundary** | Emerges from probabilistic assumptions | Directly optimized |
| **Sample Efficiency** | Often more efficient (fewer examples needed) | Generally less efficient |
| **Data Generation** | Can generate synthetic samples | Cannot generate feature data |
| **Labeled Data Requirement** | Can leverage unlabeled data (semi-supervised) | Requires labeled data |
| **Interpretability** | Interpretable class distributions | Interpretable boundaries |
| **Computational Cost** | Can be expensive (high-dimensional covariance) | Usually computationally efficient |
| **Flexibility** | Constrained by distributional assumptions | Flexible; few assumptions |
| **Typical Performance** | Good generalization with limited data | Often superior with sufficient data |
| **Examples** | Naive Bayes, GDA, HMM, Topic Models | Logistic Regression, SVM, Decision Trees, Neural Networks |

### When to Choose

**Choose Generative if:**
- Data is limited or imbalanced
- You need to generate synthetic data or perform analysis
- Interpretability of class structure is important
- Semi-supervised learning is needed
- Computational resources are abundant

**Choose Discriminative if:**
- Abundant labeled data available
- Maximum prediction accuracy is priority
- Computational efficiency is critical
- High-dimensional feature spaces
- Non-linear boundaries are needed
- Real-time inference required

### Modern Perspective

In modern machine learning, **discriminative models dominate** because:
1. **Big Data Era:** Abundant labeled data makes discriminative learning practical
2. **Deep Learning:** Neural networks (discriminative) have revolutionized many domains
3. **Task Focus:** Most applications prioritize prediction accuracy over data generation
4. **Flexibility:** No need to specify distributional assumptions

However, **generative models remain valuable** for:
- Semi-supervised learning with limited labels
- Domain adaptation and transfer learning
- Data augmentation and synthesis
- Unsupervised and self-supervised learning
- Interpretability and explainability

The **hybrid approach** combines both:
- Use discriminative models for primary prediction task
- Use generative models to understand data structure, detect anomalies, or augment training data
