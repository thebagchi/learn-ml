# CNN Architectures — LeNet, AlexNet, VGG

> Based on Stanford CS230 — C4M2, Andrew Ng / DeepLearning.AI

Three landmark architectures that defined how CNNs are designed: LeNet-5 (1998), AlexNet (2012), and VGGNet (2014). Each introduced patterns that remain standard today.

## Learning Objectives

1. Trace the shape of activations through LeNet-5, AlexNet, and VGG-16
2. Identify the design patterns each architecture introduced
3. Count parameters and explain where most parameters live
4. Understand why deeper + smaller filters outperform shallow + larger filters

## Architecture Comparison

| Architecture | Year | Input | Depth | Params | Top-5 Error | Key innovation |
|---|---|---|---|---|---|---|
| **LeNet-5** | 1998 | 32×32×1 | 5 layers | ~60 K | — | Convolution for vision |
| **AlexNet** | 2012 | 224×224×3 | 8 layers | ~60 M | 15.3 % | ReLU, dropout, GPU training |
| **VGG-16** | 2014 | 224×224×3 | 16 layers | ~138 M | 7.3 % | Deep + small (3×3) filters |

## VGG-16 Key Pattern — Two 3×3 Filters = One 5×5 Filter

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'figure.facecolor': '#f5f5f7', 'axes.facecolor': '#ffffff',
    'axes.edgecolor': '#c8ccd4', 'axes.labelcolor': '#1a1d27',
    'axes.titlecolor': '#1a1d27', 'xtick.color': '#2a2e3a',
    'ytick.color': '#2a2e3a', 'grid.color': '#e0e3ea',
    'grid.linestyle': '--', 'grid.alpha': 0.5,
    'text.color': '#1a1d27', 'font.family': 'DejaVu Sans',
    'axes.titlesize': 12, 'axes.labelsize': 11,
    'legend.facecolor': '#ffffff', 'legend.edgecolor': '#c8ccd4',
    'figure.dpi': 110,
})
P = ['#5b9bd5', '#e05c5c', '#f4b942', '#7ecba1', '#56b6c2', '#c678dd']

# ---- Shape trace through LeNet-5 and VGG-16 (simplified) ----
def conv_out(n, f, p, s): return (n + 2*p - f) // s + 1
def pool_out(n, f, s):    return (n - f) // s + 1

def trace_lenet():
    rows = []
    n, nc = 32, 1
    rows.append(('Input',        n,  n,  nc,  0))
    f=5; nc2=6;  n=conv_out(n,f,0,1);  params=(f*f*nc+1)*nc2; nc=nc2
    rows.append(('Conv1 (f=5,C=6)',   n, n, nc, params))
    n=pool_out(n,2,2)
    rows.append(('AvgPool (f=2,s=2)', n, n, nc, 0))
    f=5; nc2=16; n=conv_out(n,f,0,1);  params=(f*f*nc+1)*nc2; nc=nc2
    rows.append(('Conv2 (f=5,C=16)',  n, n, nc, params))
    n=pool_out(n,2,2)
    rows.append(('AvgPool (f=2,s=2)', n, n, nc, 0))
    flat = n*n*nc
    rows.append(('Flatten',           1, flat, 1, 0))
    rows.append(('FC-120', 1,120,1,(flat+1)*120))
    rows.append(('FC-84',  1,84, 1,(120+1)*84))
    rows.append(('Output-10', 1,10,1,(84+1)*10))
    return rows

def trace_vgg_block():
    """VGG-16 first two blocks (224→56)."""
    rows = []
    n, nc = 224, 3
    rows.append(('Input',                 n,  n,  nc, 0))
    for rep in range(2):
        nc2=64; params=(3*3*nc+1)*nc2; n=conv_out(n,3,1,1); nc=nc2
        rows.append((f'Conv (3×3,C=64) #{rep+1}', n, n, nc, params))
    n=pool_out(n,2,2)
    rows.append(('MaxPool (f=2,s=2)',  n,n,nc, 0))
    for rep in range(2):
        nc2=128; params=(3*3*nc+1)*nc2; n=conv_out(n,3,1,1); nc=nc2
        rows.append((f'Conv (3×3,C=128) #{rep+1}', n, n, nc, params))
    n=pool_out(n,2,2)
    rows.append(('MaxPool (f=2,s=2)',  n,n,nc, 0))
    return rows

print("=" * 65)
print("LeNet-5 Shape Trace")
print(f"{'Layer':30s} {'H':>4} {'W':>4} {'C':>5} {'Params':>10}")
print("-" * 65)
total = 0
for name, h, w, c, p in trace_lenet():
    print(f"{name:30s} {h:>4} {w:>4} {c:>5} {p:>10,}")
    total += p
print(f"{'Total parameters':55s} {total:>10,}")

print()
print("=" * 65)
print("VGG-16 First Two Blocks (224 → 56)")
print(f"{'Layer':35s} {'H':>4} {'W':>4} {'C':>5} {'Params':>10}")
print("-" * 65)
total = 0
for name, h, w, c, p in trace_vgg_block():
    print(f"{name:35s} {h:>4} {w:>4} {c:>5} {p:>10,}")
    total += p
print(f"{'Total params (first 2 blocks)':55s} {total:>10,}")
```

```python
# ---- Parameter distribution visualisation ----
architectures = {
    'LeNet-5 (62K)': {
        'Conv layers': 3906,
        'FC layers':   58290,
    },
    'AlexNet (60M)': {
        'Conv layers':  3747200,
        'FC layers':   58631144,
    },
    'VGG-16 (138M)': {
        'Conv layers': 14714688,
        'FC layers':  123642856,
    }
}

fig, axes = plt.subplots(1, 3, figsize=(14, 5))
fig.suptitle('Parameter Distribution — Conv vs FC Layers', fontsize=13, fontweight='bold')
colors_pie = [P[0], P[1]]

for ax, (arch_name, parts) in zip(axes, architectures.items()):
    labels = list(parts.keys())
    sizes  = list(parts.values())
    total  = sum(sizes)
    wedges, texts, autotexts = ax.pie(
        sizes, labels=labels, autopct='%1.1f%%',
        colors=colors_pie, startangle=90,
        wedgeprops={'edgecolor': 'white', 'linewidth': 1.5},
        textprops={'fontsize': 10}
    )
    for at in autotexts: at.set_fontsize(10)
    ax.set_title(f'{arch_name}\n{total/1e6:.1f}M parameters', fontsize=10)

plt.tight_layout()
plt.show()

# ---- Error rate progression ----
years = [1998, 2012, 2014, 2015, 2016]
top5  = [None, 15.3, 7.3, 3.57, 2.99]
human = 5.1

fig, ax = plt.subplots(figsize=(9, 4.5))
ax.set_title('ImageNet Top-5 Error — Architecture Progress', fontsize=12)
points = [(y, e) for y, e in zip(years, top5) if e is not None]
ax.plot([p[0] for p in points], [p[1] for p in points],
        color=P[0], lw=2.5, marker='o', ms=8, zorder=3)
labels_map = {2012: 'AlexNet', 2014: 'VGGNet', 2015: 'ResNet', 2016: 'Ensemble'}
for y, e in points:
    ax.annotate(labels_map.get(y, ''), (y, e), textcoords='offset points',
                xytext=(0, 10), ha='center', fontsize=9)
ax.axhline(human, color=P[1], lw=1.8, ls='--', label=f'Human-level error ({human}%)')
ax.set_xlabel('Year')
ax.set_ylabel('Top-5 Error (%)')
ax.legend(fontsize=9)
ax.grid(True)
plt.tight_layout()
plt.show()
```
