# ARIMA — Time Series Forecasting
**Metrics:** MAE, RMSE, MAPE

```python
from utils.time_series import ARIMA

import numpy as np
import pandas as pd
```

```python
# Simulated stationary time series
np.random.seed(42)
n = 200
t = np.arange(n)
y = 10 + 0.05 * t + np.random.normal(0, 1, n)
series = pd.Series(y, index=pd.date_range("2020-01-01", periods=n, freq="D"))

train_size = int(len(series) * 0.8)
train, test = series[:train_size], series[train_size:]

display(pd.DataFrame({"value": series}).head())
```

```python
model    = ARIMA(order=(1, 1, 1)).train(train)
forecast = model.predict(steps=len(test))
report   = model.evaluate(test.values, forecast.values)
for k, v in report.items():
    print(f"{k}: {v:.4f}")
```
