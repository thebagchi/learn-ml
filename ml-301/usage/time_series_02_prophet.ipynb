# Prophet — Time Series Forecasting
**Metrics:** MAE, RMSE, MAPE

```python
from utils.time_series import ProphetModel

from prophet import Prophet
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
np.random.seed(42)
n = 365
dates = pd.date_range("2022-01-01", periods=n)
y = 50 + 10 * np.sin(2 * np.pi * np.arange(n) / 7) + np.random.normal(0, 2, n)
df = pd.DataFrame({"ds": dates, "y": y})

split = int(n * 0.8)
train_df, test_df = df[:split], df[split:]

display(df.head())
```

```python
model    = ProphetModel().train(train_df)
forecast = model.predict(periods=len(test_df))
report   = model.evaluate(test_df["y"].values, forecast, len(test_df))
for k, v in report.items():
    print(f"{k}: {v:.4f}")
```
