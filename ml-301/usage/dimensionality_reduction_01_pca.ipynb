# Principal Component Analysis (PCA)
**Metrics:** Explained Variance Ratio, Reconstruction Error

```python
from utils.dim_reduction import PCA

from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
digits = load_digits()
X, y = digits.data, digits.target

df = pd.DataFrame(X, columns=[f"pixel_{i}" for i in range(X.shape[1])])
df["label"] = y
display(df.head())

scaler = StandardScaler()
X = scaler.fit_transform(X)
print(f"Shape before PCA: {X.shape}")

df_scaled = pd.DataFrame(X, columns=[f"pixel_{i}" for i in range(X.shape[1])])
df_scaled["label"] = y
display(df_scaled.head())
```

```python
model     = PCA(n_components=2).train(X)
X_reduced = model.predict(X)
report    = model.evaluate(X)
print(f"Shape after PCA: {X_reduced.shape}")
for k, v in report.items():
    print(f"{k}: {v}")

df_reduced = pd.DataFrame(X_reduced, columns=[f"pc_{i+1}" for i in range(X_reduced.shape[1])])
df_reduced["label"] = y
display(df_reduced.head())
```
