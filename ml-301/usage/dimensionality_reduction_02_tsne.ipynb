# t-SNE — Dimensionality Reduction
**Note:** t-SNE is for visualization; no inverse transform.
**Metrics:** KL Divergence

```python
from utils.dim_reduction import TSNE

from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
digits = load_digits()
X, y = digits.data, digits.target

display(pd.DataFrame(X, columns=[f"pixel_{i}" for i in range(X.shape[1])]).assign(label=y).head())

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
model      = TSNE(n_components=2, perplexity=30).train(X)
X_embedded = model.predict(X)
report     = model.evaluate(X_embedded)
print(f"Embedded shape: {X_embedded.shape}")
for k, v in report.items():
    print(f"{k}: {v}")

display(pd.DataFrame(X_embedded, columns=[f"dim_{i+1}" for i in range(X_embedded.shape[1])]).assign(label=y).head())
```
