# K-Means Clustering
**Metrics:** Silhouette Score, Davies-Bouldin Index, Inertia

```python
from utils.clustering import KMeans

from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.8, random_state=42)
scaler = StandardScaler()
X = scaler.fit_transform(X)

display(pd.DataFrame(X, columns=[f"feat_{i}" for i in range(X.shape[1])]).head())
```

```python
model  = KMeans(n_clusters=4).train(X)
labels = model.predict(X)
report = model.evaluate(X)
for k, v in report.items():
    print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")
```
