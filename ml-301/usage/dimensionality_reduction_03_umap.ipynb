# UMAP — Dimensionality Reduction
**Metrics:** Trustworthiness (sklearn proxy)

```python
from utils.dim_reduction import UMAP

import umap
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import trustworthiness
import numpy as np
import pandas as pd
```

```python
digits = load_digits()
X, y = digits.data, digits.target

display(pd.DataFrame(X, columns=[f"pixel_{i}" for i in range(X.shape[1])]).assign(label=y).head())

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
model      = UMAP(n_components=2).train(X)
X_embedded = model.predict(X)
report     = model.evaluate(X, X_embedded)
print(f"Embedded shape: {X_embedded.shape}")
for k, v in report.items():
    print(f"{k}: {v}")

display(pd.DataFrame(X_embedded, columns=[f"dim_{i+1}" for i in range(X_embedded.shape[1])]).assign(label=y).head())
```
