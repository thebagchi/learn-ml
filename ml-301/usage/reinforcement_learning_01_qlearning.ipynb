# Q-Learning — Reinforcement Learning
**Environment:** FrozenLake-v1
**Metrics:** Cumulative Reward, Win Rate

```python
from utils.rl import QLearning

import numpy as np
import gymnasium as gym
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False)
n_states  = env.observation_space.n
n_actions = env.action_space.n
print(f"States: {n_states}, Actions: {n_actions}")
```

```python
agent  = QLearning().train(env, episodes=5_000)
report = agent.evaluate(env, episodes=500)
for k, v in report.items():
    print(f"{k}: {v:.4f}")
print(f"\nLast 100 train episodes avg reward: {np.mean(agent.train_rewards[-100:]):.4f}")
```
