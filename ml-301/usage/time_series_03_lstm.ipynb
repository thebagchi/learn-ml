# LSTM — Time Series Forecasting
**Metrics:** MAE, RMSE, MAPE

```python
from utils.time_series import LSTMForecaster

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
```

```python
np.random.seed(42)
n = 400
t = np.linspace(0, 4 * np.pi, n)
y = np.sin(t) + 0.1 * np.random.randn(n)

display(pd.DataFrame({"t": t, "y": y}).head())

scaler = MinMaxScaler()
y_scaled = scaler.fit_transform(y.reshape(-1, 1)).flatten()

SEQ_LEN = 20
def make_sequences(data, seq_len):
    X, Y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i: i + seq_len])
        Y.append(data[i + seq_len])
    return np.array(X), np.array(Y)

X_all, y_all = make_sequences(y_scaled, SEQ_LEN)
split = int(len(X_all) * 0.8)
X_train, X_test = X_all[:split], X_all[split:]
y_train, y_test = y_all[:split], y_all[split:]

X_train_t = torch.tensor(X_train, dtype=torch.float32).unsqueeze(-1)
X_test_t  = torch.tensor(X_test,  dtype=torch.float32).unsqueeze(-1)
y_train_t = torch.tensor(y_train, dtype=torch.float32)
y_test_t  = torch.tensor(y_test,  dtype=torch.float32)
```

```python
model  = LSTMForecaster(epochs=20).train(X_train_t, y_train_t)
y_pred = model.predict(X_test_t)
report = model.evaluate(y_test, y_pred, scaler)
for k, v in report.items():
    print(f"{k}: {v:.4f}")
```
