# Recurrent Neural Networks (RNN)

> Based on Stanford CS230 — C5M1, Andrew Ng / DeepLearning.AI

Feedforward networks assume independent inputs. Sequences — speech, text, DNA, video — have **temporal dependencies**. An RNN processes one token at a time, threading a **hidden state** $a^{\langle t \rangle}$ across time steps.

## Learning Objectives

1. Explain why feedforward nets are insufficient for sequential data
2. Write the RNN forward pass equations
3. Identify the types of RNN architectures (many-to-many, many-to-one, one-to-many)
4. Understand backpropagation through time (BPTT) and vanishing gradients

## RNN Forward Pass

$$a^{\langle t \rangle} = \tanh\!\bigl(W_{aa}\,a^{\langle t-1 \rangle} + W_{ax}\,x^{\langle t \rangle} + b_a\bigr)$$
$$\hat{y}^{\langle t \rangle} = \mathrm{softmax}\!\bigl(W_{ya}\,a^{\langle t \rangle} + b_y\bigr)$$

- $x^{\langle t \rangle} \in \mathbb{R}^{n_x}$ — input at step $t$
- $a^{\langle t \rangle} \in \mathbb{R}^{n_a}$ — hidden state (memory)
- Parameters shared across all time steps

## Backpropagation Through Time (BPTT)

Gradients flow *backward through time* across all steps. For long sequences the chain rule multiplies many Jacobians, causing **vanishing** (or exploding) gradients — motivating LSTM/GRU.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    'figure.facecolor': '#f5f5f7', 'axes.facecolor': '#ffffff',
    'axes.edgecolor': '#c8ccd4', 'axes.labelcolor': '#1a1d27',
    'axes.titlecolor': '#1a1d27', 'xtick.color': '#2a2e3a',
    'ytick.color': '#2a2e3a', 'grid.color': '#e0e3ea',
    'grid.linestyle': '--', 'grid.alpha': 0.5,
    'text.color': '#1a1d27', 'font.family': 'DejaVu Sans',
    'axes.titlesize': 12, 'axes.labelsize': 11,
    'legend.facecolor': '#ffffff', 'legend.edgecolor': '#c8ccd4',
    'figure.dpi': 110,
})
P = ['#5b9bd5', '#e05c5c', '#f4b942', '#7ecba1', '#56b6c2', '#c678dd']

def softmax(z):
    e = np.exp(z - z.max(axis=0, keepdims=True))
    return e / e.sum(axis=0, keepdims=True)

class VanillaRNN:
    """Minimal RNN: forward pass only, for illustration."""
    def __init__(self, n_x, n_a, n_y, seed=0):
        np.random.seed(seed)
        s = 0.01
        self.Wax = np.random.randn(n_a, n_x) * s
        self.Waa = np.random.randn(n_a, n_a) * s
        self.Wya = np.random.randn(n_y, n_a) * s
        self.ba  = np.zeros((n_a, 1))
        self.by  = np.zeros((n_y, 1))

    def step(self, x_t, a_prev):
        a_t = np.tanh(self.Waa @ a_prev + self.Wax @ x_t + self.ba)
        y_t = softmax(self.Wya @ a_t + self.by)
        return a_t, y_t

    def forward(self, x_seq):
        """x_seq: (T, n_x, 1)  →  hidden states, outputs."""
        T = len(x_seq)
        n_a = self.Waa.shape[0]
        a   = np.zeros((n_a, 1))
        states, preds = [], []
        for t in range(T):
            a, y = self.step(x_seq[t], a)
            states.append(a.copy())
            preds.append(y.copy())
        return np.array(states), np.array(preds)

# ---- Simulate forward pass on a short sequence ----
np.random.seed(42)
T, n_x, n_a, n_y = 10, 4, 8, 3
rnn = VanillaRNN(n_x, n_a, n_y)
x_seq = [np.random.randn(n_x, 1) for _ in range(T)]
states, preds = rnn.forward(x_seq)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Vanilla RNN — Forward Pass', fontsize=13, fontweight='bold')

# Hidden states heatmap
im = axes[0].imshow(states[:, :, 0].T, aspect='auto', cmap='RdYlBu', vmin=-1, vmax=1)
axes[0].set_xlabel('Time step t')
axes[0].set_ylabel('Hidden unit index')
axes[0].set_title('Hidden state activations $a^{\\langle t \\rangle}$')
axes[0].set_xticks(range(T))
plt.colorbar(im, ax=axes[0], label='tanh activation')

# Output probabilities over time
for c in range(n_y):
    axes[1].plot(range(T), preds[:, c, 0], marker='o', color=P[c], lw=2, label=f'class {c}')
axes[1].set_xlabel('Time step t')
axes[1].set_ylabel('Probability')
axes[1].set_title('Output $\\hat{y}^{\\langle t \\rangle}$ (softmax)')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# ---- Vanishing gradient illustration ----
fig, ax = plt.subplots(figsize=(10, 4))
steps = np.arange(1, 31)
for w, label in [(0.5, '|W|=0.5 (vanishing)'), (1.0, '|W|=1.0 (stable)'), (1.5, '|W|=1.5 (exploding)')]:
    ax.semilogy(steps, w**steps, lw=2, label=label)
ax.set_xlabel('BPTT steps back')
ax.set_ylabel('Gradient magnitude (log)')
ax.set_title('Vanishing vs Exploding Gradients in BPTT')
ax.axhline(1e-3, color='#bbb', ls='--', lw=1)
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```
