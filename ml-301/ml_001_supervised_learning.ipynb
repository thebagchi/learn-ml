# Supervised Learning: Optimization Methods

This notebook introduces key optimization methods used in supervised learning. We'll cover the mathematical concepts essential for understanding each approach without going into detailed derivations.

## Mathematical Foundations

### Terminology and Notation

**Nabla ($\nabla$):** The symbol $\nabla$ (pronounced "nabla" or "del") is the gradient operator used to denote the gradient of a function.

**Gradient Vector:** The **gradient vector** (also called the **gradient**) is the vector of all first-order partial derivatives of a function with respect to its variables. It points in the direction of steepest ascent.

### Gradient and Jacobian Matrix

The **gradient is a special case of the Jacobian matrix**. Understanding their relationship is crucial for optimization:

**Jacobian Matrix:** For a vector-valued function $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^m$, the Jacobian is the matrix of all first-order partial derivatives:

$$J = \begin{bmatrix} \frac{\partial f_1}{\partial x_1} & \frac{\partial f_1}{\partial x_2} & \cdots & \frac{\partial f_1}{\partial x_n} \\ \frac{\partial f_2}{\partial x_1} & \frac{\partial f_2}{\partial x_2} & \cdots & \frac{\partial f_2}{\partial x_n} \\ \vdots & \vdots & \ddots & \vdots \\ \frac{\partial f_m}{\partial x_1} & \frac{\partial f_m}{\partial x_2} & \cdots & \frac{\partial f_m}{\partial x_n} \end{bmatrix}$$

**Gradient Vector:** The gradient vector (denoted using nabla $\nabla$) is the Jacobian of a **scalar-valued function** $J: \mathbb{R}^n \to \mathbb{R}$ (when $m = 1$):

$$\nabla J(\theta) = \begin{bmatrix} \frac{\partial J}{\partial \theta_1} \\ \frac{\partial J}{\partial \theta_2} \\ \vdots \\ \frac{\partial J}{\partial \theta_n} \end{bmatrix}$$

In optimization, we work with scalar cost functions, so **the gradient vector is the Jacobian specialized for single-output functions**. It is a column vector (or row vector by convention) containing all first-order partial derivatives, and it points in the direction of steepest increase in the function value.


## 1. Batch Gradient Descent

### Key Mathematical Concepts

**Gradient Vector ($\nabla J(\theta)$):** The vector of partial derivatives of the cost function with respect to each parameter. The nabla symbol ($\nabla$) denotes the gradient operator.
$$\nabla J(\theta) = \begin{bmatrix} \frac{\partial J}{\partial \theta_0} \\ \frac{\partial J}{\partial \theta_1} \\ \vdots \\ \frac{\partial J}{\partial \theta_n} \end{bmatrix}$$

**Update Rule:** Parameters are updated based on the gradient vector computed over the entire dataset:
$$\theta := \theta - \alpha \nabla J(\theta)$$

Where $\alpha$ is the learning rate.

### Essential Concepts
- **Cost Function:** $J(\theta)$ - typically Mean Squared Error (MSE) or Cross-Entropy
- **Learning Rate ($\alpha$):** Controls step size; too large causes divergence, too small causes slow convergence
- **Convergence:** Process repeats until gradient vector approaches zero or max iterations reached
- **Dataset Size:** Uses all $m$ training examples per iteration

### When to Use
- Small to medium-sized datasets
- When you want guaranteed convergence (with proper learning rate)
- Offline learning scenarios


## 2. Stochastic Gradient Descent (SGD)

### Key Mathematical Concepts

**Per-Sample Gradient Vector:** Updates based on the gradient vector computed from a single training example:
$$\theta := \theta - \alpha \nabla J(\theta; x^{(i)}, y^{(i)})$$

Where $(x^{(i)}, y^{(i)})$ is a single training example, and $\nabla J(\theta; x^{(i)}, y^{(i)})$ is the gradient vector for that example.

### Essential Concepts
- **Noisy Updates:** Gradient vector estimates are noisy (based on single samples)
- **Oscillation:** Parameter updates oscillate around the minimum instead of smoothly converging
- **Variance:** High variance in gradient estimates requires careful learning rate scheduling
- **Update Frequency:** More frequent updates per epoch (one update per sample)

### Key Characteristics
- **Epoch:** One complete pass through the entire dataset (containing $m$ updates)
- **Noise Benefits:** Noise can help escape local minima (beneficial in non-convex optimization)
- **Learning Rate Schedule:** Often requires decreasing learning rate over time: $\alpha_t = \frac{\alpha_0}{1 + t}$

### When to Use
- Very large datasets
- Online learning (streaming data)
- Non-convex optimization problems
- Memory-constrained environments


## 3. Mini-Batch Gradient Descent

### Key Mathematical Concepts

**Batch Gradient Vector:** Updates based on the gradient vector computed over a subset (batch) of training data:
$$\theta := \theta - \alpha \frac{1}{b} \sum_{i=1}^{b} \nabla J(\theta; x^{(i)}, y^{(i)})$$

Where $b$ is the batch size (typically 32, 64, 128, etc.), and $\nabla J$ represents the gradient vector.

### Essential Concepts
- **Batch Size ($b$):** Critical hyperparameter balancing stability and speed
- **Gradient Vector Estimation:** Average gradient vector over $b$ samples reduces variance compared to SGD
- **Computational Efficiency:** Leverages vectorized operations for faster computation than pure SGD
- **Variance Reduction:** Lower variance than SGD, higher variance than batch GD

### Mini-Batch Properties
- **Stability:** More stable convergence than SGD due to gradient vector averaging
- **Updates per Epoch:** $\frac{m}{b}$ updates per epoch (between SGD and Batch GD)
- **Hardware Utilization:** Optimal balance for GPU/TPU processing
- **Number of Iterations:** Epoch = $\frac{m}{b}$ iterations

### When to Use
- Large datasets (most practical scenarios)
- Deep learning with modern hardware
- When you need a balance between stability and efficiency


## 4. Normal Equation

### Key Mathematical Concepts

**Closed-Form Solution:** Directly compute optimal parameters without iteration:
$$\theta = (X^T X)^{-1} X^T y$$

Where:
- $X$ is the design matrix (size: $m \times (n+1)$, with column of 1s for intercept)
- $y$ is the target vector (size: $m \times 1$)
- $\theta$ is the parameter vector

### Essential Concepts
- **Derivation Basis:** Derived by setting gradient vector to zero: $\nabla J(\theta) = 0$
- **Matrix Inversion:** Requires $(X^T X)$ to be invertible (full rank)
- **Computational Complexity:** $O(n^3)$ due to matrix inversion (where $n$ is number of features)
- **No Hyperparameters:** No learning rate or iteration count needed

### Essential Conditions
- **Invertibility:** $(X^T X)$ must be invertible (full rank)
- **Applicability:** Primarily for linear regression problems
- **Feature Scaling:** Not required (unlike gradient descent)

### Handling Non-Invertible Matrices: Pseudoinverse

**When $(X^T X)$ is singular or nearly singular**, the **Moore-Penrose pseudoinverse** $X^+$ provides a practical solution:

$$\theta = X^+ y$$

Where $X^+ = (X^T X)^{-1} X^T$ is replaced by the pseudoinverse.

**Key Benefits:**
- **Always computable:** Works even when $(X^T X)$ is singular or ill-conditioned
- **Least-squares solution:** Minimizes $\|X\theta - y\|^2$ even when exact solution doesn't exist
- **Numerical stability:** More robust to rounding errors and near-singular matrices
- **Unique solution:** Provides minimum-norm solution when multiple solutions exist

**When Singularity Occurs:**
- Redundant features (linearly dependent columns in $X$)
- More features than training examples ($n > m$)
- Features with perfect multicollinearity

### When to Use
- Small to medium datasets ($n < 10,000$ features)
- When you need exact solution (for linear regression)
- When interpretability is important
- Offline learning scenarios

### When NOT to Use
- Very large feature sets (matrix inversion becomes prohibitively expensive)
- Streaming/online learning
- Non-linear models
- When numerical stability is critical and pseudoinverse cannot be computed reliably


## 5. Locally Weighted Linear Regression (LWLR)

### Key Mathematical Concepts

**Weighted Cost Function:** Fit a separate local model for every query point $x$ by assigning a weight $w^{(i)}$ to each training example:
$$J(\theta) = \sum_{i=1}^{m} w^{(i)}\bigl(y^{(i)} - \theta^T x^{(i)}\bigr)^2$$

**Gaussian Kernel (weight function):**
$$w^{(i)} = \exp\!\left(-\frac{\bigl(x^{(i)} - x\bigr)^2}{2\tau^2}\right)$$

Where $\tau$ is the **bandwidth** — controls how quickly weights fall off with distance.

**Closed-Form Solution:** Same as the Normal Equation but with weight matrix $W = \text{diag}(w^{(1)}, \ldots, w^{(m)})$:
$$\theta = \left(X^T W X\right)^{-1} X^T W y$$

### Essential Concepts
- **Weighted Least Squares:** Extension of OLS where each residual is scaled by a sample weight $w^{(i)}$
- **Diagonal Weight Matrix ($W$):** $m \times m$ matrix with kernel weights on the diagonal; encodes per-sample importance
- **Gaussian / RBF Kernel:** Exponential function mapping distance to similarity; produces values in $(0, 1]$
- **Bandwidth ($\tau$):** Controls the effective neighbourhood size — the key bias-variance trade-off parameter
- **Matrix Inversion / Solve:** $(X^TWX)^{-1}$ must be computed per query; requires the matrix to be full rank
- **Bias-Variance Trade-off:** Small $\tau$ → low bias, high variance (overfit); large $\tau$ → high bias, low variance (underfit)
- **Non-parametric Learning:** No fixed parameter vector; model complexity grows with the training set
- **Computational Cost:** $O(mn^2 + n^3)$ per prediction — matrix solve dominates for large feature count

### When to Use
- Non-linear relationships where a single linear model is insufficient
- Small to medium datasets

### When NOT to Use
- Large datasets (per-query cost scales with $m$)
- High-dimensional inputs

## 5. Variants of Linear Regression

Beyond standard Linear Regression, several important variants address different problem characteristics and constraints:

| Variant | Objective Function | Key Feature | Use Case |
|---------|-------------------|-------------|----------|
| **Ridge Regression (L2)** | $J(\theta) = \frac{1}{2m}\|X\theta - y\|^2 + \lambda\|\theta\|^2$ | Adds L2 penalty to shrink coefficients | Handles multicollinearity; prevents overfitting |
| **Lasso Regression (L1)** | $J(\theta) = \frac{1}{2m}\|X\theta - y\|^2 + \lambda\|\theta\|_1$ | Adds L1 penalty; performs feature selection | Automatic feature selection; sparse solutions |
| **Elastic Net** | $J(\theta) = \frac{1}{2m}\|X\theta - y\|^2 + \lambda_1\|\theta\|_1 + \lambda_2\|\theta\|^2$ | Combines L1 and L2 penalties | Balance between Ridge and Lasso benefits |
| **Polynomial Regression** | $J(\theta) = \frac{1}{2m}\|X'\theta - y\|^2$ (with polynomial features) | Fits polynomial curves to data | Non-linear relationships; higher-degree patterns |
| **Weighted Least Squares** | $J(\theta) = \frac{1}{2m}(X\theta - y)^T W(X\theta - y)$ | Different weights for different samples | Heteroscedasticity; some samples more reliable |
| **Robust Regression** | $J(\theta) = \sum_{i=1}^{m} \rho(y^{(i)} - \theta^T x^{(i)})$ | Uses robust loss functions (Huber, absolute value) | Outlier-resistant fitting |
| **Quantile Regression** | Minimizes weighted sum of absolute errors | Fits specific quantiles (median, quartiles) | Capturing distribution properties beyond mean |
| **Logistic Regression** | $J(\theta) = -\frac{1}{m}\sum_{i=1}^{m}[y^{(i)}\log(\hat{y}^{(i)}) + (1-y^{(i)})\log(1-\hat{y}^{(i)})]$ | Sigmoid activation for binary classification | Binary/multi-class classification problems |

### Key Regularization Concepts

**Regularization Parameter ($\lambda$):** Controls the trade-off between fitting the training data and model complexity:
- $\lambda = 0$: Standard Linear Regression (no penalty)
- $\lambda \to \infty$: Strong penalty, coefficients shrink toward zero

**L1 Penalty:** $\|\theta\|_1 = \sum_{j=1}^{n}|\theta_j|$
- Promotes sparsity (some coefficients exactly zero)
- Useful for automatic feature selection

**L2 Penalty:** $\|\theta\|^2 = \sum_{j=1}^{n}\theta_j^2$
- Shrinks all coefficients proportionally
- Helps with multicollinearity

## Parametric vs Non-Parametric Learning

### Parametric Learning

A **parametric** model summarises the data into a **fixed, finite set of parameters** $\theta$. Once trained, the training data can be discarded — all information is captured in $\theta$.

- Model complexity is **fixed** regardless of dataset size
- Fast at prediction time (just evaluate $\theta^T x$)
- Strong assumptions about the functional form of the data

| Algorithm | Parameters |
|---|---|
| Linear Regression | Weights $\theta \in \mathbb{R}^{n+1}$ |
| Logistic Regression | Weights $\theta \in \mathbb{R}^{n+1}$ |
| Linear SVM | Weight vector $\mathbf{w}$, bias $b$ |
| Naive Bayes | Class priors + feature likelihoods |
| Neural Networks | All layer weights and biases |

---

### Non-Parametric Learning

A **non-parametric** model makes **no fixed assumption** about the functional form. The model complexity grows with the data — the training set itself is retained at prediction time.

- More flexible; can fit complex, irregular patterns
- Slower at prediction time (depends on training data)
- Requires more data to generalise well

| Algorithm | What is stored |
|---|---|
| Locally Weighted Linear Regression (LWLR) | Entire training set; $\theta$ recomputed per query |
| k-Nearest Neighbours (k-NN) | Entire training set |
| Kernel SVM | Support vectors |
| Gaussian Process Regression | Training inputs + kernel parameters |
| Decision Trees (deep, unpruned) | Full tree structure grows with data |

---

### Key Differences

| Aspect | Parametric | Non-Parametric |
|---|---|---|
| **Model size** | Fixed ($\theta$ only) | Grows with training data |
| **Training data needed after fit** | No | Yes (instance-based) |
| **Assumptions** | Strong (e.g. linearity) | Weak / none |
| **Flexibility** | Lower | Higher |
| **Prediction speed** | Fast | Slower (scales with $m$) |
| **Risk** | Underfitting if assumptions wrong | Overfitting with small datasets |

## 6. Classification and Logistic Regression

### Key Mathematical Concepts

**Sigmoid (Logistic) Function:** Squashes any real value into $(0, 1)$, interpreted as a probability:
$$h_\theta(x) = \sigma(\theta^T x) = \frac{1}{1 + e^{-\theta^T x}}$$

**Decision Boundary:** The threshold separating predicted classes:
$$\hat{y} = \begin{cases} 1 & \text{if } h_\theta(x) \ge 0.5 \\ 0 & \text{otherwise} \end{cases}$$

**Cost Function (Log-Loss / Binary Cross-Entropy):** Derived from maximum likelihood estimation; penalises confident wrong predictions heavily:
$$J(\theta) = -\frac{1}{m}\sum_{i=1}^{m}\left[y^{(i)}\log h_\theta(x^{(i)}) + (1 - y^{(i)})\log\bigl(1 - h_\theta(x^{(i)})\bigr)\right]$$

**Gradient Update Rule:** Same form as linear regression but $h_\theta(x)$ is the sigmoid output:
$$\theta_j := \theta_j - \alpha \frac{1}{m}\sum_{i=1}^{m}\bigl(h_\theta(x^{(i)}) - y^{(i)}\bigr)x_j^{(i)}$$

### Essential Concepts
- **Sigmoid Function:** Differentiable, monotonic; maps $\mathbb{R} \to (0,1)$; output is the predicted probability $P(y=1 \mid x;\theta)$
- **Log-Loss:** Convex cost function; ensures gradient descent finds a global minimum
- **Maximum Likelihood Estimation (MLE):** Log-loss is derived by maximising the likelihood of observed labels under the Bernoulli distribution
- **Decision Boundary:** Linear in feature space ($\theta^T x = 0$ is the boundary); can be made non-linear with polynomial features
- **Softmax (Multi-class):** Generalises sigmoid to $K$ classes: $P(y=k \mid x) = \frac{e^{\theta_k^T x}}{\sum_{j=1}^K e^{\theta_j^T x}}$
- **Gradient:** $\nabla J(\theta) = \frac{1}{m} X^T (h - y)$ — identical in form to linear regression
- **Regularisation:** L1 / L2 penalty added to $J(\theta)$ to prevent overfitting

### When to Use
- Binary or multi-class classification
- When a probabilistic output is needed
- Linearly (or near-linearly) separable classes

### When NOT to Use
- Complex non-linear decision boundaries without feature engineering
- When classes are highly imbalanced without adjustment

## 7. Batch Gradient Ascent

### Key Mathematical Concepts

**Update Rule:** Instead of minimising a cost, gradient ascent **maximises** an objective $\ell(\theta)$ by stepping in the direction of the gradient:
$$\theta := \theta + \alpha \nabla \ell(\theta)$$

The only difference from gradient descent is the **`+`** sign.

**Log-Likelihood for Logistic Regression:** The canonical use case — maximise the log-likelihood of the training labels:
$$\ell(\theta) = \sum_{i=1}^{m}\left[y^{(i)}\log h_\theta(x^{(i)}) + (1 - y^{(i)})\log\bigl(1 - h_\theta(x^{(i)})\bigr)\right]$$

**Gradient of Log-Likelihood:**
$$\nabla \ell(\theta) = X^T\bigl(y - h_\theta(X)\bigr)$$

### Essential Concepts
- **Ascent vs Descent:** Gradient descent minimises $J(\theta)$; gradient ascent maximises $\ell(\theta)$; equivalent since $J = -\ell$
- **Log-Likelihood:** Objective derived from MLE — measures how well $\theta$ explains the observed labels
- **Same Convergence Properties:** Learning rate, convergence criteria, and batch behaviour are identical to gradient descent
- **Dataset Size:** Uses all $m$ training examples per update (batch)

### When to Use
- Maximising a likelihood or posterior (MLE / MAP estimation)
- Logistic regression trained via maximum likelihood
- Any probabilistic model where the objective is a log-likelihood

### When NOT to Use
- When the objective is a cost / loss to be minimised (use gradient descent instead)

## 8. Newton's Method

### Key Mathematical Concepts

**Update Rule:** Uses both the gradient **and** the Hessian (second-order information) to take a more informed step toward the optimum:
$$\theta := \theta - H^{-1} \nabla_\theta \ell(\theta)$$

**Hessian Matrix:** Matrix of all second-order partial derivatives of the objective:
$$H_{jk} = \frac{\partial^2 \ell(\theta)}{\partial \theta_j \,\partial \theta_k}$$

For logistic regression the Hessian is:
$$H = X^T D X$$

where $D$ is a diagonal matrix with $D_{ii} = h_\theta(x^{(i)})(1 - h_\theta(x^{(i)}))$.

### Essential Concepts
- **Second-Order Optimisation:** Gradient descent uses only $\nabla \ell$ (first derivative); Newton's method also uses $H$ (second derivative) to account for curvature
- **Hessian Matrix ($H$):** Captures the curvature of the objective; must be computed and inverted at every iteration — $O(n^3)$ cost
- **No Learning Rate:** Step size is determined automatically by $H^{-1}$; no $\alpha$ hyperparameter needed
- **Quadratic Convergence:** Converges much faster than gradient descent near the optimum (quadratic vs linear rate)
- **Newton-Raphson:** The scalar version; Newton's method generalises it to vector-valued $\theta$
- **Positive Definiteness:** Requires $H$ to be negative definite (for maximisation) to guarantee a descent/ascent step

### When to Use
- Small to medium feature sets ($n$ not too large — Hessian inversion is $O(n^3)$)
- When fast convergence is needed and few iterations are preferred
- Logistic regression and GLMs (Hessian has a closed form)

### When NOT to Use
- Large number of features (Hessian inversion becomes prohibitively expensive)
- Non-convex objectives (may converge to saddle points or diverge)
- Deep learning (use quasi-Newton methods like L-BFGS instead)
